### Description

This is a chronic kidney disease dataset that can help us predict chronic kidney disease in patients based on symptoms. 

The `unique_id` column refers to a patient. The `Target` which has values of either `ckd` or `notckd` is a boolean label telling us whether the patient contracted chronic kidney disease or not. The rest of the columns are the features and the complete dataset has both categorical and numerical features. The ones in the `numerical` table above are numerical features and the ones in the `categorical` table above are categorical features. 

For example, `age`, `bp`, and `bgr` from `numerical` are numerical features, and `pcc`, `htn`, and `dm` from `categorical` are categorical features. 



# Part 3: Data Preprocessing

## Part 3.1 : Drop Duplicate rows

````python
# Checking for Duplicates
num_dupe = numerical["unique_id"].duplicated().sum()
cat_dupe = categorical["unique_id"].duplicated().sum()


# Drop these duplicate rows
numerical = numerical.drop_duplicates(subset=['unique_id'])
categorical = categorical.drop_duplicates(subset=['unique_id'])


print("Duplicates in Numerican, Categorical:")
print(cat_dupe, num_dupe)
print("New Lengths of Numerical, Categorical:")
print(len(numerical), len(categorical))
````

## Part 3.2: Combine two differents datasets

All `unique_id`'s are not present across both tables because `categorical` has 400 rows whereas `numercial` has 390, after removing duplicate rows. This means that `numerical` is missing 10 rows, and if we joined the two tables, there would be 10 `unique_id`'s that have catagorical features but don't have numerical features. 

````python
# Merge numerical and categorical tables based on unique_id

features = pd.merge(numerical, categorical, on="unique_id") 
features
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Percentage of rows that contain atleast one missing value
missing = features[features.isnull().any(axis=1)]
len(missing)/len(features) * 100
````

````python
features.dropna(how='any', inplace=True)
features
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to 'Target'
features = features.sort_values(by='Target')
features.reset_index(drop=True, inplace=True)
features
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.

````python
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']

df_encoded = pd.get_dummies(features, columns=categorical_columns, dtype=int)

df_encoded.columns.tolist()
````
