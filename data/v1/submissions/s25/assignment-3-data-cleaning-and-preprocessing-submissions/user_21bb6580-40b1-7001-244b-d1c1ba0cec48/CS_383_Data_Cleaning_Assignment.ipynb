## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows
duplicate_rows_numerical = numerical_data.duplicated(subset=['unique_id']).sum()
print(f"Number of duplicate rows in numerical dataset: {duplicate_rows_numerical}")

# Drop these duplicate rows
numerical_data_no_duplicates = numerical_data.drop_duplicates(subset=['unique_id'], keep='first')
print(f"Shape of numerical dataset after removing duplicates: {numerical_data_no_duplicates.shape}")

# Repeat the same for categorical dataset. Print the duplicate rows and drop them
duplicate_rows_categorical = categorical_data.duplicated(subset=['unique_id']).sum()
print(f"Number of duplicate rows in categorical dataset: {duplicate_rows_categorical}")

# Drop these duplicate rows
categorical_data_no_duplicates = categorical_data.drop_duplicates(subset=['unique_id'], keep='first')
print(f"Shape of categorical dataset after removing duplicates: {categorical_data_no_duplicates.shape}")

````

## Part 3.2: Combine two differents datasets

A good skill to have is to know how to combine 2 different datasets.

Are all the unique ids are present in both datasets? Why do you think so? If not, what do the rows that are missing from one of the datasets look like in the combined table?

Answer: Not all unique IDs are present in both datasets. This is likely because some patients may have completed only part of their medical tests or examinations. When examining the merged data, we can see that some IDs appear only in the numerical dataset while others appear only in the categorical dataset. In the combined table, rows with missing IDs from one dataset will have null values for all columns from that missing dataset. This creates incomplete patient profiles where either all categorical or all numerical information is missing. These partial records highlight the challenge of working with real-world medical data, where complete information isn't always available for every patient. This pattern of missingness needs to be addressed during data cleaning to ensure our model works with complete and reliable information.


````python
# Check if all unique IDs are present in both datasets
numerical_ids = set(numerical_data_no_duplicates['unique_id'])
categorical_ids = set(categorical_data_no_duplicates['unique_id'])

# Find IDs that are in numerical but not in categorical
only_in_numerical = numerical_ids - categorical_ids
print(f"Number of IDs only in numerical dataset: {len(only_in_numerical)}")
if len(only_in_numerical) > 0:
    print(f"Example IDs only in numerical: {list(only_in_numerical)[:5]}")

# Find IDs that are in categorical but not in numerical
only_in_categorical = categorical_ids - numerical_ids
print(f"Number of IDs only in categorical dataset: {len(only_in_categorical)}")
if len(only_in_categorical) > 0:
    print(f"Example IDs only in categorical: {list(only_in_categorical)[:5]}")

# Find common IDs
common_ids = numerical_ids.intersection(categorical_ids)
print(f"Number of common IDs: {len(common_ids)}")

# Merge the two datasets based on unique_id
combined_data = pd.merge(numerical_data_no_duplicates, categorical_data_no_duplicates, 
                         on='unique_id', how='inner')

# Print the shape of the combined dataset
print(f"Shape of combined dataset: {combined_data.shape}")

# Print the first few rows of the combined dataset
print("\nCombined dataset preview:")
print(combined_data.head())
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain at least one missing value
rows_with_missing = combined_data.isna().any(axis=1).sum()
total_rows = combined_data.shape[0]
percentage_missing = (rows_with_missing / total_rows) * 100

# Print percentage
print(f"Number of rows with at least one missing value: {rows_with_missing}")
print(f"Percentage of rows with at least one missing value: {percentage_missing:.2f}%")

# Drop these rows from the dataset
cleaned_data = combined_data.dropna()

# Print the Dataset
print(f"\nShape of dataset after dropping rows with missing values: {cleaned_data.shape}")
print(f"Number of rows removed: {total_rows - cleaned_data.shape[0]}")
print("\nCleaned dataset preview:")
print(cleaned_data.head())
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting
sorted_data = cleaned_data.sort_values(by='Target').reset_index(drop=True)

# Print the dataset
print("Shape of sorted dataset:", sorted_data.shape)
print("\nSorted dataset preview:")
print(sorted_data.head())

# Check the distribution of Target values
print("\nTarget value counts:")
print(sorted_data['Target'].value_counts())
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']

# Apply one-hot encoding to categorical columns
encoded_data = pd.get_dummies(sorted_data, columns=categorical_columns, drop_first=False)

# Print the dataset
print("Shape of encoded dataset:", encoded_data.shape)
print("\nEncoded dataset preview:")
print(encoded_data.head())

# Check the new columns created by encoding
print("\nNew columns after encoding:")
categorical_encoded_cols = [col for col in encoded_data.columns if col not in sorted_data.columns]
print(categorical_encoded_cols)
````

In the example we went through above, another solution is to have a single column for the binary variable. In the downstream modeling would this be equivalent? What effect would this have if the categorical variable could take more than 2 values? For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

Answer: Using a single column for binary variables (like 0/1 encoding) is equivalent to one-hot encoding in downstream modeling only when the variable has exactly two categories. However, this approach becomes problematic with multi-valued categorical variables. For a "type of condiment" feature with 5 values (like ketchup, mustard, mayo, relish, and hot sauce), using a single numeric column would incorrectly impose an ordinal relationship between condiments. This would mislead the model into thinking that condiment "4" is somehow more similar to condiment "5" than to condiment "1", when no such relationship exists. One-hot encoding creates separate binary columns for each condiment type, allowing the model to learn the unique effect of each condiment on sandwich ratings without imposing artificial ordering. This prevents the model from making incorrect assumptions about relationships between different categorical values, leading to more accurate predictions of sandwich ratings.


## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

# Create a copy of the data to avoid modifying the original
data_no_outliers = encoded_data.copy()

# Define a function to identify outliers (values more than 3 standard deviations from the mean)
def remove_outliers(df, column):
    mean = df[column].mean()
    std = df[column].std()
    outlier_threshold = 3 * std
    
    # Create a mask for non-outlier values
    mask = (df[column] >= mean - outlier_threshold) & (df[column] <= mean + outlier_threshold)
    return mask

# Apply the outlier removal to each numerical column
for column in numerical_columns:
    mask = remove_outliers(data_no_outliers, column)
    data_no_outliers = data_no_outliers[mask]

# Print the dataset
print(f"Shape of dataset before removing outliers: {encoded_data.shape}")
print(f"Shape of dataset after removing outliers: {data_no_outliers.shape}")
print(f"Number of rows removed: {encoded_data.shape[0] - data_no_outliers.shape[0]}")
print("\nDataset preview after removing outliers:")
print(data_no_outliers.head())
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
from sklearn.preprocessing import MinMaxScaler

# Create a scaler object
scaler = MinMaxScaler()

# Create a copy of the data to avoid modifying the original
normalized_data = data_no_outliers.copy()

# Apply normalization to numerical columns
normalized_data[numerical_columns] = scaler.fit_transform(normalized_data[numerical_columns])

# Print the dataset
print("Shape of normalized dataset:", normalized_data.shape)
print("\nNormalized dataset preview:")
print(normalized_data.head())

# Compare before and after normalization for a few columns
print("\nBefore normalization (first 5 rows):")
print(data_no_outliers[numerical_columns].head())
print("\nAfter normalization (first 5 rows):")
print(normalized_data[numerical_columns].head())

# Check the range of values in normalized columns
print("\nRange of values in normalized columns:")
for column in numerical_columns:
    print(f"{column}: Min = {normalized_data[column].min()}, Max = {normalized_data[column].max()}")
````
