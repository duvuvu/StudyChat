## Part 3.1 : Drop Duplicate rows

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value
total_rows = len(combined_df)
num_rows_with_missing = combined_df.isnull().any(axis = 1).sum()

rows_missing_percentage = (num_rows_with_missing / total_rows) * 100

# Print %
print(f"Percentage of rows with at least one missing value: {rows_missing_percentage}%")
# Drop these rows from the dataset
cleaned_combined_df = combined_df.dropna()

# Print the Dataset
print("Cleaned, combined, dataset: \n", cleaned_combined_df)
````

In the example we went through above, another solution is to have a single column for the binary variable. In the downstream modeling would this be equivalent? What effect would this have if the categorical variable could take more than 2 values? For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

Answer: When encoding categorical variables for modeling, the approach differs significantly between binary and multi-class categories. For binary variables (e.g., "Yes" and "No"), using a single column is effective and straightforward, allowing models to interpret the variable correctly as two distinct states, often encoded as `0` and `1`. However, for multi-class categorical variables, such as a "type of condiment" that can take multiple values (e.g., "Ketchup," "Mustard," "Mayo," "Hot Sauce," "Relish"), using a single column can be problematic as it may lead to a misinterpretation of categories as ordinal, potentially skewing model predictions. Instead, one-hot encoding is the preferred method for handling such multi-class categories, as it creates separate binary columns for each category, allowing the model to treat them independently without implying any ordinal relationship. While one-hot encoding increases dimensionality and the risk of multicollinearity in certain types of models, it provides a clearer representation of categorical data that enhances model performance and interpretability.

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
def remove_outliers(df, column):
    mean = df[column].mean()  
    std = df[column].std()  
    lower_bound = mean - 3 * std  
    upper_bound = mean + 3 * std 
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

# Apply the function to each numerical column
for column in numerical_columns:
    cleaned_combined_df = remove_outliers(cleaned_combined_df, column)

# Print the cleaned dataset
print("Dataset after removing outliers:")
print(cleaned_combined_df)
# Print the dataset
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python

scaler = MinMaxScaler()


cleaned_combined_df[numerical_columns] = scaler.fit_transform(cleaned_combined_df[numerical_columns])

# Print the normalized dataset
print("Dataset after normalization:")
print(cleaned_combined_df.head())
````

To do this, start by providing clear and specific instructions, including examples of the desired output format to reduce ambiguity.
