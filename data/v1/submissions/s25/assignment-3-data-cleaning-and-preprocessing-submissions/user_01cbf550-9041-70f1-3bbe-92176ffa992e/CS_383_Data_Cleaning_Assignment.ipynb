## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows
num_dupes_size = len(numerical_data[numerical_data.duplicated()])
print(f'Number of duplicate rows in numerical_data: {num_dupes_size}')

# Drop these duplicate rows
numerical_data_deduped = numerical_data.drop_duplicates()

# Repeat the same for categorical dataset. Print the duplicate rows and drop them
cat_dupes_size = len(categorical_data[categorical_data.duplicated()])
print(f'Number of duplicate rows in categorical_data: {cat_dupes_size}')
categorical_data_deduped = categorical_data.drop_duplicates()
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value
percentage_with_atleast_one_missing = merged.isnull().any(axis=1).mean() * 100
# Print %
print(f'Percentage of rows that contain at least one missing value: {percentage_with_atleast_one_missing}')
# Drop these rows from the dataset
merged_without_nan = merged.dropna()
# Print the Dataset
print(merged_without_nan.columns)
merged_without_nan
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting
sorted_data = merged_without_nan.sort_values(by='Target').reset_index(drop=True)
# Print the dataset
sorted_data
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']
sorted_with_dummy_columns = pd.get_dummies(sorted_data, columns=categorical_columns, dtype=int)
# Print the dataset
sorted_with_dummy_columns
````

For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

Answer: 

This would be equivalent in downstream modelling if the single column had only 2 values (binary). This is because splitting it into 2 columns versus having only two options for values is both 2 dimensions, meaning it would have the same effect.

However, when adding more than 2 values, like the condiment example, this changes. Keeping those 5 different values within one column might cause later modelling to interpret that column as a numerical column rather than a categorical one since there would be more that 2 different values to analyze. ## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
means = sorted_with_dummy_columns[numerical_columns].mean()
std_devs = sorted_with_dummy_columns[numerical_columns].std()

mask = (sorted_with_dummy_columns[numerical_columns] <= means + (3 * std_devs)) & \
(sorted_with_dummy_columns[numerical_columns] >= means - (3 * std_devs))

mask = mask.all(axis=1)
sorted_without_outliers = sorted_with_dummy_columns.loc[mask, :].reset_index(drop=True)

sorted_without_outliers
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
normalized = sorted_without_outliers.copy()
normalized[numerical_columns] = (normalized[numerical_columns] - normalized[numerical_columns].min()) / (normalized[numerical_columns].max() - normalized[numerical_columns].min())
# Print the dataset
normalized
````
