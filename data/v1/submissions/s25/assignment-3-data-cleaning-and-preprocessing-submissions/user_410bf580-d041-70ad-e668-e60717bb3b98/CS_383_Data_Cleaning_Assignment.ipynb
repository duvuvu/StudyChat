## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows
duplicates = numerical.duplicated(subset=['unique_id'])
num_duplicates_numerical = duplicates.sum()
print(f"Number of duplicate rows in numerical dataset: {num_duplicates_numerical}")
# Drop these duplicate rows
print(f"Number of rows of numerical chart before dropping duplicates: {len(numerical)}")
numerical_cleaned = numerical.drop_duplicates(subset=['unique_id'])
print(f"Number of rows of numerical chart after dropping duplicates: {len(numerical_cleaned)}")

# Repeat the same for categorical dataset. Print the duplicate rows and drop them
duplicates = categorical.duplicated(subset=['unique_id'])
num_duplicates_categorical = duplicates.sum()
print(f"Number of duplicate rows in categorical dataset: {num_duplicates_categorical}")
# Drop these duplicate rows
print(f"Number of rows of categorical chart before dropping duplicates: {len(categorical)}")
categorical_cleaned = categorical.drop_duplicates(subset=['unique_id'])
print(f"Number of rows of categorical chart after dropping duplicates: {len(categorical_cleaned)}")

````

## Part 3.2: Combine two differents datasets

A good skill to have is to know how to combine 2 different datasets.

Are all the unique ids are present in both datasets? Why do you think so? If not, what do the rows that are missing from one of the datasets look like in the combined table?

Answer: No, since two datasets have different number of rows, so there will be ids not present in both datasets. The rows are missing from one would have NaN in features they do not have.

````python
# Merge the two given numerical and categorical datasets based on their unique_ID.
merged_chart = pd.merge(numerical, categorical, on='unique_id', how='outer')

# 'inner' – keeps only rows with matching 'unique_id' in both DataFrames
# 'outer' – keeps all rows from both DataFrames, fills with NaN where there are no matches
# 'left' – keeps all rows from df1, including all rows from df2 that match
# 'right' – keeps all rows from df2, including all rows from df1 that match

#Print the combined dataset
print(merged_chart)
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value

rows_with_nan = merged_chart.isna().any(axis=1)  # Create a boolean Series
num_rows_with_nan = rows_with_nan.sum()  # Count True values

print(f"\nNumber of rows containing at least one NaN: {num_rows_with_nan}")
num_rows = len(merged_chart)

# Print %
percent_NaN = 100*(num_rows_with_nan)/num_rows
print(f"\nPercentage of rows with missing values: {percent_NaN}%")

# Drop these rows from the dataset
merged_chart_cleaned = merged_chart.dropna()


# Print the Dataset
print("\nCleaned DataFrame (without NaN):")
print(merged_chart_cleaned)
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting
merged_chart_cleaned_sorted = merged_chart_cleaned.sort_values(by='Target')

# Step 2: Reset the index after sorting
merged_chart_cleaned_sorted.reset_index(drop=True, inplace=True)

# Print the dataset
print("\nSorted DataFrame (by 'Target' column):")
print(merged_chart_cleaned_sorted)
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']

merged_chart_cleaned_sorted_with_dummies = pd.get_dummies(merged_chart_cleaned_sorted, columns=categorical_columns)
# If the dummy values are still boolean, ensuring conversion to integers
merged_chart_cleaned_sorted_with_dummies = merged_chart_cleaned_sorted_with_dummies.astype(int)

# Print the dataset
print(merged_chart_cleaned_sorted_with_dummies)
````

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

merged_predropped = merged_chart_cleaned_sorted_with_dummies

mean = merged_predropped[numerical_columns].mean()
std_dev = merged_predropped[numerical_columns].std()

# Define the upper and lower bounds for outlier detection
upper_limit = mean + 3 * std_dev
lower_limit = mean - 3 * std_dev

# Filter out rows with outliers
# Create a boolean condition for retaining rows within the limits
condition = (merged_predropped[numerical_columns] <= upper_limit) & (merged_predropped[numerical_columns] >= lower_limit)

# Use .all(axis=1) to ensure all specified columns meet the condition for each row
merged_dropped_outlier = merged_predropped[condition.all(axis=1)]
merged_dropped_outlier = merged_dropped_outlier.reset_index()
# Print the dataset
print(merged_dropped_outlier)
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
scaler = MinMaxScaler()
normalized = merged_dropped_outlier
normalized[numerical_columns] = scaler.fit_transform(merged_dropped_outlier[numerical_columns])

# Print the dataset
print(normalized)

````
