## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows
numerical_duplicates = numerical_df[numerical_df.duplicated()]
print(f"Total number of duplicate rows: {numerical_duplicates.shape[0]}")

# Drop these duplicate rows
numerical_df = numerical_df.drop_duplicates()

# Repeat the same for categorical dataset. Print the duplicate rows and drop them
categorical_duplicates = categorical_df[categorical_df.duplicated()]
print(f"Total number of duplicate rows: {categorical_duplicates.shape[0]}")

# Drop duplicate rows from the categorical dataset
categorical_df = categorical_df.drop_duplicates()
````

## Part 3.2: Combine two differents datasets

A good skill to have is to know how to combine 2 different datasets.

Are all the unique ids are present in both datasets? Why do you think so? If not, what do the rows that are missing from one of the datasets look like in the combined table?

````python
#  Testing to see if ids are present in both datasets
# Assuming the unique ID column is named 'id'
unique_ids_cat = set(categorical_df['unique_id'])
unique_ids_num = set(numerical_df['unique_id'])

print("Number of Unique Ids in categorical dataset")
print(len(unique_ids_cat))
print("Number of Unique Ids in numerical dataset")
print(len(unique_ids_num))

missing_in_num = unique_ids_cat - unique_ids_num
print("The exact missing unique ids:")
print(missing_in_num)

# Outer Merge:
outer_combined_df = numerical_df.merge(categorical_df, left_on='unique_id', right_on='unique_id', how="outer")

print("Example of a row that is missing from one of the datasets looks like in the combined table if OUTER JOIN is used:")
print(outer_combined_df.loc[outer_combined_df['unique_id'] == 208034])


# Inner Merge:
inner_combined_df = numerical_df.merge(categorical_df, left_on='unique_id', right_on='unique_id', how="inner")
print("Example of a row that is missing from one of the datasets looks like in the combined table if INNER JOIN is used:")
print(inner_combined_df.loc[inner_combined_df['unique_id'] == 208034])


````

Answer: 
As you can see in the code above, the categorical dataset has 10 more unique ids than the numerical dataset. Therefore, all unique ids are not present in both datasets. We calculated this by counting the total number of rows in the unique_id columns of both datasets. 

What the rows missing from one of the datasets looks like in the combined dataset depends on how the datasets are merged, i.e. what join operation is used. 
If we use outer join: then the data of the missing unique id is still added to the table but since it has no numerical values (all missing ids are present in cat and NOT in numerical), it has NaN value for all numerical columns. 

If we use inner join: the rows for the missing unique ids are not added to table itself, that's why it returns a empty dataframe if particualrly searched for. 


For the purpose of clean, complete, and ocnsistent data we will proeed with an inner merge.



````python
# Merge the two given numerical and categorical datasets based on their unique_ID.
combined_df = numerical_df.merge(categorical_df, left_on='unique_id', right_on='unique_id', how="inner")

#Print the combined dataset
print(combined_df)
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value
missing_percentage = combined_df.isnull().any(axis=1).mean() * 100
# Print %
print(missing_percentage)

# Drop these rows from the dataset
cleaned_df = combined_df.dropna()

# Print the Dataset
print(cleaned_df)
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting
sorted_df = cleaned_df.sort_values(by=['Target'])

# Reset Index
sorted_df.reset_index(drop=True, inplace=True)

# Print the dataset
print(sorted_df)
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']

# get_dummies to sepereate the columns
dummy_df = pd.get_dummies(sorted_df,columns=categorical_columns, prefix=categorical_columns,)

# changing the boolean values to binary
dummy_columns = [col for col in dummy_df.columns if '_' in col and col!="unique_id"]
dummy_df[dummy_columns] = dummy_df[dummy_columns].astype(int)

# Print the dataset
print(dummy_df)
````

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

nooutlier_df = dummy_df.copy()

for col in numerical_columns:
    mean = dummy_df[col].mean()
    std_dev = dummy_df[col].std()

    threshold_upper = mean + (3 * std_dev)
    threshold_lower = mean - (3 * std_dev)

    nooutlier_df = nooutlier_df[(nooutlier_df[col] <= threshold_upper) & (nooutlier_df[col] >= threshold_lower)]

# restting index
nooutlier_df.reset_index(drop=True, inplace=True)

# Print the dataset
print(nooutlier_df)
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
df_min_max_scaled = nooutlier_df.copy()

for column in numerical_columns:
    df_min_max_scaled[column] = (df_min_max_scaled[column] - df_min_max_scaled[column].min()) / (df_min_max_scaled[column].max() - df_min_max_scaled[column].min())

# Print the dataset
print(df_min_max_scaled)

````

````python
# Code to rename all the columns in the dataset
print("Initial column names")
print(df.columns)

# Column name mapping
abbreviation_mapping = {
    'age': 'age',
    'bp': 'blood pressure',
    'sg': 'specific gravity',
    'al': 'albumin',
    'su': 'sugar',
    'rbc': 'red blood cells',
    'pc': 'pus cell',
    'pcc': 'pus cell clumps',
    'ba': 'bacteria',
    'bgr': 'blood glucose random',
    'bu': 'blood urea',
    'sc': 'serum creatinine',
    'sod': 'sodium',
    'pot': 'potassium',
    'hemo': 'hemoglobin',
    'pcv': 'packed cell volume',
    'wc': 'white blood cell count',
    'rc': 'red blood cell count',
    'htn': 'hypertension',
    'dm': 'diabetes mellitus',
    'cad': 'coronary artery disease',
    'appet': 'appetite',
    'pe': 'pedal edema',
    'ane': 'anemia',
    'class': 'class'
}

# Function to rename columns
def rename_columns(column_name):
    for abbr, full_name in abbreviation_mapping.items():
        if column_name.startswith(abbr):
            # Replace the abbreviation with the full name, while preserving the suffix
            return column_name.replace(abbr, full_name, 1)  # Replace only the first occurrence
    return column_name

# Rename the columns in the DataFrame
df.columns = [rename_columns(col) for col in df.columns]

# Display the renamed columns
print("Renamed Columns:")
print(df.columns)
````
