```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score, KFold
from sklearn.linear_model import LinearRegression
import numpy as np
from sklearn.preprocessing import PolynomialFeatures



```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset
dataset = pd.read_csv("science_data_large.csv")
# Output the first 15 rows of the data
dataset.head(15)
```

```python
# Display a summary of the table information (number of datapoints, etc.)
dataset.describe(include='all')
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = dataset.drop('Size nm^3', axis=1)
y = dataset['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=5)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression().fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
prediction = model.predict(np.array([[827,341]]))
print(prediction)
```

```python
# Report on the score for that model, in your own words (markdown, not code) explain what the score means
model.score(X_test, y_test)
```

This score represents how well the linear model fits the test dataset. For example, this score of 0.89, means that the model will accurately predict the label of the test data 89% of the time. The closer this number is to 1, the better the linear regression fit, basically implying that the model will always predict the correct answer. Alas, in this context the linear model correctly labels the test data 89% of the time. Also, the score represents the $R^2$ value.

```python
# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
print(model.coef_, model.intercept_)
```

My equation: $y = 869.62x_1+1029.26x_2-414313.64$

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
kf = KFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X, y, cv=kf)
print(scores, "\n")
print(scores.mean(), "\n")
print(scores.std())
```

Report on their finding and their significance: 
Each score represents the accuracy of the model on producing accurate labels based on 5 different partitions of data. This cross validation displays the model's capacity to accurately predict labels for different partitions of the datset. Thus each element in the printed array scores, represents a score for a model on accurately representing that dataset. The cross_val_score method allows for a more dynamic interpreation of the models strength based on different partitions of training and test data. The average score is ~85% accuracy wth a standard deviation of about 1.8%. The more consistant the model's scores are, the smaller the standard deviation, the better the model is. Therefore it is a useful metric to look at the standard deviation of the scores so see how universally accurate the model is. 

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
x_poly = PolynomialFeatures(degree=2).fit_transform(X)
polyModel = LinearRegression().fit(x_poly, y)
# Report on the metrics and output the resultant equation as you did in Part 3.
polyModel.score(PolynomialFeatures(degree=2).fit_transform(X_test), y_test)
```

An $R^2$ value of 1, implies that the polynomial regression perfectly fits the data. It will accurately label the test data 100% of the time. 

The resultant equation is

$y = 12x - x^2 - x^3 + 2x^4 + 0.0286 x^5$

```python
print(polyModel.coef_, polyModel.intercept_ )
```
