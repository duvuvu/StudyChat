```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Using pandas load the dataset (load remotely, not locally)
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
df = pd.read_csv('science_data_large.csv')
print(df.head(15))
print(df.info())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df.iloc[:, :-1]
y = df.iloc[:, -1]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = pd.DataFrame([[523, 764]], columns=['Temperature °C', 'Mols KCL'])
sample_prediction = model.predict(sample_data)
print(sample_prediction)

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
score = model.score(X_test, y_test)
print(score)

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_
print(coefficients, intercept)
```

The score explains how well the model explains the variance in the data. If the score is 1 then the model can perfectly predict a variable. If it is 0 then the model cannot explain any varaiablity so it will just give the mean. If it is negative then the model is completely inaccurate to the point it gives a worse result than the mean. In this case we get a 0.8552472077276095 which is inbetween 0 and 1, so it can explain around 85% of the variablity of the model.

Sample equation: $E = mc^2$

$h(x) = -409391.47958340764 + (866.14641337)T + (1032.69506649)Mols$

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
scores = cross_val_score(model, X, y, cv=20)

# Report on their finding and their significance
print(scores)
meanScore = sum(scores)/ 20

standardDeviationScore = (sum(((i - meanScore)**2) for i in scores)/20)**0.5

print(meanScore, standardDeviationScore)
```

The mean score I got was 85.17% variabilty and the scores had a very small standard deviation of 3.95%. So our current model can pretty accurately predict 85% of the variabilty of the model.

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.1, random_state=42)

model_poly = LinearRegression()
model_poly.fit(X_train, y_train)

poly_score = model_poly.score(X_test, y_test)
print(poly_score)

coefficients = model_poly.coef_
intercept = model_poly.intercept_

# Report on the metrics and output the resultant equation as you did in Part 3.
print(coefficients, intercept)
```
