```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
path = '/workspaces/assignment-4-equation-of-a-slime-<redacted>/science_data_large.csv'
data = pd.read_csv(path)

# Output the first 15 rows of the data
print(data.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print(data.info())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data[['Temperature °C', 'Mols KCL']]
y = data['Size nm^3'] 

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1)

#confirming splits
print("Training set size (X_train):", X_train.shape)
print("Test set size (X_test):", X_test.shape)
print("Training labels size (y_train):", y_train.shape)
print("Test labels size (y_test):", y_test.shape)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report on the score for that model, in your own words (markdown, not code) explain what the score means

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX

# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)


# Create a sample datapoint and predict the output of that sample with the trained model
sample = pd.DataFrame([[500, 600]], columns=['Temperature °C', 'Mols KCL'])
predicted_size = model.predict(sample)
print(predicted_size)

#sample was 500 temp, 600 mol, output ranging between 635000 to 645000

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
score = model.score(X_test, y_test)
print(f"model R-squared score: {score}")

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

The score ranges between 0.85 to 0.90 when run again. The function .score outputs the R^2 score, which basically represents "The proportion of the variance in the target variable that is predictable from the feature variables" (temp and mol). Additionally, the score ranges from 0 - 1. When the score is 1 it means that the model perfectly predicts the outcomes, when the score is 0 it means that the model predicts the outcomes the same as the mean of the data, and when the score is negative  it means that the model is worse than predicting the mean (horizontal line). In my case, my R^2 value is pretty good, it is at 0.85-0.9 which shows that our linear regression model is a fairly good fit for our data.

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
cross_val_scores = cross_val_score(model, X, y, cv=10)
print(cross_val_scores)

# Report on their finding and their significance
mean_score = cross_val_scores.mean()
std_score = cross_val_scores.std()

print(mean_score)
print(std_score)
# Report on their finding and their significance

```

# Report on their finding and their significance
The cross-valudation score shows how well the lin regression works on different splits on the data, the scores show the proportion of the variance in the target variable. Values with 1 indicate a better fit. The mean cross-validation score is 85.5% which is pretty good. The significance of this is that on average, the model can explain 85.5% of the variance in the target variable (size) based on the features. Hence, i'd say it's a good fit. The standard deviation was approximately 0.0315, this value is pretty low. It shows that the model is quite consistent across different splits of data and hence the model can generalize pretty well and is not overly sensitive to specific parts of the dataset. The low st. dev shows that the model is not heavily overfitting or underfitting the data. Also, the cv parameter in teh cross_val_score function stands for eg. cv = 10, 10-fold cross validation. This basically means that the dataset is split into 10 equal sized subsets. Hence, the model is trained and tested 10 times, each time using a different fold as the test set. 

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)
X_poly_train = poly.fit_transform(X_train)
X_poly_test = poly.transform(X_test)

poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_train)

y_poly_pred = poly_model.predict(X_poly_test)
poly_r2_score = r2_score(y_test, y_poly_pred)

poly_coefficients = poly_model.coef_
poly_intercept = poly_model.intercept_

print(poly_r2_score, poly_coefficients, poly_intercept)

cross_val_scores_poly = cross_val_score(poly_model, poly.fit_transform(X), y, cv=10)

mean_poly_score = cross_val_scores_poly.mean()
std_poly_score = cross_val_scores_poly.std()

(mean_poly_score, std_poly_score)
```
