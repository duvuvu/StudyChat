```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

url = "https://raw.githubusercontent.com/COMPSCI-383-Fall2024/assignment-4-equation-of-a-slime-<redacted>/refs/heads/main/science_data_large.csv?token=GHSAT0AAAAAACXIVX5R3S77TCZ6YZB2A5HGZYQP74Q"
data = pd.read_csv(url)

print("First 15 rows of the dataset:")
print(data.head(15))

print("\nSummary of the dataset:")
print(data.info())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

X = data[['Temperature °C', 'Mols KCL']]
y = data['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = [[420, 365]]
predicted_output = model.predict(sample_data)
print(f"Predicted output for the sample data {sample_data}: {predicted_output[0]}")

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
model_score = model.score(X_test, y_test)
print(f"Model R^2 score: {model_score}")

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_
print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

The score reported by the LinearRegression model is the R^2 score, which is the coefficient of determination. This score measures how well the model explains the variance in the target variable, which is: Size nm^3. An R^2 score of 1 indicates a perfect fit, meaning the model explains 100% of the variance in the data. A score of 0 means the model does not explain any variance beyond what the mean of the target would explain. For our model, a higher R² score means the features (Temperature °C, Mols KCL nm^3) can accurately predict the target (Size nm^3).

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
cross_val_scores = cross_val_score(model, X, y, cv=10, scoring='r2')
# Using a (90% train, 10% test) split in each fold
# On each iteration, 1 fold is test and remaining 9 folds are for training the model

# Report on their finding and their significance
print(f"Cross-validation R² scores: {cross_val_scores}") # Shows the R^2 values for each fold
print(f"Mean R² score: {cross_val_scores.mean()}") # Displays the average R^2 score from all folds, indicating overall model performance
print(f"Standard deviation of R² scores: {cross_val_scores.std()}") # Indicates the variability in model performance across different folds
print(f"Minimum R² score: {cross_val_scores.min()}") # Identifies the lowest R^2 score, revealing the worst performance among the folds
```

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Do 90% train, 10% split again
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.1, random_state=42)

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)
coefficients = model.coef_
intercept = model.intercept_

# Report on the metrics and output the resultant equation as you did in Part 3.
print(f"Polynomial Regression R² score: {r2}")
print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

The polynomial regression model achieved an R^2 score of 1.0, meaning that it perfectly explains the variance in the target variable. This suggests that the model captures all underlying patterns in the data, and all the data points are explained accurately. This model is further correct because when we switch to a degree of 1 (not shown in code), the output matches the linear regression model, which is accurate.
