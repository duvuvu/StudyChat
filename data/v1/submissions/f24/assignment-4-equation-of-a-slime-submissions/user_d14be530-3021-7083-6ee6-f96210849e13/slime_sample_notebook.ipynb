```python
# Imports section
import pandas as pd
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

data = pd.read_csv('C:\\<redacted>\\383\\Assignment2\\assignment-4-equation-of-a-slime-<redacted>\science_data_large.csv')

print(data.head(15))

data.info()
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
from sklearn.model_selection import train_test_split

X = data[['Temperature °C', 'Mols KCL']]
y = data[['Size nm^3']]

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
from sklearn.linear_model import LinearRegression
import numpy as np

linear_reg = LinearRegression()

linear_reg.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model

sample_data = np.array([[400,600]])

predicted_size = linear_reg.predict(sample_data)

print(f"Predicted size for 400°C, 600 Mols KCL: {predicted_size[0]}")

# Report on the score for that model, in your own words (markdown, not code) explain what the score means

score = linear_reg.score(X_test, y_test)
print(f"Model score: {score}")

coefficients = linear_reg.coef_.flatten()  # Flatten the coefficients to ensure it's a 1D array
intercept = linear_reg.intercept_.flatten()  # Flatten the intercept if needed

# Print the coefficients and intercept
print("Coefficients:", coefficients)
print("Intercept:", intercept)

# Ensure there are two coefficients (one for each feature)
if len(coefficients) == 2:  
    temperature_coef = coefficients[0]
    kcl_coef = coefficients[1]

    # Print the linear regression equation
    print(f"h(x) = {temperature_coef:.5f} * Temperature + {kcl_coef:.5f} * Mols KCL + {intercept[0]:.5f}")
else:
    print("The number of coefficients is not what was expected. Please check the model.")
```

### Report on the score for the Linear Regression model

The **model score** is a measure of how well the Linear Regression model fits the data. Specifically, it represents the coefficient of determination, \(R^2\), which indicates the proportion of the variance in the dependent variable (slime size) that is predictable from the independent variables (temperature and Mols KCL).

In this case, the model score is **0.8552**, meaning that approximately 85.5% of the variance in the size of the slime can be explained by the temperature and the amount of Potassium Chloride (KCL). The remaining 14.5% is due to factors not captured by the model, such as noise or other unaccounted variables. This is generally a good score, indicating a strong linear relationship between the input variables and the output (slime size).

The linear regression equation is:

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(linear_reg, X, y, cv=5)

# Report on their finding and their 
print(f"Cross-validation scores: {cv_scores}")
print(f"Average cross-validation score: {cv_scores.mean()}")
```
