```python
# Imports section
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
url = './science_data_large.csv'
df = pd.read_csv(url)
print(df.head(15))
print(df.info())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df.drop(columns=['Size nm^3'])
y = df['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
# Create a sample datapoint and predict the output of that sample with the trained model
model.fit(X_train, y_train)
sample = df.iloc[0].drop('Size nm^3').values.reshape(1, -1)
print(sample)
#TODO: ask whether i should use the test set or the training set for the sample
prediction = model.predict(sample)
print('Prediction:', prediction)
# Report on the score for that model, in your own words (markdown, not code) explain what the score means
modelScore = model.score(X_test, y_test)
print('Model Score:', modelScore)
# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
cofficents = model.coef_
intercept = model.intercept_
print('Coefficients:', cofficents)
print('Weight of feature one:', cofficents[0])
print('Weight of feature two:', cofficents[1])
print('Intercept:', intercept)
```

## Report on the score of the model
The $(R^2)$ score of **0.8552** indicates that the regression line approximates the actual data quite well. This score means that about **85.52%** of the variance in the actual data is captured by the regression line, leaving roughly **14.48%** of the variance unexplained. In practical terms, the regression line fits closely to the data points, but it doesn't perfectly match all of them. The higher the $(R^2)$ value, the better the line represents the actual data, so this value shows a strong, but not perfect, approximation.

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
scores = cross_val_score(model, X, y, cv=10)
print('Cross Validation Scores:', scores)
print('The mean/average of the cross validation scores: ', scores.mean())
# Report on their finding and their significance
```

## Findings and significance of the cross-validation results:

The cross-validation scores obtained from the model represent the $R^2$ values for each of the 10 folds. These scores demonstrate that the model performs consistently across different portions of the dataset, with values ranging between **0.811** and **0.887**. This consistency indicates that the model is reliable and also might not overfit to any particular subset of data. The average score of **0.855** closely aligns with the previously reported $R^2$ score of **0.8552**, confirming that the model generalizes well and maintains stable performance. The small variance in scores further suggests that the model is robust and not overly sensitive to how the data is split for training and testing.

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
Poly = PolynomialFeatures(degree=2)
X_poly = Poly.fit_transform(X)
X_train_poly, X_test_poly, y_train, y_test = train_test_split(X_poly, y, test_size=0.1, random_state=42)
poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)
y_pred = poly_model.predict(X_test_poly)
poly_model_score = poly_model.score(X_test_poly, y_test)
print('Poly Model Score:', poly_model_score)
poly_cofficients = poly_model.coef_
poly_intercept = poly_model.intercept_
print('Poly Coefficients:', poly_cofficients)
print('Poly Intercept:', poly_intercept)
# Report on the metrics and output the resultant equation as you did in Part 3.
```
