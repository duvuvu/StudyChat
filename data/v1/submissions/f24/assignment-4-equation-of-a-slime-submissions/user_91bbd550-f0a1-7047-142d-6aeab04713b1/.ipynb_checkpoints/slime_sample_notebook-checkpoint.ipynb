```python
# Imports section
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
import numpy as np
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
df = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(df.info())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df[['Temperature °C', 'Mols KCL']]
y = df['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
sample = pd.DataFrame([[500, 600]], columns=['Temperature °C', 'Mols KCL'])
predicted_size = model.predict(sample)
print(f"Predicted slime size for sample: {predicted_size[0]:.2f} nm^3")
# Report on the score for that model, in your own words (markdown, not code) explain what the score means
score = model.score(X_test, y_test)
print(f"score: {score:.2f}")
# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_
print(f"coefficients: {coefficients}")
print(f"intercept: {intercept}")
equation_latex = rf"h(x) = {intercept:.2f} + {coefficients[0]:.2f} \cdot \text{{Temperature}} + {coefficients[1]:.2f} \cdot \text{{Mols KCL}}"
print(f"Equation: {equation_latex}")
```

Sample equation: $E = mc^2$

Explanation for score: 
* The score is 0.86. This means that 86% of the changes in slime size can be predicted by our model using temperature and Mols of KCL. This shows us that the model is good at finding a strong connection between the 2 factors and the size of the slime.
* As the score is near 1, it shows that the model fits the data well.

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
#lets take cv = 10 --> Using 10-fold cross-validation
scores = cross_val_score(model, X, y, cv=10)
# Report on their finding and their significance
print(f"Cross-validation scores: {scores}")
print(f"Mean of all scores: {scores.mean():.2f}")
```

Finding and significance: 
* Cross-validation scores show the performance of a linear regression model across multiple data subsets.
* In our case, the scores (R^2) `range` from `0.78` to `0.89` with a `mean` of `0.86`.
* The score of 0.86 shows that on average, the model is able to generalize well with unseen data. The consistent performance of the model shows that the model is reliable.

Conclusion:
* The cross-validation results shows us that the linear regression model is a good fit for the data and the right choice.
