```python
# Imports section
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split, cross_val_score  
from sklearn.linear_model import LinearRegression 
from sklearn.preprocessing import PolynomialFeatures 
from sklearn.metrics import mean_squared_error
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
url = 'https://raw.githubusercontent.com/COMPSCI-383-Fall2024/assignment-4-equation-of-a-slime-<redacted>/refs/heads/main/science_data_large.csv?token=GHSAT0AAAAAACYTEJ3MQQVOUYCWK552DET4ZYXWV4Q'
data = pd.read_csv(url)
# Output the first 15 rows of the data
print(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
data.info()
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data[['Mols KCL', 'Temperature °C']]  
y = data['Size nm^3'] 
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

print(f"Training set shape: X_train: {X_train.shape}, y_train: {y_train.shape}")
print(f"Test set shape: X_test: {X_test.shape}, y_test: {y_test.shape}")
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

print("Coefficients:", linear_model.coef_)
print("Intercept:", linear_model.intercept_)

# Create a sample datapoint and predict the output of that sample with the trained model

sample_data = [[2, 25]]
predicted_size = linear_model.predict(sample_data)

print(f"Predicted size of slime for 2 mols of KCl and 25°C: {predicted_size[0]}")

# Report on the score for that model, in your own words (markdown, not code) explain what the score means

score = linear_model.score(X_test, y_test)

print(f"Model R^2 Score: {score}")

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX

```

### Model Equation

The equation for our model is: 

$$

$$

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
cv_scores = cross_val_score(linear_model, X_train, y_train, cv=5)

print("Cross-validation scores for each fold:", cv_scores)
print("Mean cross-validation score:", cv_scores.mean())
print("Standard deviation of cross-validation scores:", cv_scores.std())
# Report on their finding and their significance
```

### Cross-Validation Findings:

The model achieved a mean cross-validation score of `0.8590794402339516`, which suggests that the linear regression model performs reasonably well in predicting the size of the slime based on the KCl concentration and temperature. The standard deviation of `0.022493119105236842` indicates that the model's performance is fairly consistent across different folds, showing minimal variability.

A lower standard deviation signifies that the model's performance is stable, meaning it is likely to generalize well to unseen data. If the standard deviation were higher, it would suggest that the model's predictions may vary more depending on the specific split of the data, potentially indicating overfitting or underfitting.

Overall, the cross-validation results suggest that the model is reliable for predicting the growth of slime under different conditions.


## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2

poly = PolynomialFeatures(degree=2)

X_poly_train = poly.fit_transform(X_train)
X_poly_test = poly.transform(X_test)

poly_model = LinearRegression()
poly_model.fit(X_poly_train, y_train)

y_poly_pred = poly_model.predict(X_poly_test)

mse = mean_squared_error(y_test, y_poly_pred)
r2 = poly_model.score(X_poly_test, y_test)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")

# Report on the metrics and output the resultant equation as you did in Part 3.
```

### Polynomial Regression Findings:

The polynomial regression model (degree 2) achieved a **Mean Squared Error** of `1.4693515832582168e-08` and an **R² Score** of `1.0`. Compared to the linear regression model, this suggests that the polynomial regression model fits the data perfectly, explaining all of the variance in the target variable (slime size).

However, an **R² score** of `1.0` often indicates that the model may be overfitting, especially in cases where there’s no error. This means the model is capturing not only the underlying patterns in the data but also the noise, which may lead to poor generalization on unseen data.
