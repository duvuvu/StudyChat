```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import r2_score
import numpy as np
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
url = "/workspaces/assignment-4-equation-of-a-slime-<redacted>/science_data_large.csv"
df = pd.read_csv(url)

# Output the first 15 rows of the data
print("First 15 rows of the dataset:")
print(df.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print("\nDataset Summary:")
print(df.info())

num_data_points = df.shape[0]
print(f"Number of data points: {num_data_points}")
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df[['Temperature °C','Mols KCL']]  # Features: KCl and Temperature
y = df['Size nm^3']           # Label: Size_Change
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_point = np.array([[25, 5]])  # Example: 5 KCl, 25 Temperature, temp is 470
prediction = model.predict(sample_point)
print(f"Prediction for sample point {sample_point[0]}: {prediction[0]}")

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
#train_score = model.score(X_train, y_train)
test_score = model.score(X_test, y_test)

#print(f"Training set score: {train_score}")
print(f"Test set score: {test_score}")

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

## Model Score Explanation

The score reported by the model is the coefficient of determination (R-squared). 
For the test set, we got a score of 0.8552472077276095.

This score represents the proportion of the variance in the dependent variable (Size_Change) 
that is predictable from the independent variables (KCl and Temperature). 
A score of 1.0 indicates perfect prediction, while a score of 0 indicates that the model 
is just predicting the mean of the target variable.

Our score indicate that the model performs well on the test set. This suggests that our linear regression model has captured the underlying relationship between the input variables (KCl and Temperature) and the output variable (Size_Change) effectively. The model explains a large portion of the variance in the data and is likely to make good predictions on new, unseen data.

The linear regression model can be represented by the following equation:

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)
X_poly_train = poly.fit_transform(X_train)
X_poly_test = poly.transform(X_test)

# Train the model
model_poly = LinearRegression()
model_poly.fit(X_poly_train, y_train)

# Make predictions
y_pred_test = model_poly.predict(X_poly_test)

# Report on the metrics and output the resultant equation as you did in Part 3.
# Calculate R2 scores
r2_test = model_poly.score(X_poly_test, y_test)

#print(f"Training R2 score: {r2_train}")
print(f"Test R2 score: {r2_test}")

coefficients = model_poly.coef_
intercept = model_poly.intercept_

print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```


## Polynomial Regression Results (Degree 2)

### Process
1. We used PolynomialFeatures to transform our original features into polynomial features of degree 2.
2. We applied this transformation to both our training and test sets.
3. We then trained a linear regression model on these polynomial features.
4. Finally, we made predictions on the test set and calculated the R2 score.

### Metrics
- Test R2 score: 1.0

This R2 score indicates how well our polynomial model fits the test data. An R2 score closer to 1 indicates a better fit. This score is higher than the linear model's score from Part 3, suggesting better performance.

### Equation
The polynomial regression model of degree 2 can be represented by the following equation:

$$h(x) = a_0 + a_1x_1 + a_2x_2 + a_3x_1^2 + a_4x_1x_2 + a_5x_2^2$$

Where:
- $x_1$ represents the KCl concentration
- $x_2$ represents the Temperature
- $h(x)$ is the predicted Size_Change

Substituting the coefficients:

$$h(x) = 2.04771[10^{-5}] + [1.2][10^{1}]x_1 - [1.27195][10^{-7}]x_2 + [1.26494][10^{-11}]x_1^2 + [2.0]x_1x_2 + [2.85714][10^{-2}]x_2^2$$

### Interpretation

1. Constant term ([2.047710]): This is the predicted Size_Change when both KCl and Temperature are zero.
2. Linear terms ($x_1$ and $x_2$): These represent the linear effects of KCl and Temperature on Size_Change.
3. Quadratic terms ($x_1^2$ and $x_2^2$): These capture any non-linear relationships between each variable and Size_Change.
4. Interaction term ($x_1x_2$): This represents how the effect of KCl on Size_Change depends on the Temperature, and vice versa.

The magnitude and sign of each coefficient tell us about the strength and direction of each effect:
- Positive coefficients indicate that increasing that term increases Size_Change.
- Negative coefficients indicate that increasing that term decreases Size_Change.
- Larger absolute values indicate stronger effects.
