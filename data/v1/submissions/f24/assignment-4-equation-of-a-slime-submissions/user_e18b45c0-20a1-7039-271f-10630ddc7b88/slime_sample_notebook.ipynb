```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
dataset = pd.read_csv('science_data_large.csv')

# Output the first 15 rows of the data
print(dataset.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print(dataset.info())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = dataset[['Temperature °C', 'Mols KCL']] # Features
Y = dataset['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, Y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = [[427, 593]]
predicted_output = model.predict(sample_data)
print(f"Predicted output for the sample data {sample_data}: {predicted_output[0]}")

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print(f"coefficients: {coefficients}")
print(f"intercept: {intercept}")
```

### Equation for h(x):

$$
h(x) = 866.14641 \cdot \text{Temperature °C} - 1032.69506 \cdot \text{Mols KCL} + (-409391.47958)
$$

### What Does the Score Mean?

The score R^2 represents the variance among the dependent variable of the dataset, which in this case is "Size nm^3", that is explained by the indipendent variables "Temperature °C" and "Mols KCL".

A score closer to 1 indicates that the model explains most of the variance in the dataset, or that the model fits well. Vice versa applies when the score is close to 0 or negative. 

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
scores = cross_val_score(model, X, Y, cv=5, scoring='r2') 

print(f"Cross-validation R^2 scores: {scores}")
print(f"Mean R^2 score: {scores.mean()}")
print(f"Standard deviation of R^2 scores: {scores.std()}")
```

### Findings & Significance:

The cross validation scores are consistent with each other across the board, with a range from 0.839-0.872. Since the R^2 value is near 1, we can assume the model fits the data well. The mean score of R^2 is a representation of the average performance of the model across all scores, meaning that on average, the model explains 85.7% variance in the target variable. The standard deviation of the R^2 scores is very small, about 0.013, meaning that the performance is consistent across all splits of data. This means the model generalizes very well.

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)

# Train the model
poly_model = LinearRegression()
poly_model.fit(X_poly, Y)

# Evaulate model w/ transformed dataset
poly_scores = cross_val_score(poly_model, X_poly, Y, cv=5, scoring='r2')

print(f"Polynomial Regression R² scores: {poly_scores}")
print(f"Mean R² score: {poly_scores.mean()}")
print(f"Standard deviation of R² scores: {poly_scores.std()}")

# Make predictions using the polynomial regression model
sample_data_poly = poly.transform([[427, 593]])  # Use the same sample point but transform it to polynomial features
predicted_output_poly = poly_model.predict(sample_data_poly)
print(f"Predicted output for the sample data (polynomial regression): {predicted_output_poly[0]}")

# Extract coefficients and intercept
coefficients_poly = poly_model.coef_
intercept_poly = poly_model.intercept_

print(f"Polynomial coefficients: {coefficients_poly}")
print(f"Polynomial intercept: {intercept_poly}")
```

### Equation for h(x):
$$

+ 2.85714 \times 10^{-2} \cdot (\text{Temperature °C} \cdot \text{Mols KCL})
$$
