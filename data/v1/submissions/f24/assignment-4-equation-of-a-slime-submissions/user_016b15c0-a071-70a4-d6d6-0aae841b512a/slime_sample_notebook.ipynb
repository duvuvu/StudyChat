```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

data = pd.read_csv('science_data_large.csv')
print(data.head(15))
print(data.describe())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X, y = data[['Temperature °C', 'Mols KCL']], data['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, train_size=0.9, random_state=44)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
reg = LinearRegression().fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model

sample_datapoint = X_test.head(1)
sample_y = y_test.head(1)
predicition = reg.predict(sample_datapoint)
print(f'Predicition: {predicition[0]}, Actual: {sample_y.to_numpy()[0]}')

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
print(f'Score: {reg.score(X_test, y_test)}')
# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
print(f'Coefficients: {reg.coef_}')
print(f'Intercepts: {reg.intercept_}')
```

Report: The score is approximately 0.85385, which is a good but not perfect score. According to the SciKit-Learn documentation, the score represents the coefficient of determination, $R^2$. The best score would be 1.0, whereas the worst score would be 0. Mathematically, $R^2$ is the sum of the squared residuals divided by the sum of squared differences of each label from the label mean, subtracting that quotient from 1. So since the score is close to 1, but not extremely close, that means our predictions from our linear regression model are slightly off from the actual values.

Rounding to 5 significant digits: $h(x_1, x_2) = -417080 + 881.14x_1 + 1024.4x_2$. Where $x_1$ represents temperature (°C) and $x_2$ represents Mols KCL.

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
print(cross_val_score(LinearRegression(), X, y))
# Report on their finding and their significance
```

Report: The return value of the `cross_val_score` function is array of scores of our estimator for each run of the cross validation according to the SciKit-Learn documentation. Cross validation essentially subdivides the training data into several sets (in this case 5) and will train it each time on 4 out of the 5 data sets and use the remaining one as a test set. Each score is the represents one instance of cross validation. All of the scores are roughly the same, around 0.85, which was similar to the score in the previous section. It is good, but still has room for improvement as it could be closer to 1.

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
X_new = PolynomialFeatures(2).fit_transform(X_train)
X_test_poly = PolynomialFeatures(2).fit_transform(X_test)
poly = LinearRegression().fit(X_new, y_train)
# Report on the metrics and output the resultant equation as you did in Part 3.
print(f'Score: {poly.score(X_test_poly, y_test)}')
print(f'Coefficients: {poly.coef_}')
print(f'Intercept: {poly.intercept_}')
```

Report: The score for this model is exactly 1. This means that this polynomial regression model exactly fits the test data, meaning our predicitions are completely accurate.
