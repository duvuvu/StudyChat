```python
# Imports section
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import linear_model
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import KFold
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
dataset = pd.read_csv("science_data_large.csv")

# Output the first 15 rows of the data
print(dataset.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print(dataset.info())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]

print(X)
print(y)
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
train_model = model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = pd.DataFrame({'Temperature °C': [500], 'Mols KCL': [333]})
predicted_output = model.predict(sample_data)

print(f"Predicted output for the sample point {sample_data.values[0]}: {predicted_output[0]:.2f}")

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
r2_score = model.score(X_test, y_test)
print("r2 score:", r2_score)

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
n_splits = 5  # Number of folds
kf = KFold(n_splits=n_splits, shuffle=True)

# Perform cross-validation and calculate R² scores
cv_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')

print(f"Shuffle 1: Cross-Validation R^2 Scores: {cv_scores}")
print(f"Shuffle 1: Mean R^2 Score: {np.mean(cv_scores):.4f}")
print(f"Shuffle 1: Standard Deviation of R^2 Scores: {np.std(cv_scores):.4f}")
print("")

n_splits = 5 
kf = KFold(n_splits=n_splits, shuffle=True)

# Perform cross-validation and calculate R² scores
cv_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')

print(f"Shuffle 2: Cross-Validation R^2 Scores: {cv_scores}")
print(f"Shuffle 2: Mean R^2 Score: {np.mean(cv_scores):.4f}")
print(f"Shuffle 2: Standard Deviation of R^2 Scores: {np.std(cv_scores):.4f}")
print("")

n_splits = 10  
kf = KFold(n_splits=n_splits, shuffle=True)

# Perform cross-validation and calculate R² scores
cv_scores = cross_val_score(model, X, y, cv=kf, scoring='r2')

print(f"Shuffle 3: Cross-Validation R^2 Scores: {cv_scores}")
print(f"Shuffle 3: Mean R^2 Score: {np.mean(cv_scores):.4f}")
print(f"Shuffle 3: Standard Deviation of R^2 Scores: {np.std(cv_scores):.4f}")

# Report on their finding and their significance

```
