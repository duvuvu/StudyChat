```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures

```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
df = pd.read_csv("science_data_large.csv")
# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(df.shape)
print(df.describe())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df.drop(columns=['Size nm^3'])
y = df['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
linearModel = LinearRegression()
linearModel.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
datapoint = [[469, 647]] #first data point from the table
prediction = linearModel.predict(datapoint)
print(f"Prediction for sample datapoint: {prediction}")

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
modelScore = linearModel.score(X_test, y_test)
print(f"Model's r^2 score: {modelScore}")

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
print(f"Coefficients: {linearModel.coef_}")
print(f"Intercept: {linearModel.intercept_}")
```

The .score() method returns the $r^2$ for for a given model. The $r^2$ score is an indication of how well the model's predictions match the actual data, and it is on a range from 0 to 1, where a number closer to 1 indicates that the model has high predictive accuracy. 

For this model, the $r^2$ score is approximately 0.855. This indicates that the Linear Regression model explains about 85.5% of the variability in the slime's size, which suggests a relatovely strong predictive accuracy.


$h(x) = -409391.48 + 866.146 \cdot$ X1 + $1032.695 \cdot$ X2 , where X1 is the Temperature and X2 is the Mols.

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
scores = cross_val_score(linearModel, X, y, cv=100)
print(scores)
# Report on their finding and their significance
```

Findings: 

As I increase the number of folds (cv hyperparameter value), there are some higher scores in the iterations. For example, when cv=5, the highest score is ~0.872. When cv=10, the highest is ~0.887, and when cv=100, the highest is ~0.966. I also noticed that the scores increase and decrease in a varying manner across the iterations. For example, when for cv=100, the range is between ~0.023 to ~0.966, indicating that the model's performance is inconsistent across the folds, and is potentially overfitting in some folds, and underfitting in others.

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)

X_poly = poly.fit_transform(X)
X_train_poly, X_test_poly, y_train_poly, y_test_poly = train_test_split(X_poly, y, test_size=0.1, random_state=42)

polynomialRegression = LinearRegression()
polynomialRegression.fit(X_train_poly, y_train_poly)


# Report on the metrics and output the resultant equation as you did in Part 3.
print(f"Feature names: {poly.get_feature_names_out()}\n")

print(f"Model's score: {polynomialRegression.score(X_test_poly, y_test_poly)}\n")

print(f"Coefficients: {polynomialRegression.coef_}\n")
print(f"Intercept: {polynomialRegression.intercept_}")
```

The model's $r^2$ score is 1.0, indicating that it explains 100% of the variability in the data, indicating that it has a 100% predictive accuracy. While this may seem optimistic, it is important to also consider that the model might be overfitting to the training data, and may not generalize and perform well on new, unseen data.
