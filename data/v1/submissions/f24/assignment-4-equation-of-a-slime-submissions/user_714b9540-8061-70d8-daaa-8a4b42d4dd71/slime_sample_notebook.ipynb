```python
# Imports section
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, root_mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
science_data_large = os.path.join(os.getcwd(), "science_data_large.csv")
df = pd.read_csv(science_data_large)
# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(df.info())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df[['Temperature °C', 'Mols KCL']]  #features
y = df['Size nm^3'] #label

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression() # Create the  LinearRegression model object
model.fit(X_train, y_train) # Fit the model to the training data 
prediction = model.predict(X_test)  #  Make predictions on the test data 

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = [[500, 500]]
sample_prediction = model.predict(sample_data)
print(sample_prediction)
```

```python
# Report on the score for that model, in your own words (markdown, not code) explain what the score means
r2 = r2_score(y_test, prediction)
print(r2)
```

The R² score for the model indicates the proportion of the variance in the dependent variable (Size nm³) that is predictable from the independent variables (Temperature °C and Mols KCL). An R² score of 1 indicates a perfect fit, while a score of 0 means that the model does not explain any of the variability of the response data around its mean. The R² score for this score is around 0.855 which indicates that it is quite close to being a good fit.

```python
# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

# Function to round to 5 significant figures
def round_to_significant_figures(num, sig_figs=5):
    if num == 0:
        return 0
    else:
        return round(num, sig_figs - int(np.floor(np.log10(abs(num)))) - 1)
    
rounded_intercept = round_to_significant_figures(intercept)
rounded_coefficients = [round_to_significant_figures(coef) for coef in coefficients]

# Equation in LaTeX format:
# h(x) = intercept + coefficient1 * Temperature + coefficient2 * Mols KCL
print(f"h(x) = {rounded_intercept} + {rounded_coefficients[0]} * Temperature + {rounded_coefficients[1]} * Mols KCL")
```

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
scores = cross_val_score(model, X, y, cv=5) #5 fold

# Report on their finding and their significance
print("Cross-Validation Scores:", scores)
print("Average Cross-Validation Score:", np.mean(scores))
print("Standard Deviation of Scores:", np.std(scores))
```

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2) # Create Polynomial Features of degree 2
X_poly = poly.fit_transform(X)

# Split the dataset into training and testing sets (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, test_size=0.1, random_state=42)

# Train linear regression model on the polynomial features
poly_model = LinearRegression()
poly_model.fit(X_train, y_train)

# Predictions using the polynomial regression model
poly_prediction = poly_model.predict(X_test)

# Report on the metrics and output the resultant equation as you did in Part 3.
r2 = r2_score(y_test, poly_prediction)
print(r2)
```

The R² score for the model indicates the proportion of the variance in the dependent variable (Size nm³) that is predictable from the independent variables (Temperature °C and Mols KCL). An R² score of 1 indicates a perfect fit, while a score of 0 means that the model does not explain any of the variability of the response data around its mean. The R² score for this score is a perfect 1 which indicates that it is quite close to being a good fit.
