```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1. Loading the dataset

```python
url= '/workspaces/assignment-4-equation-of-a-slime-<redacted>/science_data_large.csv'
data = pd.read_csv(url)
print(data.head(15))
print(data.info())
# Using pandas load the dataset (load remotely, not locally)
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data.drop(columns=['Size nm^3'])  # features
y = data['Size nm^3']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=87)
#print(X_test);
```

## Part 3. Perform a Linear Regression

```python
model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_data = {
    'Temperature °C': [650], 
    'Mols KCL': [519],   
}
sample_df = pd.DataFrame(sample_data)
predicted_output = model.predict(sample_df)
print(f"Predicted output for the sample data point: {predicted_output[0]}")

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
y_pred = model.predict(X_test)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"Mean Absolute Error (MAE): {mae}")
print(f"R-squared (R²): {r2}")
print("\nthe MAE score gives the mean margin of error between the predicitoin and the actual values as an absolute value.\nit calculates the absolute value of each of the entries (actual-predicted) and returns the mean of all reasults\nin our case it seems like MAE is more than 12,000 wich is high and not optimal")
print("\nthe R² score measured between 0 and 1 and measure the strength of the relationship between the features and the target variable\nit evaluates if the the model fits the data and if there is a close relationship between the different features and the label.\nin our case the score is considerbly high(closer to 1) ehich means that there is a strong relationship and the model fits the data.")
print("\nwe can conclude from the scores that the model does fit our data, although it needs more data to train on in order to get more acurate reasults.")

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
intercept = model.intercept_

print("Intercept:", intercept)
print("\nCoefficients of the model:")
for i, col in enumerate(X.columns):
    print(f"{col}: {model.coef_[i]}")
```

equation: $h(x) = -406960.74749 + 861.91499 \cdot \text{temp} + 1029.15641\cdot \text{KCL} $


## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
scores = cross_val_score(model, X, y, cv=10, scoring='neg_mean_absolute_error')

# Report on their finding and their significance
scores = -scores
print(scores)
print("the major segnificance of finding is is the consistanct of the reasult, we can see that the range of the reasult is norrower than the actual errors which imply")
print("that there is no big segnificance to the order and the model give similar reasults while using different data.")
print("as already mentioned, the model needs to be trained on bigger data set to achive better reasults but the findings shows that the logic is correct")
```

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)  
X_test_poly = poly.transform(X_test)        # 

poly_model = LinearRegression()
poly_model.fit(X_train_poly, y_train)

y_pred_poly = poly_model.predict(X_test_poly)

mae_poly = mean_absolute_error(y_test, y_pred_poly)
r2_poly = r2_score(y_test, y_pred_poly)

print(f"R-squared (R²): {r2_poly}")
print(f"Mean Absolute Error (MAE) for Polynomial Regression (degree 2): {mae_poly}")
print("we can see that the R² score is 1 now(highest) which indicate that the relation between thwe features and the label is obvios and clean")
print("the MAE got readuced to a really small number, basically 0, which indicate that the model fits perfectly to the data and that the model")
print("will probably correctly predict the size of the slime given a temperature and amount of KCL")
# Report on the metrics and output the resultant equation as you did in Part 3.

intercept_poly = poly_model.intercept_

print("\n\nIntercept:", intercept_poly)
print("\nCoefficients of the model:")
for i, col in enumerate(X.columns):
    print(f"{col}: {poly_model.coef_[i]}")
```
