```python
# Imports section
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
from sklearn.metrics import  r2_score
from sklearn.metrics import root_mean_squared_error
from sklearn.model_selection import RepeatedKFold
from sklearn.metrics import make_scorer
from sklearn.pipeline import make_pipeline
# used 383gpt
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)/workspaces/assignment-4-equation-of-a-slime-<redacted>/science_data_large.csv

data = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print(f'First 15 rows: \n{data.head(15)}')
# Display a summary of the table information (number of datapoints, etc.)
print(f'Summary of data (rows, columns): \n{data.shape}')
print(f'Summary of data (column names): \n {data.columns}')
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

X = data[['Mols KCL', 'Temperature °C']] 
y = data['Size nm^3']                   

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set (used 383gpt for this entire section to help me understand)
linear_model = LinearRegression()
linear_model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample = pd.DataFrame([[500, 600]], columns=['Mols KCL', 'Temperature °C'])
predicted_size = linear_model.predict(sample)
print(f"Predicted size for [500 mols KCl, 600°C] is: {predicted_size[0]:.2f} nm^3")
y_pred = linear_model.predict(X_test)
r2 = r2_score(y_test, y_pred)
rmse = root_mean_squared_error(y_test, y_pred)


print(f"R² score: {r2:.4f}")
print(f"Root mean Squared Error: {rmse:.2f}")
# Report on the score for that model, in your own words (markdown, not code) explain what the score means

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = linear_model.coef_
intercept = linear_model.intercept_
print(f'Coefficients: {coefficients}')
print(f'Intercept: {intercept}')
```

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
# RepeatedKFold method
cv = RepeatedKFold(n_splits=5, n_repeats=10, random_state=42)
cv_rmse_scores = cross_val_score(linear_model, X, y, cv=cv, scoring='neg_root_mean_squared_error')
cv_rmse_scores = -cv_rmse_scores

print(f"Cross-Validation RMSE scores: {cv_rmse_scores}")
print(f"Mean RMSE score: {cv_rmse_scores.mean():.4f}")
print(f"Standard Deviation of RMSE scores: {cv_rmse_scores.std():.4f}")
```

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
model = make_pipeline(PolynomialFeatures(degree=2), LinearRegression())
model.fit(X_train, y_train)
y_pred_poly = model.predict(X_test)
rmse_poly = np.sqrt(root_mean_squared_error(y_test, y_pred_poly))

print(f'Polynomial Regression RMSE: {rmse_poly:.4f}')

linear_regressor = model.named_steps['linearregression']
coefficients = linear_regressor.coef_
intercept = linear_regressor.intercept_
print(f'Coefficients: {coefficients}')
print(f'Intercept: {intercept}')

# Report on the metrics and output the resultant equation as you did in Part 3.
```

# Polynomial Regression Equation

Polynomial Regression RMSE: 0.0110

The polynomial regression performed much better at capturing the non-linear relationships between the input variables and the size of the slime, as reflected in the RMSE (lower score).

The polynomial regression model can be expressed as:

$
h(x) = \text{intercept} + c_1 \cdot (\text{Mols KCL}) + c_2 \cdot (\text{Temperature °C}) + c_3 \cdot (\text{Mols KCL})^2 + c_4 \cdot (\text{Temperature °C})^2 + c_5 \cdot (\text{Mols KCL} \cdot \text{Temperature °C})
$

Where:
- Intercept = 0.0110
- Coefficients:
  - $ c_1 = -1.27196088 \times 10^{-7}$ (coefficient for Mols KCL)
  - $c_2 = 12.0$ (coefficient for Temperature °C)
  - $ c_3 = 0.0285714287$ (coefficient for {Mols KCL}^2)
  - $ c_4 = 2.0$ (coefficient for {Temperature °C}^2)
  - $c_5 = 1.26471056 \times 10^{-11}$ (coefficient for {Mols KCL} $\cdot$ {Temperature °C})
