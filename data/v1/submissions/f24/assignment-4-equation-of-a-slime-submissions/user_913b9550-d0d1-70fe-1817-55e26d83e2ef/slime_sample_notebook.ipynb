```python
# Imports section
import pandas as pd
from sklearn import *
# just using for mean function
import statistics as s
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

sci_data = pd.read_csv("science_data_large.csv")
print(sci_data.head(15))
sci_data.info()
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)



X = sci_data.iloc[:,0:2]
Y = sci_data.iloc[:,2]


# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = sklearn.linear_model.LinearRegression().fit(X_train, Y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sample_X = X_test.iloc[1:3,:]

#print(sample_X)
sample_prediction = model.predict(sample_X)
sample_score = model.score(sample_X, sample_prediction)
print(sample_prediction)
print(sample_score)

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
# Done

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficient = model.coef_
intercept = model.intercept_
print(coefficient)
print(intercept)
```

The score for that model is 1.0. This means the model exactly fits the data given. This makes sense, as I am using data from the same place both to predict and to score.

A: Temperature in $^{\circ} C$ \
B: Moles KCl \
h(A,B): Size nm<sup>3</sup> \
Function:  $h(A,B)= 866.14641A + 1032.69507B - 409391.47958$ 


## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
scores = sklearn.model_selection.cross_val_score(model, X_test, Y_test)
print(scores)
print(s.mean(scores))
# Report on their finding and their significance
```

The model seems to have a R<sup>2</sup> value of around 0.85
This means the regression is about an 85% fit to the data.

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
poly_test = sklearn.preprocessing.PolynomialFeatures(degree = 2).fit(X_test, Y_test)
ptest_data = poly.fit_transform(X_test, Y_test)

poly = sklearn.preprocessing.PolynomialFeatures(degree = 2).fit(X_train, Y_train)
poly_data = poly.fit_transform(X_train, Y_train)
#print(poly_data)
print(poly.powers_)
poly_model = sklearn.linear_model.LinearRegression().fit(poly_data, Y_train)


base_score = poly_model.score(ptest_data, Y_test)
p_scores = sklearn.model_selection.cross_val_score(poly_model, ptest_data, Y_test)
print(base_score)
print(p_scores)
print(s.mean(p_scores))



# Report on the metrics and output the resultant equation as you did in Part 3.
p_coefficients = poly_model.coef_
p_intercept = model.intercept_
#print(p_powers)
print(p_coefficients)
print(p_intercept)
```

The R^2 score for that model is 1.0, both in simple scoring as well as cross-validated scoring. This means we have identified a function that perfectly matches the data.\
