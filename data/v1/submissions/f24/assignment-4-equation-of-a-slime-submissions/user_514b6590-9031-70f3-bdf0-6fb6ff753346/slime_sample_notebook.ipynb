```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
data = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
print(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(data.describe())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data.drop(columns=['Size nm^3'])
Y = data['Size nm^3']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, Y_train)

# Create a sample datapoint and predict the output of that sample with the trained model
sampleData = [[500.5, 471.53]]
prediction = model.predict(sampleData)
print(f'Predicted size: {prediction}')

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
testPredict = model.predict(X_test)
score = r2_score(Y_test, testPredict)
print(f'Model score: {score}')

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coef = model.coef_
print(f'Coefficients: {coef}')
inter = model.intercept_
print(f'Intercept: {inter}')
```

The score (r^2 value), represents what portion of the variance in the Y is predictable by the X. A higher score indicates that more variability of the Y is explainable by the X. The 0.85525 received here indicates that 85.5% of the variation is explainable, meaning the models predictions make 85.5% sense. A perfect score would be 1 or 100%. 

Sample equation: $E = mc^2$

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data
scores = cross_val_score(model, X_test, Y_test, cv=5)

# Report on their finding and their significance
print(f'Cross value scores: {scores}')
```

The cross value scores are extremely similar to the single score, meaning the initial score was not a fluke or a lucky outcome. The model seems to consistently predict with an accuracy of above 80%.

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2

# Initialize and fit data
polynom = PolynomialFeatures(degree=2)
X_train_polynom = polynom.fit_transform(X_train)
X_test_polynom = polynom.transform(X_test)

# Train model
polyModel = LinearRegression()
polyModel.fit(X_train_polynom, Y_train)

# Report on the metrics and output the resultant equation as you did in Part 3.

# Use the old sample datapoint and predict the output of that sample with the polynomial model
polySample = polynom.fit_transform(sampleData)
polyPrediction = polyModel.predict(polySample)
print(f'Predicted size: {polyPrediction}')

# Report on the score for that model, in your own words (markdown, not code) explain what the score means
polyTestPredict = polyModel.predict(X_test_polynom)
polyScore = r2_score(Y_test, polyTestPredict)
print(f'Model score: {polyScore}')

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
polyCoef = polyModel.coef_
print(f'Coefficients: {polyCoef}')
polyInter = polyModel.intercept_
print(f'Intercept: {polyInter}')
```
