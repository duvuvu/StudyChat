```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
df = pd.read_csv("/workspaces/assignment-4-equation-of-a-slime-<redacted>/science_data_large.csv")
# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(df.info())
```

```python
print(df.describe())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = df.drop("Size nm^3", axis=1)
y = df["Size nm^3"]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.9, test_size=0.1, random_state=42)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set
model = LinearRegression()
model.fit(X_train, y_train)
# Create a sample datapoint and predict the output of that sample with the trained model
# sample_dataPoint = [[95,499]]
sample_dataPoint = pd.DataFrame([[95, 499]], columns=["Temperature °C", "Mols KCL"])
predicted_dataPoint = model.predict(sample_dataPoint)
print("predicted_dataPoint: ", predicted_dataPoint)
# Report on the score for that model, in your own words (markdown, not code) explain what the score means
model_score = model.score(X_test, y_test)
print("model_score: ", model_score)
# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

Sample equation:  $E = MC^2$ 

The R² score ranges from 0 to 1:
- A score of 1 indicates that the model explains all the variance in the target variable, meaning it is a perfect fit.

The model's R² score is 0.855, which means it explains 85.5% of the variation in the target variable. This is a good score, showing that the model does a good job predicting "Size $nm^3$" based on the features (Temperature and Mols KCL). However, it's not perfect, so there may still be some patterns or information that the model isn't capturing completely.

The model equtation is given by $h(x) = \theta_0 + \theta_1 \cdot x_1 + \theta_2 \cdot x_2$

$h(x) = -409391.48 + 866.15 \cdot x_1 + 1032.70 \cdot x_2$ 

OR

$h(x) = -409391.48 + 866.15 \cdot  Temperature °C  + 1032.70 \cdot Mols KCL$


## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data

# Performing cross-validation with 5 folds.
scores = cross_val_score(model, X, y, cv=5)

# Output the cross-validation scores and the mean score
print("Cross-validation scores:", scores)
print("Mean of cross-validation score:", scores.mean())

# Report on their finding and their significance
# Report given in the Markdown.
```

The individual scores show how well the model worked on different parts of the data after it was shuffled. Since the scores are close to 1, it means the model does a good job of explaining most of the differences in the data across these different shuffles, which suggests it works well for different sets of data.


The average score is 0.857, meaning that on average, the model explains about 85.7% of the variations in the target variable. This is a strong result for a regression model and indicates that the model has learned important patterns from the data.

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2
# Step 1: Creating a pipeline with PolynomialFeatures and LinearRegression
pipeline = Pipeline([
    ('poly', PolynomialFeatures(degree=2)),
    ('linear', LinearRegression())
])

pipeline.fit(X_train, y_train)
# printing the R-squared score
r2_score = pipeline.score(X_test, y_test)
print("R-squared score:", r2_score)
sample_dataPoint = pd.DataFrame([[95, 499]], columns=["Temperature °C", "Mols KCL"])
predicted_dataPoint = pipeline.predict(sample_dataPoint)
print("predicted_dataPoint: ",predicted_dataPoint)
# Report on the metrics and output the resultant equation as you did in Part 3.
coefficients = pipeline.named_steps['linear'].coef_
intercept = pipeline.named_steps['linear'].intercept_

# # printing the coefficients and intercept
print("Coefficients:", coefficients)
print("Intercept:", intercept)
# Getting the output feature names for transformation.
feature_names = pipeline.named_steps['poly'].get_feature_names_out()
print("Feature Names:", feature_names)

# wiriting the equation of h(x) in the markdown.
```

The R² score is 1.0, meaning the model perfectly explains all the variance in the target variable (Size nm³) based on the features. A score of 1 indicates a perfect fit, meaning the model captures the relationship between the features and the target variable without any unexplained variation.

I noticed that the "Size $nm^3$" predicted by the linear regression model was slightly different from the value given in the dataset for the same data point. However, when I used the same data point with the polynomial regression model, it predicted the exact same "Size $nm^3$" value as in the dataset.

This suggests that the polynomial regression model is better at capturing the underlying relationship between the features (Temperature °C and Mols KCL) and the target variable ("Size $nm^3$")
