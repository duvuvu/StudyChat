```python
# Imports section
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
data = pd.read_csv('science_data_large.csv')
# Output the first 15 rows of the data
display(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
data.info()
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)
X = data[['Temperature °C', 'Mols KCL']]
y = data[['Size nm^3']]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, train_size=0.9, random_state=42)
print(X_train)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

model = LinearRegression()
model.fit(X_train, y_train)

# Create a sample datapoint and predict the output of that sample with the trained model

sample_dp = pd.DataFrame({'Temperature °C': [663], 'Mols KCL': [295]})
sample_pred = model.predict(sample_dp)

print("sample prediction =", sample_pred)

# Report on the score for that model, in your own words (markdown, not code) explain what the score means

train_score = model.score(X_train, y_train)
print("score of training set =", train_score)

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX

coeffs = model.coef_
inter = model.intercept_

print("coefficients =", coeffs)
print("intercept =", inter)
```

The score of 0.861 suggests that the model is a pretty good fit for the data considering it is on a scale of 0 to 1.0. This score means that the model explains $\approx$ 86% of the variation of the size of the slime based on the inputs for temperature and mols of KCL

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data

scores = cross_val_score(model, X_train, y_train, cv=10)
print("scores =", scores)

# Report on their finding and their significance

print("mean of scores =", np.mean(scores))
print("standard deviation =", np.std(scores))
```

Overall, each fold of the data has a very similar score and the model fits very well for each fold, as the lowest score of all of the iterations is 0.815. The mean of all the scores signifies that the average score is $\approx$ 0.85827 which is very close to the training score in part 3. 


## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2

poly = PolynomialFeatures(2)
poly_X = poly.fit_transform(X_train)

poly_model = LinearRegression()
poly_model.fit(poly_X, y_train)

poly_score = poly_model.score(poly_X, y_train)
print("score of poly_training set =", poly_score)

poly_coef = poly_model.coef_
poly_inter = poly_model.intercept_

print("poly_coefficients =", poly_coef)
print("poly_intercepts =", poly_inter)

# Report on the metrics and output the resultant equation as you did in Part 3.

poly_cross_scores = cross_val_score(poly_model, poly_X, y_train, cv=10)
print("cross scores =", poly_cross_scores)
print("mean of scores =", np.mean(poly_cross_scores))
print("standard deviation of scores =", np.std(poly_cross_scores))
```
