```python
# Imports section
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score, train_test_split
from sklearn.preprocessing import PolynomialFeatures

```

## Part 1. Loading the dataset

```python
# Using pandas load the dataset (load remotely, not locally)
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

url = 'science_data_large.csv'
# Load the dataset
data = pd.read_csv(url)

# Output the first 15 rows of the data
print(data.head(15))

# Display a summary of the table information
print(data.info())
```

## Part 2. Splitting the dataset

```python
# Take the pandas dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X = data.drop(columns=['Size nm^3'])  # Features
y = data['Size nm^3']                  # Labels

# Split the data into training and test sets (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Output the shapes to confirm the split
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

## Part 3. Perform a Linear Regression

```python
# Use sklearn to train a model on the training set

# Create a sample datapoint and predict the output of that sample with the trained model

# Report on the score for that model, in your own words (markdown, not code) explain what the score means

# Extract the coefficents and intercept from the model and write an equation for your h(x) using LaTeX

# Initialize the model
model = LinearRegression()

# Train the model on the training set
model.fit(X_train, y_train)

```

```python
# Create a sample datapoint (for example)
sample_datapoint = pd.DataFrame({'Temperature °C': [650], 'Mols KCL': [500]})

# Predict the output using the trained model
predicted_output = model.predict(sample_datapoint)
print("Predicted Output:", predicted_output)
```

```python
# Calculate the R^2 score on the training set
train_score = model.score(X_train, y_train)
print("Training set R^2 score:", train_score)

# Optionally, calculate the R^2 score on the test set to evaluate generalization
test_score = model.score(X_test, y_test)
print("Test set R^2 score:", test_score)
```

In this model, the R2 scode is 0.86. The score is on a scale of 0 to 1, with 1 being completely accurate and 0 being completely inaccurate. A score of 0.86 shows the model is fairlty accurate, and works well.

```python
# Get coefficients and intercept
coefficients = model.coef_
intercept = model.intercept_

print("coefficients", coefficients)
print("intercept",intercept)
```

## Part 4. Use Cross Validation

```python
# Use the cross_val_score function to repeat your experiment across many shuffles of the data

# Report on their finding and their significance

# Use cross_val_score to evaluate the model
# Let's say we want to use 10-fold cross-validation
scores = cross_val_score(model, X, y, cv=10, scoring='r2')  # Use 'r2' scoring for regression models

# Report findings
print("Cross-validation scores for each fold:", scores)
print("Mean R^2 score:", scores.mean())
print("Standard Deviation of R^2 scores:", scores.std())
```

The cross validation score breakes up the data into multiple sets, I chose 10, and then calculates an R2 score for each one of those sets, figuring out the accuracy. Then it shows that the the mean aligns with our overall accuracy score, showing that different subsets also average out to a similar accuracy. The standard deviation is also fairly low, showing that the R2 score remains quite consistant

## Part 5. Using Polynomial Regression

```python
# Using the PolynomialFeatures library perform another regression on an augmented dataset of degree 2

# Report on the metrics and output the resultant equation as you did in Part 3.

# Create polynomial features of degree 2 for the training set

# Split features and target variable
X = data.drop(columns=['Size nm^3'])
y = data['Size nm^3']

# Split the original data into training and test sets (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

poly = PolynomialFeatures(degree=2, include_bias=False)  # No bias term since LinearRegression adds it
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)

# Initialize and fit the model on the training set
model_poly = LinearRegression()
model_poly.fit(X_train_poly, y_train)

# Evaluate the model using cross-validation on the training set
scores = cross_val_score(model_poly, X_train_poly, y_train, cv=10, scoring='r2')

# Report the metrics
print("Cross-validation scores for each fold:", scores)
print("Mean R^2 score for polynomial regression:", scores.mean())
print("Standard Deviation of R^2 scores for polynomial regression:", scores.std())

# Get coefficients and intercept for the equation
coefficients = model_poly.coef_
intercept = model_poly.intercept_
print("coefficients", coefficients)
print("intercept",intercept)
```
