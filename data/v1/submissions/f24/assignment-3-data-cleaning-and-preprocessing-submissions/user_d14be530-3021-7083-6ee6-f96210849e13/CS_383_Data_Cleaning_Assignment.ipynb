## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows

numerical_cols1 = df1.select_dtypes(include=['int64', 'float64']).columns
numerical_cols2 = df2.select_dtypes(include=['int64', 'float64']).columns

numerical_duplicates_df1 = df1[numerical_cols1].duplicated()
numerical_duplicates_df2 = df2[numerical_cols2].duplicated()

print(f"Total duplicate rows in numerical dataset 1: {numerical_duplicates_df1.sum()}")
print(f"Total duplicate rows in numerical dataset 2: {numerical_duplicates_df2.sum()}")


# Drop these duplicate rows

df1_numerical_cleaned = df1[~numerical_duplicates_df1]
df2_numerical_cleaned = df2[~numerical_duplicates_df2]

# Confirm that duplicates are removed
print(f"Dataset 1 shape after removing duplicates: {df1_numerical_cleaned.shape}")
print(f"Dataset 2 shape after removing duplicates: {df2_numerical_cleaned.shape}")



# Repeat the same for categorical dataset. Print the duplicate rows and drop them

# For the categorical dataset, check if there are duplicate rows
# First, identify the categorical columns
categorical_cols1 = df1.select_dtypes(include=['object']).columns
categorical_cols2 = df2.select_dtypes(include=['object']).columns

# Check for duplicate rows in the categorical dataset
categorical_duplicates_df1 = df1[categorical_cols1].duplicated(keep='first')
categorical_duplicates_df2 = df2[categorical_cols2].duplicated(keep='first')

# Print the total number of duplicate rows
print(f"Total duplicate rows in categorical dataset 1: {categorical_duplicates_df1.sum()}")
print(f"Total duplicate rows in categorical dataset 2: {categorical_duplicates_df2.sum()}")

# Drop these duplicate rows for df1
df1_categorical_cleaned = df1[~categorical_duplicates_df1].reset_index(drop=True)

# For df2, since there are no duplicates, we don't need to clean it
df2_categorical_cleaned = df2.copy()

# Confirm that duplicates are removed (df1 only)
print(f"Dataset 1 shape after removing categorical duplicates: {df1_categorical_cleaned.shape}")
print(f"Dataset 2 shape remains unchanged (no categorical duplicates found): {df2_categorical_cleaned.shape}")


````

## Part 3.2: Combine two differents datasets

A good skill to have is to know how to combine 2 different datasets.

Are all the unique ids are present in both datasets? Why do you think so? If not, what do the rows that are missing from one of the datasets look like in the combined table?

Answer: No, not all unique IDs are present in both datasets. This is likely because the two datasets represent different types of data, collected from different tests or assessments performed on the same or different subsets of patients

````python
# Merge the two given numerical and categorical datasets based on their unique_ID.

combined_df = pd.merge(df1, df2, on='unique_id', how='outer')


#Print the combined dataset

print("Combined Dataset:")
print(combined_df.head())
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value

total_rows = combined_df.shape[0]
rows_with_missing_values = combined_df.isnull().any(axis=1).sum()
percentage_missing = (rows_with_missing_values / total_rows) * 100

# Print %
print(f"Percentage of rows with at least one missing value: {percentage_missing:.2f}%")

# Drop these rows from the dataset

cleaned_df = combined_df.dropna()

# Print the Dataset

print("Dataset after dropping rows with missing values:")
print(cleaned_df.head())
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting

sorted_df = combined_df.sort_values(by='Target').reset_index(drop=True)


# Print the dataset

print("Dataset sorted by 'Target' column:")
print(sorted_df.head())
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']
df_with_dummies = pd.get_dummies(sorted_df, columns=categorical_columns, drop_first=False)

# Print the dataset
print("Dataset with dummy variables for categorical columns:")
print(df_with_dummies.head())
````

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

for col in numerical_columns:
    mean = df_with_dummies[col].mean()
    std = df_with_dummies[col].std()
    
    # Keep rows where the value is within 3 standard deviations from the mean
    df_with_dummies = df_with_dummies[(df_with_dummies[col] >= (mean - 3 * std)) & (df_with_dummies[col] <= (mean + 3 * std))]

# Print the dataset

print("Dataset after removing outliers from numerical columns:")
print(df_with_dummies.head())
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
from sklearn.preprocessing import MinMaxScaler

numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Normalize the numerical columns
df_with_dummies[numerical_columns] = scaler.fit_transform(df_with_dummies[numerical_columns])

# Print the dataset

print("Dataset after normalizing numerical columns:")
print(df_with_dummies.head())
````

*Describe what the above code does here*

1) cleaned_df.groupby('Target'): This part of the code groups the rows in the cleaned_df DataFrame by the values in the 'Target' column. The groupby() function essentially creates groups based on the unique values in the 'Target' column. Each group contains rows that share the same value for this column (e.g., 'ckd' and 'notckd').

2) .size(): After grouping the rows by 'Target', the .size() function counts the number of rows (or occurrences) in each group. This produces a Series where the index is the value of 'Target', and the corresponding value is the count of rows for that target value.

3) .reset_index(name='count'): The reset_index() function converts the resulting grouped data back into a DataFrame. It creates a new DataFrame where each unique value from the 'Target' column becomes a row, and the count of occurrences for each value is stored in a new column named 'count'.

4) print(result): This prints the final DataFrame, where each row contains a unique value of 'Target' and its corresponding count, showing how many times each target value appears in the dataset.
