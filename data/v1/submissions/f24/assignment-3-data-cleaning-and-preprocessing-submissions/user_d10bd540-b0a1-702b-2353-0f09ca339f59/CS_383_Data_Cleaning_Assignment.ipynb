## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']
df_combined_datasets = pd.get_dummies(df_combined_datasets, categorical_columns, drop_first=False)

# Print the dataset
print(df_combined_datasets)
````

In the example we went through above, another solution is to have a single column for the binary variable. In the downstream modeling would this be equivalent? What effect would this have if the categorical variable could take more than 2 values? For example, let's say we have a categorical feature that is "type of condiment" that can take 5 separate values and we are trying to predict the rating of a particular sandwich.

In the downstream modeling process, whether you use dummy variables (multiple columns) or a single binary column depends on the type of categorical variable:

#### **Binary Variables (Two Values):**
If the categorical variable is binary (e.g., 'rbc' with values 'normal' and 'abnormal'), using two dummy columns(`rbc_normal`, `rbc_abnormal`) or a single column (e.g., `rbc` as 0 or 1) would be mathematically equivalent for many models, such as linear regression, logistic regression, and tree-based models. 

#### **Effect of Single Column for Binary Variables:**
In practice, it's common to reduce binary variables to a single column (0 or 1), as this saves memory and computation. For example:
- `rbc` could be represented as 1 for "normal" and 0 for "abnormal".
- This doesn't lose any information because having two columns would be redundant: if `rbc_normal` is 1, you know `rbc_abnormal` must be 0.

For most machine learning models, this would have no effect on predictive performance, and the interpretation remains straightforward. It’s more efficient because you’re not adding unnecessary columns.

### **Categorical Variables with More Than Two Values:**
For categorical variables with more than two unique values (e.g., "type of condiment" with values like `ketchup`, `mustard`, `mayo`, `bbq`, and `ranch`), the situation changes.

#### **Effect of Single Column for Multiclass Variables:**
If you try to use a single column for multiclass categorical variables, this column would typically contain integer values (e.g., 0 for ketchup, 1 for mustard, 2 for mayo, etc.). However, this approach would not be suitable for most models because:
1. Ordinal Interpretation: A model might treat the values as having an ordinal relationship (e.g., `mayo > mustard > ketchup`), even though the categories are unrelated. This could lead to incorrect model assumptions.
2. Loss of Information: Assigning a single integer value to each category prevents the model from understanding that the different categories are independent of each other (i.e., they are not ordered).

Therefore, for categorical variables with more than two unique values, you typically use one-hot encoding (i.e., creating a separate dummy column for each unique value). This way, the model treats each category as independent, without assuming any implicit order.

#### **Example with "Type of Condiment":**
Let’s say the categorical variable "condiment" has five unique values:
- `condiment`: ketchup, mustard, mayo, bbq, ranch.

Using one-hot encoding, you would get five separate columns:
- `condiment_ketchup`
- `condiment_mustard`
- `condiment_mayo`
- `condiment_bbq`
- `condiment_ranch`

Each row would have one column set to 1 (the actual condiment used), and the rest would be 0.

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
for cols in numerical_columns:
  mean = df_combined_datasets[cols].mean()
  std_dev = df_combined_datasets[cols].std()
  upper_limit = mean + 3 * std_dev
  lower_limit = mean - 3 * std_dev
  df_combined_datasets = df_combined_datasets[(df_combined_datasets[cols] >= lower_limit) & 
  (df_combined_datasets[cols] <= upper_limit)]

# Print the dataset
print(df_combined_datasets)
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_combined_datasets[numerical_columns] = scaler.fit_transform(df_combined_datasets[numerical_columns])

print(df_combined_datasets[numerical_columns])
````

These hallucinations occur because the model may attempt to infer missing or ambiguous details, resulting in fabricated or incorrect information. Here’s how to mitigate such issues:
1. Cross-Validation with Original Data: Always compare the model's output (converted data) against the original data. Ensure that the numbers, text, or structures match exactly.
2. Manual Spot-Checking: After using an LLM for conversions, manually review a subset of the data. For example, in the top 15 rows, you can manually check whether 3-5 rows are accurately represented.
3. Use Small Batches for Verification: If you’re dealing with a large dataset, break it down into smaller batches (e.g., 10-20 rows) and ask the model to convert one batch at a time. This way, it’s easier to spot and fix potential errors.
