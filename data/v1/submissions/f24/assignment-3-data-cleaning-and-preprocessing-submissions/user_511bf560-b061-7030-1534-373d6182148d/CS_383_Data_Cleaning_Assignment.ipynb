## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows
duplicate_rows_dataset1 = dataset1[dataset1.duplicated(subset='unique_id', keep=False)]

if not duplicate_rows_dataset1.empty:
    print("Duplicate rows in Dataset 1:")
    print(duplicate_rows_dataset1)
    print(f"\nTotal number of duplicate rows in Dataset 1: {duplicate_rows_dataset1.shape[0]}")
# Drop these duplicate rows
dataset1_cleaned = dataset1.drop_duplicates(subset='unique_id', keep='first')

# Repeat the same for categorical dataset. Print the duplicate rows and drop them
duplicate_rows_dataset2 = dataset2[dataset2.duplicated(subset='unique_id', keep=False)]

if not duplicate_rows_dataset2.empty:
    print("Duplicate rows in Dataset 2:")
    print(duplicate_rows_dataset2)
    print(f"\nTotal number of duplicate rows in Dataset 2: {duplicate_rows_dataset2.shape[0]}")

dataset2_cleaned = dataset2.drop_duplicates(subset='unique_id', keep='first')

````

## Part 3.2: Combine two differents datasets

A good skill to have is to know how to combine 2 different datasets.

Are all the unique ids are present in both datasets? Why do you think so? If not, what do the rows that are missing from one of the datasets look like in the combined table?

Answer:

````python
# Merge the two given numerical and categorical datasets based on their unique_ID.
unique_ids_dataset1 = set(dataset1_cleaned['unique_id'])
unique_ids_dataset2 = set(dataset2_cleaned['unique_id'])

# Finding unique IDs that are not in both datasets
missing_in_dataset2 = unique_ids_dataset1 - unique_ids_dataset2
missing_in_dataset1 = unique_ids_dataset2 - unique_ids_dataset1

# Printing missing unique IDs
if missing_in_dataset2:
    print(f"Unique IDs present in Dataset 1 but missing in Dataset 2: {missing_in_dataset2}")
else:
    print("All unique IDs from Dataset 1 are present in Dataset 2.")

if missing_in_dataset1:
    print(f"Unique IDs present in Dataset 2 but missing in Dataset 1: {missing_in_dataset1}")
else:
    print("All unique IDs from Dataset 2 are present in Dataset 1.")

combined_dataset = pd.merge(dataset1_cleaned, dataset2_cleaned, on='unique_id', how='outer')
print("\nCombined Dataset:")
print(combined_dataset.head())

#Print the combined dataset
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value
total_rows = combined_dataset.shape[0]  
rows_with_missing = combined_dataset.isnull().any(axis=1).sum()
percentage_missing = (rows_with_missing / total_rows) * 100

# Print %
print(f"Percentage of rows with at least one missing value: {percentage_missing:.2f}%")

# Drop these rows from the dataset
cleaned_combined_dataset = combined_dataset.dropna()

# Print the Dataset
print("\nCleaned Dataset (after dropping rows with missing values):")
print(cleaned_combined_dataset.head())  # Show the first 5 rows of the cleaned dataset
print(f"\nNumber of rows in the cleaned dataset: {cleaned_combined_dataset.shape[0]}")
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting
sorted_dataset = cleaned_combined_dataset.sort_values(by='Target')
sorted_dataset.reset_index(drop=True, inplace=True)
# Print the dataset
print("\nSorted Dataset by 'Target' column:")
print(sorted_dataset.head())
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']
encoded_dataset = pd.get_dummies(sorted_dataset, columns=categorical_columns, drop_first=True)

# Print the dataset
print("\nEncoded Dataset with Dummy Variables:")
print(encoded_dataset.head()) 
````

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
def remove_outliers(df, columns):
    for column in columns:
        mean = df[column].mean()
        std_dev = df[column].std()
        
        # Define thresholds for outliers
        upper_threshold = mean + 3 * std_dev
        lower_threshold = mean - 3 * std_dev
        
        # Filter out the outliers
        df = df[(df[column] <= upper_threshold) & (df[column] >= lower_threshold)]
    
    return df
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']
cleaned_dataset_no_outliers = remove_outliers(encoded_dataset, numerical_columns)

# Print the dataset
print("\nCleaned Dataset (after removing outliers from numerical columns):")
print(cleaned_dataset_no_outliers.head())
print(f"\nNumber of rows in the cleaned dataset after outlier removal: {cleaned_dataset_no_outliers.shape[0]}")
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
scaler = StandardScaler()
normalized_data = scaler.fit_transform(cleaned_dataset_no_outliers[numerical_columns])
normalized_df = pd.DataFrame(normalized_data, columns=numerical_columns)
final_normalized_dataset = cleaned_dataset_no_outliers.copy()
final_normalized_dataset[numerical_columns] = normalized_df
print("\nNormalized Dataset:")
print(final_normalized_dataset.head())
````

## Part 3.8: Remove Unnecessary columns

Are there any columns in this dataset which are not appropriate for modeling and predictions? Which column(s)? Justify their exclusion and remove them

Answer: unique_id is the column in this dataset that does not hold any value in terms of modeling and predictions. Its purpose was to provide an identifier to each row, which was helpful when it came to cleaning the data. 

````python
#Remove that column
unnecessary_columns = ['unique_id']
final_dataset_cleaned = final_normalized_dataset.drop(columns=unnecessary_columns)
# Print the dataset
print("\nDataset after removing unnecessary columns:")
print(final_dataset_cleaned.head())
````

Now that you've completed these cleaning steps you should have a pandas dataframe which is much cleaner and ready for modeling. Our final step is to save our work. Export the DataFrame to a two new formats: csv and json.
