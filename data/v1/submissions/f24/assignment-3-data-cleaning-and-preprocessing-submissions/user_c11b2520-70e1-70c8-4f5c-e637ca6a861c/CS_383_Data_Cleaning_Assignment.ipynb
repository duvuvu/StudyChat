## Part 3.1 : Drop Duplicate rows

Let's start by checking if the given datasets have any duplicate rows (same Unique Id). Use pandas to identify and remove these duplicate rows from the given dataset

````python
# For the numerical dataset, check if there are duplicate rows in the dataset. If yes, print total number of duplicate rows
duplicate_rows_numerical = numerical_data[numerical_data.duplicated()]
print(f"Duplicate rows in numerical dataset:\n{duplicate_rows_numerical}")
# Drop these duplicate rows
numerical_data_cleaned = numerical_data.drop_duplicates()

# Repeat the same for categorical dataset. Print the duplicate rows and drop them
duplicate_rows_categorical = categorical_data[categorical_data.duplicated()]
print(f"Duplicate rows in categorical dataset:\n{duplicate_rows_categorical}")

categorical_data_cleaned = categorical_data.drop_duplicates()
````

## Part 3.2: Combine two differents datasets

A good skill to have is to know how to combine 2 different datasets.

Are all the unique ids are present in both datasets? Why do you think so? If not, what do the rows that are missing from one of the datasets look like in the combined table?

Answer: No, it's likely that not all unique IDs are present in both datasets. This could be as we are dealing with medical or observational data, where not all patients or samples may have complete data across all variables (numerical and categorical). Missing data can occur for several reasons, such as incomplete tests or different recording procedures.

When performing a merge (specifically an "outer" merge that keeps all rows from both datasets), the rows that are missing from one dataset would have NaN (Not a Number) values in the columns corresponding to the dataset they are missing from. For example: If a patient’s numerical data is missing but their categorical data exists, the columns corresponding to numerical values (like blood pressure) would show NaN. Similarly, if categorical data is missing for some patients, the corresponding categorical columns (such as "hypertension") would show NaN values in those rows.

````python
# Merge the two given numerical and categorical datasets based on their unique_ID.
combined_data = pd.merge(numerical_data_cleaned, categorical_data_cleaned, on='unique_id', how='inner')

#Print the combined dataset
print("Combined Dataset:")
print(combined_data.head())
````

## Part 3.3: Rows with Missing values

Removing missing values from a dataset is important for classification because it ensures the model is trained on complete and accurate data, leading to better performance and reliable predictions. Incomplete data can introduce bias and errors, negatively impacting the model's effectiveness.

````python
# Calculate the percentage of rows that contain atleast one missing value

# Print %

missing_percentage = (combined_data.isnull().sum(axis=1) > 0).mean() * 100
print(f"Percentage of rows with missing values: {missing_percentage:.2f}%")

# Drop these rows from the dataset
cleaned_data = combined_data.dropna()
# Print the Dataset
print(cleaned_data)
````

## Part 3.4: Sort the dataset according to the Labels

````python
# Sort the dataset according to the values in 'Target' column. Make sure reset the indices after sorting
sorted_data = cleaned_data.sort_values(by='Target').reset_index(drop=True)

# Print the dataset
print(sorted_data)
````

## Part 3.5: Encoding Categorical data

In this step, we identify and process the categorical columns in the sorted dataset. We map each unique value in these columns to separate "dummy" columns.

For example, the column 'rbc' will be transformed into two columns 'rbc_normal' and 'rbc_abnormal'. If a row's value in 'rbc' is 'normal', the 'rbc_normal' column will be set to 1 and 'rbc_abnormal' will be set to 0.


**Note: Find a correct pandas function to do this **

````python
# Write code here
categorical_columns = ['rbc', 'pc', 'pcc', 'ba', 'htn', 'dm', 'cad', 'appet', 'pe', 'ane', 'Target']
encoded_data = pd.get_dummies(sorted_data, columns=categorical_columns)

# Print the dataset
print(encoded_data)
````

## Part 3.6 : Remove Outliers from Numerical Columns

Outliers can disproportionately influence the fit of a regression model, potentially leading to a model that does not generalize well therefore it is important that we remove outliers from the numerical columns of the dataset.

For this dataset, we define an outlier to be 3 times the standard deviation from the mean. Drop these outliers from the dataset

````python
# Remove outliers
numerical_columns = ['age', 'bp', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wbcc', 'rbcc']

for column in numerical_columns:
    mean = sorted_data[column].mean()
    std = sorted_data[column].std()
    sorted_data = sorted_data[(sorted_data[column] >= mean - 3 * std) & (sorted_data[column] <= mean + 3 * std)]

# Print the dataset
print (sorted_data)
````

## Part 3.7 : Normalize the Numerical Columns

Normalizing numerical attributes ensures that all features contribute equally to the model by scaling them to a consistent range, which improves model performance and convergence. It prevents features with larger scales from disproportionately influencing the model's learning process.

````python
# Normalize the all Numerical Attributes in the dataset.
from sklearn.preprocessing import MinMaxScaler

# Apply MinMaxScaler
scaler = MinMaxScaler()
sorted_data[numerical_columns] = scaler.fit_transform(sorted_data[numerical_columns])

# Print the dataset
# 133 x 36

print(sorted_data[numerical_columns])
````

## Part 3.8: Remove Unnecessary columns

Are there any columns in this dataset which are not appropriate for modeling and predictions? Which column(s)? Justify their exclusion and remove them

Answer: 

Reasons for Dropping unique_id: Unique identifiers do not carry any predictive value. They are arbitrary and have no relationship to the target variable (such as the diagnosis or condition being predicted, e.g., CKD or notCKD). Since unique identifiers are random or sequential, they can't be used in any meaningful calculations (such as averages, summations, or correlations with the target label). Keeping them in the model might introduce noise rather than help the model learn meaningful patterns. Including a unique identifier in a model can result in the model memorizing rather than learning patterns from data. For instance, if a unique ID is included in training, the model might falsely "learn" that a specific ID is associated with a particular outcome, leading to overfitting.

Hence we would drop the unique_id or patient_id column because it does not contribute to predictions or calculations, and its inclusion could lead to noise in the model rather than meaningful insights. 

````python
#Remove that column
merged_data_cleaned = sorted_data.drop(columns=['unique_id'])  # Replace 'patient_id' with the actual column you want to drop

# Print the dataset
print(merged_data_cleaned)
````

## Part 3.9: Export the Cleaned Data

Now that you've completed these cleaning steps you should have a pandas dataframe which is much cleaner and ready for modeling. Our final step is to save our work. Export the DataFrame to a two new formats: csv and json.
