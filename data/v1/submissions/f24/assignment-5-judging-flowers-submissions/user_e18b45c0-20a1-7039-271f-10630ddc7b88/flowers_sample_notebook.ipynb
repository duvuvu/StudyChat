```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
dataset = pd.read_csv('iris.csv')

# Output the first 15 rows of the data
print(dataset.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print(dataset.info())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

Answer:

The features of this dataset are "sepal_length", "sepal_width", "petal_length", and "petal_width", while our label is "species". the "species" column contains 3 distinct values; setosa, virginica, and versicolor, which map to the specified species of flower in real-life.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = dataset[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
Y = dataset['species']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
log_model = LogisticRegression()
log_model.fit(X_train, Y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = X_train.iloc[0].values.reshape(1, -1)
probabilitiy = log_model.predict_proba(sample_data)
print(f"predicted probabilities for the sample datapoint: ", probabilitiy)

# iii. Report on the score for Logistic regression model, what does the score measure?
# Answer: The score measures the accuracy of the model on the test set, meaning the model predicted
# the correct class for the test set since the output is 1.0

test_score = log_model.score(X_test, Y_test)
print(f"Accuracy score of the Logistic Regression model on the test set: {test_score}")

# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = log_model.coef_
intercept = log_model.intercept_
print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc_model = SVC(probability=True)
svc_model.fit(X_train, Y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
svc_predicted_probabilities = svc_model.predict_proba(sample_data)
print(f"Predicted probabilities for the sample datapoint using SVC: {svc_predicted_probabilities}")

# iii. Report on the score for the SVM, what does the score measure?
# Answer: The score measures the accuracy of the model on the test set, meaning the model predicted
# the correct class for the test set since the output is 1.0

svc_test_score = svc_model.score(X_test, Y_test)
print(f"Accuracy score of the SVM model on the test set: {svc_test_score}")
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(hidden_layer_sizes=(40,40,40), max_iter=800, random_state=42)
mlp_model.fit(X_train, Y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
mlp_predicted_probabilities = mlp_model.predict_proba(sample_data)
print(f"Predicted probabilities for the sample datapoint using MLP: {mlp_predicted_probabilities}")

# iii. Report on the score for the Neural Network, what does the score measure?
# Answer: The score measures the accuracy of the model on the test set, meaning the model predicted
# the correct class for the test set since the output is 1.0

mlp_test_score = mlp_model.score(X_test, Y_test)
print(f"Accuracy score of the Neural Network model on the test set: {mlp_test_score}")

# iv: Experiment with different options for the neural network, report on your best configuration
# Answer: I had experimented with the hidden_layer_sizes parameter, and found that 3 hidden layers with 40 neurons each
# is the most optimal configuration that I had found during my experimenting. the max_iter parameter was first set to 100,
# but I found that increasing it to 800 improved the accuracy of the model on the test set, and increasing it further 
# did not improve the accuracy of the model. The random_state parameter was set to 42 to ensure reproducibility of the results.
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
knn_model = KNeighborsClassifier(n_neighbors=2)
knn_model.fit(X_train, Y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
knn_predicted_probabilities = knn_model.predict_proba(sample_data)
print(f"Predicted probabilities for the sample datapoint using kNN: {knn_predicted_probabilities}")

# iii. Report on the score for kNN, what does the score measure?
# Answer: The score measures the accuracy of the model on the test set, meaning the model predicted
# the correct class for the test set since the output is 1.0
knn_test_score = knn_model.score(X_test, Y_test)
print(f"Accuracy score of the kNN model on the test set: {knn_test_score}")
```
