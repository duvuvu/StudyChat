```python
# Imports and pip installations (if needed)
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier

```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
data = pd.read_csv('iris.csv')
# Output the first 15 rows of the data
display(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
data.info()
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The data has the different dimensions for the sepal and petals of a specific flower and classifies it as a type of iris. The features are sepal length, sepal width, petal length, and petal width and the label is the species of flower it is classified as. The label 'species' corresponds to the 3 different species the flower could be. The 3 species are 'setosa', 'versicolor', and 'virginica'.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)

X = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = data['species']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, train_size=0.9, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set

model = LogisticRegression()
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class

sample_dp = pd.DataFrame({'sepal_length': [6.4], 'sepal_width': [2.9], 'petal_length': [5.4], 'petal_width': [1.7]})
probs = model.predict_proba(sample_dp)

print("probability of setosa =", probs[0][0])
print("probability of versicolor =", probs[0][1])
print("probability of virginica =", probs[0][2])

# iii. Report on the score for Logistic regression model, what does the score measure?

train_score = model.score(X_train, y_train)
print("score of training set =", train_score)

# iv. Extract the coefficents and intercepts for the boundary line(s)

coeffs = model.coef_
intercept = model.intercept_

print("coefficients =", coeffs)
print("intercept =", intercept)
```

The score of the logistic regression model measures the mean accuracy of the data.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set

model = svm.SVC(probability=True)
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class

SVM_probs = model.predict_proba(sample_dp)
print("probability of setosa =", SVM_probs[0][0])
print("probability of versicolor =", SVM_probs[0][1])
print("probability of virginica =", SVM_probs[0][2])

# iii. Report on the score for the SVM, what does the score measure?

SVM_score = model.score(X_train, y_train)
print("SVM score =", SVM_score)
```

The score of the model measures the accuracy. The SVM reported a score of .97 which means it correctly classified 97% of the inputs.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set

model = MLPClassifier(hidden_layer_sizes=(100,50), max_iter=1000)
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class

neural_probs = model.predict_proba(sample_dp)
print("probability of setosa =", neural_probs[0][0])
print("probability of versicolor =", neural_probs[0][1])
print("probability of virginica =", neural_probs[0][2])

# iii. Report on the score for the Neural Network, what does the score measure?

neural_score = model.score(X_train, y_train)
print("neural network score =", neural_score)

# iv: Experiment with different options for the neural network, report on your best configuration
model = MLPClassifier(hidden_layer_sizes=(75,75), max_iter=1000)
model.fit(X_train, y_train)

neural_score = model.score(X_train, y_train)
print("new neural network score =", neural_score)

```

The score of the neural network measures the accuracy of the model to correctly classify the input. The first score showed that the model correctly classified 98% of the inputs and the second score showed that the second model correctly classified nearly 99% of the inputs. When I changed the size of the hidden layers, the second model with 2 hidden layers of equal size produced a better score. However, when I ran the code multiple times without changing any of the options, I still received differing outputs.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class

neighbor_probs = model.predict_proba(sample_dp)
print("probability of setosa =", neighbor_probs[0][0])
print("probability of versicolor =", neighbor_probs[0][1])
print("probability of virginica =", neighbor_probs[0][2])

# iii. Report on the score for kNN, what does the score measure?

neighbor_score = model.score(X_train, y_train)
print("K nearest neighbors score =", neighbor_score)
```

The score measures the accuracy of the model to correctly classify the inputs. The score varies as I changed the size of n_neighbors. With a value of 5 for n_neighbors, the model correctly classfied 97% of the inputs.
