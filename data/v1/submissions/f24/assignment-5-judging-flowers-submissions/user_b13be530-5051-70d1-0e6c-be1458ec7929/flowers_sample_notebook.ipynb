# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
iris_data = pd.read_csv('iris.csv')

# Output the first 15 rows of the data
print("First 15 rows of the dataset:")
print(iris_data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print("Summary:")
iris_data.info()
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The dataset contains physical measurements of iris flowers from 3 different species. The features are sepal_length, sepal_width, petal_length, and petal_width. The labels are setosa, versicolor and virginica. The label setosa matches to the class Setosa, the label versicolor matches to the class Versicolor and the label virginica matches to the class Virginica.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = iris_data.drop(columns=['species'])
y = iris_data['species']


# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

print(f"Training set size: {X_train.shape[0]} samples")
print(f"Test set size: {X_test.shape[0]} samples")

```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model = LogisticRegression(max_iter=200)  # Increase max_iter if the model fails to converge
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = X_test.iloc[[0]]  # Using double brackets to keep it as a DataFrame
probabilities = model.predict_proba(sample_datapoint)
print("Predicted probabilities for each class:")
print(probabilities)

# iii. Report on the score for Logistic regression model, what does the score measure?
model_score = model.score(X_test, y_test)
print("Model accuracy score:", model_score)
# the score is 1, this indicates that the model classified all test samples correctly

# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients for each class:")
print(coefficients)

print("Intercept for each class:")
print(intercept)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc_model = SVC(probability=True) 
svc_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = X_test.iloc[[0]]  # Use double brackets to keep it as a DataFrame
probabilities = svc_model.predict_proba(sample_datapoint)
print("Predicted probabilities for each class:")
print(probabilities)


# iii. Report on the score for the SVM, what does the score measure?
model_score = svc_model.score(X_test, y_test)  # This gives the accuracy score
print("SVM model accuracy score:", model_score)
# the score is 1, this indicates that the model classified all test samples correctly
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(max_iter=500, random_state=42)
mlp_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = X_test.iloc[[0]] 
probabilities = mlp_model.predict_proba(sample_datapoint)
print("Predicted probabilities for each class:")
print(probabilities)
    


# iii. Report on the score for the Neural Network, what does the score measure?
model_score = mlp_model.score(X_test, y_test) 
print("Neural Network model accuracy score:", model_score)
# the score is 1, this indicates that the model classified all test samples correctly

# iv: Experiment with different options for the neural network, report on your best configuration
best_score = model_score
best_configuration = mlp_model.hidden_layer_sizes
tested_configurations = {}
for hidden_layer_sizes in [(10,), (50,), (100,), (50, 50), (100, 50)]:
    mlp_test_model = MLPClassifier(hidden_layer_sizes=hidden_layer_sizes, max_iter=500, random_state=42)
    mlp_test_model.fit(X_train, y_train)
    score = mlp_test_model.score(X_test, y_test)
    tested_configurations[hidden_layer_sizes] = score
    
    if score > best_score:
        best_score = score
        best_configuration = hidden_layer_sizes

# Display the results of the configurations
print("\nTested configurations and their scores:")
for config, score in tested_configurations.items():
    print(f"Configuration {config}: Score {score}")

print("\nBest configuration:", best_configuration)
print("Best accuracy score:", best_score)
#Best configuration was (100,) as that was the first one that acheived a score of 1
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
knn_model = KNeighborsClassifier(n_neighbors=3)  
knn_model.fit(X_train, y_train) 
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = X_test.iloc[[0]]  # Use double brackets to keep it as a DataFrame
probabilities = knn_model.predict_proba(sample_datapoint)  # Predict probabilities
print("Predicted probabilities for each class:")
print(probabilities)

# iii. Report on the score for kNN, what does the score measure?
model_score = knn_model.score(X_test, y_test)  # This gives the accuracy score
print("k-Nearest Neighbors model accuracy score:", model_score)
# the score is 1, this indicates that the model classified all test samples correctly
```
