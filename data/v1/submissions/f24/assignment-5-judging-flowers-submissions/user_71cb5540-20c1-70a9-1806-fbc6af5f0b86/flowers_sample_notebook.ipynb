```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier

```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)

# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
df = pd.read_csv('iris.csv')

print("First 15 rows of the dataset:")
print(df.head(15))

# Display a summary of the table information
print("\nSummary of the dataset:")
print(df.info())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The dataset consists of 150 entries (rows) and 5 columns (features), which can be described as follows:
Features:
sepal_length: The length of the sepal of the flower, measured in centimeters (float).
sepal_width: The width of the sepal, measured in centimeters (float).
petal_length: The length of the petal, measured in centimeters (float).
petal_width: The width of the petal, measured in centimeters (float).

Based on the different features, we can clasify it into catagories, Species which is the label. 
Setosa is a type of flower which typically represented by a specific set of measurements. 
We are doing this based on the measurement of size of petals and sepals. 


# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# Split the dataset into features (X) and label (y)
X = df.drop(columns=['species'])  # Features: all columns except 'species'
y = df['species']  # Label: the 'species' column

# Use sklearn to split the features and labels into a training/test set (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Output the dimensions of the splits to verify
print(f'Training set size: {X_train.shape[0]} samples')
print(f'Test set size: {X_test.shape[0]} samples')
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set

# ii. For a sample datapoint, predict the probabilities for each possible class

# iii. Report on the score for Logistic regression model, what does the score measure?

# iv. Extract the coefficents and intercepts for the boundary line(s)

# i. Train a Logistic Regression model on the training set
model = LogisticRegression(multi_class='auto', solver='lbfgs', max_iter=200)
model.fit(X_train, y_train)

# ii. Predict the probabilities for a sample datapoint
sample_datapoint = X_test.iloc[12].values.reshape(1, -1)  # Get the 13th row of the test set as a sample
probabilities = model.predict_proba(sample_datapoint)

# iii. Report on the score for the Logistic Regression model
y_pred = model.predict(X_test)  # Make predictions on the test set
score = accuracy_score(y_test, y_pred)  # Calculate accuracy score
print(sample_datapoint)

print(f'Predicted probabilities for the sample data point: {probabilities}')
print(f'Accuracy score of the Logistic Regression model: {score}')

# iv. Extract the coefficients and intercepts for the boundary line(s)
coefficients = model.coef_  # Coefficients for each class
intercept = model.intercept_  # Intercepts for each class

print('Coefficients for each class:')
print(coefficients)
print('Intercepts for each class:')
print(intercept)
```

- Predictions on the test set are made using `model.predict(X_test)`.
- The accuracy score is calculated as the ratio of correctly predicted samples to the total number of samples in the test set using `accuracy_score(y_test, y_pred)`. Over here, since we have a score of 1, all the predictions were correct. For all the predictions and classification, the first one is setosa, second is vesicolor and third is virginica. this shows the intercepts for each class. 

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set

# ii. For a sample datapoint, predict the probabilities for each possible class

# iii. Report on the score for the SVM, what does the score measure?
X_train_svm, X_test_svm, y_train_svm, y_test_svm = train_test_split(X, y, test_size=0.1, random_state=42)

# i. Train a Support Vector Classifier on the training set
model = SVC(probability=True, random_state=42)  # Enable probability estimates
model.fit(X_train_svm, y_train_svm)

# ii. Predict the probabilities for a sample datapoint
sample_datapoint = X_test_svm.iloc[12].values.reshape(1, -1)  # Get the 13th row of the test set as a sample
probabilities = model.predict_proba(sample_datapoint)

# iii. Report on the score for the SVM
y_pred = model.predict(X_test_svm)  # Make predictions on the test set
score = accuracy_score(y_test_svm, y_pred)  # Calculate accuracy score

print(f'Predicted probabilities for the sample data point: {probabilities}')
print(f'Accuracy score of the Support Vector Classifier: {score}')
```

The accuracy score ranges from 0.0 to 1.0, where:
1.0 indicates perfect classification (all instances correctly predicted).
0.0 indicates that none of the instances have been predicted correctly.
In practice, the accuracy score is a straightforward measure of model performance. Over here, the SVC shows that the model is accurate, showing that it predicts correctly 100% of the time. It also says that there is a 97% probability that the flower is setosa which is the correct prediction. 

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set

# ii. For a sample datapoint, predict the probabilities for each possible class

# iii. Report on the score for the Neural Network, what does the score measure?

# iv: Experiment with different options for the neural network, report on your best configuration

# Split into training and testing sets (90% train, 10% test)
X_train_nn, X_test_nn, y_train_nn, y_test_nn = train_test_split(X, y, test_size=0.1, random_state=42)

# i. Train a Neural Network (MLP Classifier) on the training set
# You can modify 'hidden_layer_sizes', 'activation', 'solver', etc. as part of the experimentation.
model = MLPClassifier(hidden_layer_sizes=(50, 30), activation='relu', solver='adam', max_iter=400, random_state=42)
model.fit(X_train_nn, y_train_nn)

# ii. Predict the probabilities for a sample datapoint
sample_datapoint = X_test_nn.iloc[0].values.reshape(1, -1)  # Get the first row of the test set as a sample
probabilities = model.predict_proba(sample_datapoint)

# iii. Report on the score for the Neural Network
y_pred = model.predict(X_test_nn)  # Make predictions on the test set
score = accuracy_score(y_test_nn, y_pred)  # Calculate accuracy score

print(f'Predicted probabilities for the sample data point: {probabilities}')
print(f'Accuracy score of the Neural Network (MLP Classifier): {score}')

# iv: Experiment with different options for the neural network
# Trying different configurations
best_score = 0
best_config = {}
configs = [
    {'hidden_layer_sizes': (50,), 'activation': 'relu', 'max_iter': 100},
    {'hidden_layer_sizes': (60,), 'activation': 'tanh', 'max_iter': 200},
    {'hidden_layer_sizes': (50, 50), 'activation': 'logistic', 'max_iter': 400},
    {'hidden_layer_sizes': (30, 30, 30), 'activation': 'relu', 'max_iter': 100},
    {'hidden_layer_sizes': (100, 50), 'activation': 'relu', 'max_iter': 80}
]

for config in configs:
    model = MLPClassifier(hidden_layer_sizes=config['hidden_layer_sizes'], 
                          activation=config['activation'], 
                          max_iter=config['max_iter'], 
                          random_state=42)
    
    model.fit(X_train_nn, y_train_nn)
    y_pred = model.predict(X_test_nn)
    score = accuracy_score(y_test_nn, y_pred)
    print(f'Score{score}')
    print(f'config{config}')
    if score > best_score:
        best_score = score
        best_config = config

print(f'Best configuration: {best_config}')
print(f'Best accuracy score: {best_score}')
```

I tried multiple different combinations of hidden layer sizes, neural netwoks and the amount of times back propogation is allowed. From what I saw, sigmoid which is logistic did not work very well. Relu worked well however only with many hidden layers and back propogation. I belive that tan worked best, with minimal layers and lesser back propogation, giving accuracy scores of 1.0

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

# ii. For a sample datapoint, predict the probabilities for each possible class

# iii. Report on the score for kNN, what does the score measure?

X_train_kn, X_test_kn, y_train_kn, y_test_kn = train_test_split(X, y, test_size=0.1, random_state=42)

# i. Train a k-Neighbors Classifier (Note: this will load the data into the model)
model = KNeighborsClassifier(n_neighbors=3)  # Using 6 neighbors
model.fit(X_train_kn, y_train_kn)  # Fit loads the training data

# ii. Predict the probabilities for a sample datapoint
sample_datapoint = X_test_kn.iloc[0].values.reshape(1, -1)  # Get the first row of the test set as a sample
probabilities = model.predict_proba(sample_datapoint)

# iii. Report on the score for k-NN
y_pred = model.predict(X_test_kn)  # Make predictions on the test set
score = accuracy_score(y_test_kn, y_pred)  # Calculate accuracy score
print(sample_datapoint)
print(f'Predicted probabilities for the sample data point: {probabilities}')
print(f'Accuracy score of the k-NN Classifier: {score}')
```

The accuracy score is a measure of how well the model correctly predicts the labels for the test dataset. It ranges from 0.0 to 1.0:
1.0 means every prediction is correct.
0.0 means none of the predictions are correct.

I tried different neighbor sizes, and found that 3 was one of the numbers  that gave it an accuracy score of 1.0
