```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import *
from sklearn.linear_model import *
from sklearn.preprocessing import *
from sklearn.svm import *
from sklearn.neural_network import *
from sklearn.neighbors import *
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
iris_dataset = pd.read_csv("iris.csv")
# Output the first 15 rows of the data
print("Iris Dataset:")
print(iris_dataset.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print("\nSummary:")
print(iris_dataset.info())

unique_species = iris_dataset['species'].unique()
print("\nUnique Species:")
print(unique_species)
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The data of 150 entries gives the lengths and widths of both the sepal and petals of 3 different kinds of similarly looking flowers. There are 4 features: sepal_length, sepal_width, petal_length, and petal_width. The label is what kind of species of flower each row of data corresponds to and is found in the "species" column. There are three different species which are setosa, versicolor, and virginica. 

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
features = iris_dataset[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
label = iris_dataset["species"]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(features, label, test_size=0.3, train_size=0.7, random_state=13)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
logistic_model = LogisticRegression(max_iter=200)
logistic_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data_point = pd.DataFrame(X_test.iloc[0].values.reshape(1, -1), columns=X_train.columns)
logistic_probabilities = logistic_model.predict_proba(sample_data_point)
class_labels = logistic_model.classes_

print("Predicted probabilities for each class:")
for class_label, probability in zip(class_labels, logistic_probabilities[0]):
    print(f"{class_label}: {probability:.5f}")

# iii. Report on the score for Logistic regression model, what does the score measure?
accuracy_score = logistic_model.score(X_test, y_test)
print(f"\nAccuracy score of the logistic regression model: {accuracy_score}")

# iv. Extract the coefficents and intercepts for the boundary line(s)
feature_names = features.columns
coefficients_df = pd.DataFrame(logistic_model.coef_, columns=feature_names, index=class_labels)
print("\nCoefficients for each feature:")
print(coefficients_df)

intercept_df = pd.DataFrame(logistic_model.intercept_, index=class_labels, columns=['Intercept'])
print("\nIntercepts for each class:")
print(intercept_df)

# Use the cross_val_score function to repeat your experiment across many shuffles of the data
reg_cv_score = cross_val_score(logistic_model, features, label, cv=10)
print("\nCross-validation scores:")
print(reg_cv_score)

# Report on their finding and their significance
print("\nMean:")
print(reg_cv_score.mean())
print("\nStandard deviation:")
print(reg_cv_score.std())
```

The score function for the logistic regression model returns the mean accuracy on the given test data and labels. In multi-label classification, this is the subset accuracy which is a harsh metric since you require for each sample that each label set be correctly predicted. The score of 0.97778 for the regression model means that the model is performing very well and is correctly classifying about 97% of the data points.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svm_model = SVC(probability=True)
svm_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
svm_probabilities = svm_model.predict_proba(sample_data_point)

print("Predicted probabilities for each class:")
for class_label, probability in zip(class_labels, svm_probabilities[0]):
    print(f"{class_label}: {probability:.5f}")

# iii. Report on the score for the SVM, what does the score measure?
svm_score = svm_model.score(X_test, y_test)
print(f"\nAccuracy score of the Support Vector Classifier: {svm_score}")

# Use the cross_val_score function to repeat your experiment across many shuffles of the data
svm_cv_score = cross_val_score(svm_model, features, label, cv=10)
print("\nCross-validation scores:")
print(svm_cv_score)

# Report on their finding and their significance
print("\nMean:")
print(svm_cv_score.mean())
print("\nStandard deviation:")
print(svm_cv_score.std())
```

The accuracy score measures the proportion of correctly predicted instances to the total instances in the test set. Since the SVM model got a score of 0.95556, it means the model is performing pretty well and is correctly predicting the class of about 96% of the data points in the test set.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(max_iter=1000, random_state=42)
mlp_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
mlp_probabilities = mlp_model.predict_proba(sample_data_point)

print("Predicted probabilities for each class:")
for class_label, probability in zip(class_labels, mlp_probabilities[0]):
    print(f"{class_label}: {probability:.5f}")

# iii. Report on the score for the Neural Network, what does the score measure?
mlp_score = mlp_model.score(X_test, y_test)
print(f"\nAccuracy score of the Neural Network: {mlp_score}\n")

# Use the cross_val_score function to repeat your experiment across many shuffles of the data
mlp_cv_score = cross_val_score(mlp_model, features, label, cv=10)
print("\nCross-validation scores:")
print(mlp_cv_score)

# Report on their finding and their significance
print("\nMean:")
print(mlp_cv_score.mean())
print("\nStandard deviation:")
print(mlp_cv_score.std())
print()

# iv: Experiment with different options for the neural network, report on your best configuration
best_accuracy = 0
best_config = None
configurations = [
    {"hidden_layer_sizes": (10,), "activation": "relu"},
    {"hidden_layer_sizes": (20,), "activation": "relu"},
    {"hidden_layer_sizes": (10, 10), "activation": "relu"},
    {"hidden_layer_sizes": (20, 20), "activation": "relu"},
    {"hidden_layer_sizes": (10,), "activation": "logistic"},
    {"hidden_layer_sizes": (10, 10), "activation": "logistic"},
    {"hidden_layer_sizes": (30, 30, 30), "activation": "tanh"}
]

for config in configurations:
    mlp_model = MLPClassifier(hidden_layer_sizes=config["hidden_layer_sizes"],
                               activation=config["activation"],
                               max_iter=10000,
                               random_state=42)
    mlp_model.fit(X_train, y_train)
    score = mlp_model.score(X_test, y_test)
    print(f"Configuration: {config}, Score: {score:.5f}")
    
    if score > best_accuracy:
        best_accuracy = score
        best_config = config

print(f"\nBest Configuration: {best_config} with Accuracy Score: {best_accuracy:.5f}")
```

Similarly to the above two classifying models, the neaural network also got a score of 0.97778 which means that it is correctly classifying almost every data point in the test set. This means the model is well trained for this data.

With the testing of a few more neural networks, the best performing setup seemed to be when the hidden layer size was 10 and the activation was set to logistic. This model performed at the highest level and got a score of 1.0 which means that it was able to classify 100% of the data correctly which is very impressive.

# Part 6: K-Nearest Neighbors

The KNN model has the same lowest score as the SVM. The score is 0.95556 and while it is correctly classifying about 96% of the data, it is still performing at a lower rate than logistic regression and the neural network.
