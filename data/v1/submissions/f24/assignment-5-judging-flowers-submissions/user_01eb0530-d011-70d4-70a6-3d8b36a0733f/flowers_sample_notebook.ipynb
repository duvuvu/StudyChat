```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
data = pd.read_csv('iris.csv')

# Output the first 15 rows of the data
print("First 15 rows of data:")
print(data.head(15))

# Display a summary of the table information (number of datapoints, etc.)
print("\nSummary of table information:")
print(data.info())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The data specifies a number of qualities for many different iris flowers, noting various anatomical iris measurements (sepal length/width and petal length/width), as well as the type of iris species. The features are sepal_length, sepal_width, petal_length, and petal_width. The label is simply the species. The mapping of the label (species) to the actual classes is as follows. Class O corresponds to setosa, class 1 corresponds to versicolor, and class 2 corresponds to virginica.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = data['species']

print("Features (X):")
print(X)
print("\nLabel (y):")
print(y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=56)

print("\nX training set:")
print(X_train)
print("\nX testing set:")
print(X_test)
print("\ny training set:")
print(y_train)
print("\ny testing set:")
print(y_test)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
label_encoder = LabelEncoder()
y_train_encoded = label_encoder.fit_transform(y_train)
y_test_encoded = label_encoder.transform(y_test)

model = LogisticRegression(max_iter=200)  
model.fit(X_train, y_train_encoded)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_point = X_test.iloc[[0]] 
probabilities = model.predict_proba(sample_point)

print("Sample datapoint probabilities for each possible class:")
print(probabilities)

# iii. Report on the score for Logistic regression model, what does the score measure?
# SEE MARKDOWN BELOW FOR EXPLANATION OF SCORE
y_pred = model.predict(X_test)
score = accuracy_score(y_test_encoded, y_pred)

print("\nLogistic regression model score:")
print(score)

# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = model.coef_
intercepts = model.intercept_

print("\nBoundary line coefficients:")
print(coefficients)
print("\nBoundary line intercepts:")
print(intercepts)
```

Boundary line for class 0: $h(x_1, x_2, x_3, x_4) = 10.1645 - 0.4559 \cdot x_1 + 0.8991 \cdot x_2 - 2.5016 \cdot x_3 - 1.0112 \cdot x_4$

Boundary line for class 1: $h(x_1, x_2, x_3, x_4) = 1.9198 + 0.5667 \cdot x_1 - 0.2604 \cdot x_2 - 0.2695 \cdot x_3 - 0.8326 \cdot x_4$

Boundary line for class 2: $h(x_1, x_2, x_3, x_4) = -12.0843 - 0.1108 \cdot x_1 - 0.6387 \cdot x_2 + 2.7711 \cdot x_3 + 1.8438 \cdot x_4$



$x_1 = \text{sepal length}$

$x_2 = \text{sepal width}$

$x_3 = \text{petal length}$

$x_4 = \text{petal width}$


The score for the logistic regression model is 0.867. Since this value is close to 1 (there is a strong correlation between the features and label), it means that the model performs relatively well in terms of making predictions about the species of a given iris based off of its measurements; the model fits the data well. About 86.7% of the predictions were correct.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svm_model = SVC(probability=True)
svm_model.fit(X_train, y_train_encoded)

# ii. For a sample datapoint, predict the probabilities for each possible class
svm_sample_point = X_test.iloc[[0]] 
svm_probabilities = svm_model.predict_proba(svm_sample_point)

print("Sample datapoint probabilities for each possible class:")
print(svm_probabilities)

# iii. Report on the score for the SVM, what does the score measure?
#SEE MARKDOWN BELOW FOR EXPLANATION OF SCORE
svm_y_pred = svm_model.predict(X_test)
svm_score = accuracy_score(y_test_encoded, svm_y_pred)

print("\nSVM score:")
print(svm_score)
```

The score for the SVM is 0.933. Since this value is close to 1 (there is a strong correlation between the features and label), it means that the model performs very well in terms of making predictions about the species of a given iris based off of its measurements; the model fits the data well. About 93.3% of the predictions were correct.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(max_iter=1000, random_state=56) 
mlp_model.fit(X_train, y_train_encoded)

# ii. For a sample datapoint, predict the probabilities for each possible class
mlp_sample_point = X_test.iloc[[0]] 
mlp_probabilities = mlp_model.predict_proba(mlp_sample_point)

print("Sample datapoint probabilities for each possible class:")
print(mlp_probabilities)

# iii. Report on the score for the Neural Network, what does the score measure?
#SEE MARKDOWN BELOW FOR EXPLANATION OF SCORE
mlp_y_pred = mlp_model.predict(X_test)
mlp_score = accuracy_score(y_test_encoded, mlp_y_pred)

print("\nNeural Network score:")
print(mlp_score)

# iv: Experiment with different options for the neural network, report on your best configuration
#SEE MARKDOWN BELOW FOR REPORT ON BEST CONFIGURATION    
best_score = 0
best_config = None

print("\nTesting configurations...")
for hidden_layers in [(10,), (20,), (10, 10), (20, 20), (10, 10, 10)]:
    for activation in ['relu', 'tanh']:
        temp_mlp = MLPClassifier(hidden_layer_sizes=hidden_layers, activation=activation, max_iter=2000, random_state=56, early_stopping=True, validation_fraction=0.1, n_iter_no_change=10)
        temp_mlp.fit(X_train, y_train_encoded)
        
        temp_y_pred = temp_mlp.predict(X_test)
        temp_score = accuracy_score(y_test_encoded, temp_y_pred)
        
        print(f"Configuration: hidden_layers={hidden_layers}, activation='{activation}' - Accuracy: {temp_score}")
        
        if temp_score > best_score:
            best_score = temp_score
            best_config = (hidden_layers, activation)

print(f"\nBest configuration: Hidden layers={best_config[0]}, Activation='{best_config[1]}' - Score: {best_score}")
```

The score for the first neural network is 0.933. Since this value is close 1 (there is a strong correlation between the features and label), it means that the model performs well in terms of making predictions about the species of a given iris based off of its measurements; the model fits the data very well. About 93.3% of the predictions were correct.

In order to experiment with different options for the neural network, a for loop was utilized to iterate through various configurations with different activation functions and sizes of hidden layers. Each of these configurations was evaluated based on the model score. Some models scored as low as 0.267, 0.333, and 0.4, while some had higher scores of 0.8 and 0.933. Ultimately, the best configuration of the options explored utilized the tanh activation function and had 3 hidden layers with 10 nodes each. This configuration had a model score of 0.933. Since this value is close to 1 (there is a strong correlation between the features and label), it means that this configuration performs well in terms of making predictions about the species of a given iris based off of its measurements; the model fits the data well. About 93.3% of the predictions were correct.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn_model = KNeighborsClassifier(n_neighbors=60) 
knn_model.fit(X_train, y_train_encoded) 

# ii. For a sample datapoint, predict the probabilities for each possible class
knn_sample_point = X_test.iloc[[0]]  # Use double brackets to retain DataFrame structure
knn_probabilities = knn_model.predict_proba(knn_sample_point)

print("Sample datapoint probabilities for each possible class:")
print(knn_probabilities)

# iii. Report on the score for kNN, what does the score measure?
#SEE MARKDOWN BELOW FOR EXPLANATION OF SCORE
knn_y_pred = knn_model.predict(X_test)
knn_score = accuracy_score(y_test_encoded, knn_y_pred)

print("\nkNN score:")
print(knn_score)
```

The score for the kNN is 0.8. Since this value is relatively close to 1 (there is a strong correlation between the features and label), it means that the model performs relatively well in terms of making predictions about the species of a given iris based off of its measurements; the model fits the data well. About 80,0% of the predictions were correct.
