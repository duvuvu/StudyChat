```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.datasets import load_iris
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
iris = load_iris()
data = pd.read_csv('iris.csv')
data['target'] = iris.target


# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
data.head(15)
```

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
from sklearn.model_selection import train_test_split

X = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = data['species']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

# Part 3: Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
# i. Use sklearn to train a LogisticRegression model on the training set
logistic_model = LogisticRegression(max_iter=1000, random_state=42)
logistic_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = X_test.iloc[0:1]
class_probabilities = logistic_model.predict_proba(sample_data)
print("Class Probabilities for sample data: ", class_probabilities)

# iii. Report on the score for Logistic regression model, what does the score measure?
logistic_score = logistic_model.score(X_test, y_test)
print("Logistic Regression Model Score: ", logistic_score)

# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficents = logistic_model.coef_
intercepts = logistic_model.intercept_
print("Coefficients: ", coefficents)
print("Intercepts: ", intercepts)
```

# Part 4: Support Vector Machine

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# i. Use sklearn to train a Support Vector Classifier on the training set
svm_model = SVC(kernel='linear',probability=True, random_state=42)
svm_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = X_test.iloc[0:1]
class_probabilities = svm_model.predict_proba(sample_data)
print("Class Probabilities for Sample Data (SVM): ", class_probabilities)

# iii. Report on the score for the SVM, what does the score measure?
svm_score = svm_model.score(X_test, y_test)
print("SVM Model Score: ", svm_score) 
```

# Part 5: Neural Network

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score

# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, random_state=42)
mlp_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = X_test.iloc[0:1]  # Take the first sample from the test set
class_probabilities_mlp = mlp_model.predict_proba(sample_data)
print("Class Probabilities for Sample Data (Neural Network):", class_probabilities_mlp)

# iii. Report on the score for the Neural Network, what does the score measure?
mlp_score = mlp_model.score(X_test, y_test)
print("Neural Network Score:", mlp_score)

# Explanation: The score here measures the accuracy of the Neural Network model,
# which is the proportion of correct predictions on the test set.

# iv: Experiment with different options for the neural network, report on your best configuration
# Testing different configurations for hidden layers and activation functions
best_score = 0
best_config = {}

# Using valid layer configurations
for hidden_layers in [(5,), (10,), (10, 5), (20, 10)]:
    for activation in ['relu', 'tanh', 'logistic']:
        model = MLPClassifier(hidden_layer_sizes=hidden_layers, activation=activation, max_iter=1000, random_state=42)
        model.fit(X_train, y_train)
        score = model.score(X_test, y_test)
        print(f"Configuration: {hidden_layers} layers, Activation: {activation}, Score: {score}")
        
        # Update the best configuration if current score is higher
        if score > best_score:
            best_score = score
            best_config = {'hidden_layers': hidden_layers, 'activation': activation}

print("\nBest Configuration:", best_config)
print("Best Score:", best_score)
```

# Part 6: K-Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)


# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = X_test.iloc[0:1]
class_probabilities_knn = knn_model.predict_proba(sample_data)
print("Class Probabilities for Sample Data (kNN): ", class_probabilities_knn)

# iii. Report on the score for kNN, what does the score measure?
knn_score = knn_model.score(X_test, y_test)
print("knn Model Score: ", knn_score)

# Explanation: The score here measures the accuracy of the k-NN model,
# which is the proportion of correct predictions on the test set.
```
