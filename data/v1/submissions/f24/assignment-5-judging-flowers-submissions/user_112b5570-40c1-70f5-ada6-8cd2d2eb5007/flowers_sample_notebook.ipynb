```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
iris_dataset = pd.read_csv("iris.csv")

# Output the first 15 rows of the data
iris_dataset.head(15)

# Display a summary of the table information (number of datapoints, etc.)
#iris_dataset.info()
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

Each row of the data represents the characteristics of each flower found in the iris dataset. The features of the dataset are the sepal_length, sepal_width, petal_length, petal_width and the label is the species. Upon looking up the scikit documentation of the dataset, there are three possible classes for each row: setosa, versicolor, and virginica. Each of the labels corresponds to the distinct species of the iris flower (Iris setosa, Iris versicolor and Iris virginica)

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = iris_dataset.iloc[:,:-1] # our feature columns (as a Dataframe)
y = iris_dataset.iloc[:,-1] # our label columns (as a Series)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=41)
X_test
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
clf_logreg = LogisticRegression().fit(X=X_train, y=y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
print("Probabilities for a random datapoint: ", clf_logreg.predict_proba(X_test.sample())) # returns [[9.81141289e-01 1.88586775e-02 3.30699410e-08]], for setosa, versicolor, virginica respectively

# iii. Report on the score for Logistic regression model, what does the score measure?
print("Score:", clf_logreg.score(X_test, y_test))
print("This score measures the mean accuracy on the given train data and labels. A number closer to 1 is more accurate.")

# iv. Extract the coefficents and intercepts for the boundary line(s)
print("Boundary line coefficients are", clf_logreg.coef_)
print("Boundary line intercepts are", clf_logreg.intercept_)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
clf_svm = make_pipeline(StandardScaler(), SVC(gamma="auto", probability=True))
clf_svm.fit(X=X_train, y=y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
print("Probabilities for a random datapoint: ", clf_svm.predict_proba(X_test.sample()))


# iii. Report on the score for the SVM, what does the score measure?
print("Score:", clf_svm.score(X_test, y_test))
print("This score measures the mean accuracy on the given train data and labels. A number closer to 1 is more accurate.")

```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp = MLPClassifier(random_state=5, solver="adam")
clf_mlp = mlp.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
print("Probabilities for a random datapoint: ", clf_mlp.predict_proba(X_test.sample()))

# iii. Report on the score for the Neural Network, what does the score measure?
print("Score (solver=\"adam\"):", clf_mlp.score(X_test, y_test))
print("This score measures the mean accuracy on the given train data and labels. A number closer to 1 is more accurate. I don't know why I have 1.0!!!")


# iv: Experiment with different options for the neural network, report on your best configuration
mlp = MLPClassifier(random_state=5, solver="lbfgs", max_iter=500)
clf_mlp = mlp.fit(X_train, y_train)
print("Score (solver=\"lbfgs\"): " , clf_mlp.score(X_train, y_train))
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

neigh = KNeighborsClassifier(n_neighbors=6)
neigh.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
print("Probabilities for a random datapoint: ", neigh.predict_proba(X_test.sample()))
# iii. Report on the score for kNN, what does the score measure?
print("Score:", neigh.score(X_test, y_test))
print("This score measures the mean accuracy on the given train data and labels. A number closer to 1 is more accurate. I finally have a number that isn't 1 but close to it, yay!")

```
