# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
from sklearn.model_selection import train_test_split
X = df.iloc[:,:-1]
Y = df.iloc[:,-1]

#Y.columns
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.10,train_size= 0.90, random_state=42)

```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler


scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train) 
x_test_scaled = scaler.transform(x_test) 

model = LogisticRegression(random_state=42) 
model.fit(x_train_scaled, y_train)  

# ii. For a sample datapoint, predict the probabilities for each possible class

sample_data = scaler.transform(pd.DataFrame({'sepal_length': [5.1], 'sepal_width' : [3.5], 'petal_length' : [1.4], 'petal_width' : [0.2]}))

probabilities = model.predict_proba(sample_data)[0]

classes = model.classes_
for i in range(0,3):
    print(f'Class {classes[i]}: {probabilities[i]:.5f}')


# iii. Report on the score for Logistic regression model, what does the score measure?

y_pred = model.predict(x_test_scaled)  
score = accuracy_score(y_test, y_pred)
print("Accuracy score:", score)

print("The accuracy score from the Logistic regression model is 1.0 which means that its correctly predicting the label (Iris type), based on its features(Length and width of sepal and petal), for all of the data points")






# iv. Extract the coefficents and intercepts for the boundary line(s)

coeff = model.coef_
inter = model.intercept_
print("Coefficients:", coeff)
print("Intercept:", inter)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
from sklearn.svm import SVC

scaler = StandardScaler() 
x_train_scaled = scaler.fit_transform(x_train) 
x_test_scaled = scaler.transform(x_test)  

modelSVC = SVC(probability=True)  
modelSVC.fit(x_train_scaled, y_train)


# ii. For a sample datapoint, predict the probabilities for each possible class

sample_dataSVC = scaler.transform(pd.DataFrame({'sepal_length': [5.1], 'sepal_width' : [3.5], 'petal_length' : [1.4], 'petal_width' : [0.2]}))

probabilitiesSVC = modelSVC.predict_proba(sample_dataSVC)[0]

classesSVC = modelSVC.classes_
for i in range(0,3):
    print(f'Class {classesSVC[i]}: {probabilitiesSVC[i]:.5f}')


# iii. Report on the score for the SVM, what does the score measure?
y_pred = modelSVC.predict(x_test_scaled) 
score = accuracy_score(y_test, y_pred)
print("Accuracy score", score)
print("The accuracy score from the SVM model is 1.0 which means that its correctly predicting the \nlabel (Iris type), based on its features(Length and width of sepal and petal), for all of the data points")


```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
from sklearn.neural_network import MLPClassifier
scaler = StandardScaler() 
x_train_scaled = scaler.fit_transform(x_train) 
x_test_scaled = scaler.transform(x_test)  

modelNN = MLPClassifier(activation='relu', solver='adam', random_state=42, max_iter=10000) 
modelNN.fit(x_train_scaled, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class

sample_dataNN = scaler.transform(pd.DataFrame({'sepal_length': [5.1], 'sepal_width' : [3.5], 'petal_length' : [1.4], 'petal_width' : [0.2]}))

probabilitiesNN = modelNN.predict_proba(sample_dataNN)[0]

classesNN = modelNN.classes_
for i in range(0,3):
    print(f'Class {classesNN[i]}: {probabilitiesNN[i]:.5f}')

# iii. Report on the score for the Neural Network, what does the score measure?
y_pred = modelNN.predict(x_test_scaled)  
score = accuracy_score(y_test, y_pred)
print("Accuracy score:", score)
print("The accuracy score from the NN/MLP model is 1.0 which means that its correctly predicting \nthe label (Iris type), based on its features(Length and width of sepal and petal), for all of the data points\n\n\n\n")



# iv: Experiment with different options for the neural network, report on your best configuration

best_score = 0
best_configs =[]
best_config = None

#configs by 383 gpt!
configurations = [(1,),
                 (5,), 
                 (10,), 
                 (50,), 
                 (100,), 
                 (125,),
                 (100, 50), 
                 (150, 100), 
                 (100, 100, 50)]
configResults = []


for config in configurations:
    modelNN1 = MLPClassifier(hidden_layer_sizes=config,
                             activation='relu',  
                             solver='adam',
                             random_state=42, max_iter=10000)

 
    modelNN1.fit(x_train_scaled, y_train)

    
    y_pred = modelNN1.predict(x_test_scaled)
    
    
    score1 = accuracy_score(y_test, y_pred)
    
    
    configResults.append((config,round(score1,5)))
    #print(f"Configuration: {config}, Accuracy score: {score1:.5f}")

    if score1 > best_score:
        best_score = score1
        best_config = config
    elif score1 == best_score:
        if len(best_configs)==0:
            best_configs.append(config)
            best_configs.append(best_config)
        else:
            best_configs.append(config)
    
print(configResults)
\
if len(best_configs) == 0:
    print(f"\nBest Configuration: {best_config}, Best Accuracy score: {best_score}")
else:
    print(f"\nBest Configurations: {best_configs}, Best Accuracy score: {best_score}")

```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier

# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

from sklearn.neighbors import KNeighborsClassifier


# i. Use sklearn to 'train' a k-Neighbors Classifier
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)  
x_test_scaled = scaler.transform(x_test)  


modelKNN = KNeighborsClassifier(n_neighbors=5) 
modelKNN.fit(x_train_scaled, y_train) 

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_dataKNN = scaler.transform(pd.DataFrame({'sepal_length': [5.1], 'sepal_width' : [3.5], 'petal_length' : [1.4], 'petal_width' : [0.2]}))

probabilitiesKNN= modelKNN.predict_proba(sample_dataKNN)[0]

classesKNN = modelKNN.classes_
for i in range(0,3):
    print(f'Class {classesKNN[i]}: {probabilitiesKNN[i]:.5f}')



# iii. Report on the score for kNN, what does the score measure?
y_pred = modelKNN.predict(x_test_scaled)  
score = accuracy_score(y_test, y_pred)
print("Accuracy score:", round(score,4))
print("The accuracy score from the Knn model is 1.0 which means that its correctly predicting \nthe label (Iris type), based on its features(Length and width of sepal and petal), for all f the data points\n\n\n\n")

```
