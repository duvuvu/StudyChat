```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
iris_df = pd.read_csv('iris.csv')  # Update the path to your local file

# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

# Display the first 15 rows of the dataset
print("First 15 Rows of the Iris Dataset:")
print(iris_df.head(15))

# Display a summary of the table information
print("\nSummary of the Dataset Information:")
print(iris_df.info())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The Iris dataset contains 150 samples of iris flowers characterized by four features: sepal length, sepal width, petal length, and petal width. The target variable (label) represents the species of the iris, which is encoded as integers: 0 (Iris Setosa), 1 (Iris Versicolor), and 2 (Iris Virginica). This dataset is widely used for testing classification algorithms due to its simplicity and distinct classes.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)

# Load the Iris dataset
iris = load_iris()

# Convert to DataFrame for easier handling if needed
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['species'] = iris.target

# Features (X) and Labels (y)
X = iris_df[iris.feature_names]  # All feature columns
y = iris_df['species']            # Target column

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
# Split the dataset into training and testing sets (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Display the shapes of the resulting datasets
print(f'Training feature set shape: {X_train.shape}')
print(f'Testing feature set shape: {X_test.shape}')
print(f'Training label set shape: {y_train.shape}')
print(f'Testing label set shape: {y_test.shape}')
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Initialize the Logistic Regression model
logistic_model = LogisticRegression(max_iter=200)

# Train the model on the training set
logistic_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
# Select a sample datapoint from the test set (e.g., first observation)
sample_data_point = X_test.iloc[0].values.reshape(1, -1)

# Predict probabilities for each possible class
predicted_probabilities = logistic_model.predict_proba(sample_data_point)

print("Predicted probabilities for the classes:")
print(predicted_probabilities)

# iii. Report on the score for Logistic regression model, what does the score measure?
# Predict the test set results
y_pred_logistic = logistic_model.predict(X_test)

# Calculate the accuracy score
accuracy_logistic = accuracy_score(y_test, y_pred_logistic)

print(f'Logistic Regression Model Accuracy: {accuracy_logistic}')

# iv. Extract the coefficents and intercepts for the boundary line(s)
# Extract the coefficients and intercept
coefficients = logistic_model.coef_
intercept = logistic_model.intercept_

print("Coefficients for each feature:")
print(coefficients)

print("Intercept:")
print(intercept)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
from sklearn.svm import SVC

# Initialize the Support Vector Classifier
svm_model = SVC(probability=True)  # Enable probability estimates

# Train the model on the training set
svm_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
# Select a sample data point from the test set (e.g., first observation)
sample_data_point = X_test.iloc[0].values.reshape(1, -1)

# Predict probabilities for each possible class
predicted_probabilities_svm = svm_model.predict_proba(sample_data_point)

print("Predicted probabilities for the classes:")
print(predicted_probabilities_svm)

# iii. Report on the score for the SVM, what does the score measure?
from sklearn.metrics import accuracy_score

# Predict the test set results
y_pred_svm = svm_model.predict(X_test)

# Calculate the accuracy score
accuracy_svm = accuracy_score(y_test, y_pred_svm)

print(f'Support Vector Classifier Model Accuracy: {accuracy_svm}')
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
from sklearn.neural_network import MLPClassifier

# Initialize the MLP Classifier
mlp_model = MLPClassifier(max_iter=200, random_state=42)  # You can adjust max_iter for convergence

# Train the model on the training set
mlp_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
# Select a sample data point from the test set (e.g., first observation)
sample_data_point = X_test.iloc[0].values.reshape(1, -1)

# Predict probabilities for each possible class
predicted_probabilities_mlp = mlp_model.predict_proba(sample_data_point)

print("Predicted probabilities for the classes:")
print(predicted_probabilities_mlp)

# iii. Report on the score for the Neural Network, what does the score measure?
# Predict the test set results
y_pred_mlp = mlp_model.predict(X_test)

# Calculate the accuracy score
accuracy_mlp = accuracy_score(y_test, y_pred_mlp)

print(f'MLP Classifier Model Accuracy: {accuracy_mlp}')

# iv: Experiment with different options for the neural network, report on your best configuration
# Trying various configurations
configurations = [
    {'hidden_layer_sizes': (5,), 'activation': 'relu'},
    {'hidden_layer_sizes': (10,), 'activation': 'tanh'},
    {'hidden_layer_sizes': (10, 5), 'activation': 'logistic'},
    {'hidden_layer_sizes': (10, 10), 'activation': 'relu'},
    {'hidden_layer_sizes': (5, 5), 'activation': 'tanh'},
]

best_accuracy = 0
best_config = None

for config in configurations:
    mlp_model = MLPClassifier(**config, max_iter=500, random_state=42)  # Increase iterations for convergence
    mlp_model.fit(X_train, y_train)
    accuracy = mlp_model.score(X_test, y_test)
    
    print(f"Configuration: {config}, Accuracy: {accuracy}")

    # Update best configuration
    if accuracy > best_accuracy:
        best_accuracy = accuracy
        best_config = config

print(f"Best Configuration: {best_config}, Best Accuracy: {best_accuracy}")
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
from sklearn.neighbors import KNeighborsClassifier

# Initialize the k-Neighbors Classifier with a specified number of neighbors
k = 3  # You can choose any number of neighbors (k)
knn_model = KNeighborsClassifier(n_neighbors=k)

# Fit the k-NN model using the training data (load the data into the model)
knn_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
# Select a sample data point from the test set (e.g., first observation)
sample_data_point = X_test.iloc[0].values.reshape(1, -1)

# Predict probabilities for each possible class
predicted_probabilities_knn = knn_model.predict_proba(sample_data_point)

print("Predicted probabilities for the sample data point:")
print(predicted_probabilities_knn)

# iii. Report on the score for kNN, what does the score measure?
from sklearn.metrics import accuracy_score

# Predict the results on the test set
y_pred_knn = knn_model.predict(X_test)

# Calculate the accuracy score
accuracy_knn = accuracy_score(y_test, y_pred_knn)

print(f'k-Neighbors Classifier Model Accuracy: {accuracy_knn}')
```
