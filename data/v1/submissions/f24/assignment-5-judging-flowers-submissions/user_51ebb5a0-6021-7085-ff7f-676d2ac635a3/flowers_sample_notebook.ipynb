# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)

# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
df = pd.read_csv("iris.csv")
display(df.head(15))
df.info()
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The data is describing the picture based on the sepal's and petal's width and length. The labels are the different species of flowers. We will use the lengths of each part to try and predict what the species is

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
x = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.35, random_state=2)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model = LogisticRegression(max_iter=150).fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample = pd.DataFrame({'sepal_length': [5.5], 'sepal_width': [2.4], 'petal_length': [3.7], 'petal_width': [1]})
species_prediction = model.predict_proba(sample)
print(f"Sample prediction: {species_prediction}")

# iii. Report on the score for Logistic regression model, what does the score measure?
y_prediction = model.predict(X_test)
score = accuracy_score(y_test, y_prediction)

print(f"Accuracy score of the model: {score}")

# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = model.coef_
intercept = model.intercept_

print(f"Coefficients: {coefficients}")
print(f"Intercepts: {intercept}")
```

The score of .96 represents a 96% accuracy score for the predictions made. In other words, the model predicts the correct species of flower 96% of the time. 

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc_model = SVC(probability=True).fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
class_prob = svc_model.predict_proba(sample)
print(f"Class probabilities: {class_prob}")

# iii. Report on the score for the SVM, what does the score measure?
svc_y_prediction = svc_model.predict(X_test)
svc_score = accuracy_score(y_test, svc_y_prediction)
print(f"SVC Score: {svc_score}")
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
nn_model = MLPClassifier(hidden_layer_sizes=(10,10), random_state=1,max_iter=2000).fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
nn_probabilities = nn_model.predict_proba(sample)
print(f"Neural Network Sample Probabilities: {nn_probabilities}")

# iii. Report on the score for the Neural Network, what does the score measure?
nn_y_prediction = nn_model.predict(X_test)
nn_score = accuracy_score(y_test, nn_y_prediction)
print(f"Neural Network Score: {nn_score}")

# iv: Experiment with different options for the neural network, report on your best configuration
best_config = None
config_score = 0

config = [{'hidden_layer_size':(10, 10, 10), 'activation': 'logistic', 'solver': 'lbfgs'},
          {'hidden_layer_size':(5, 5), 'activation': 'tanh', 'solver': 'sgd'},
          {'hidden_layer_size':(20), 'activation': 'relu', 'solver': 'adam'},
          {'hidden_layer_size':(10, 5, 10), 'activation': 'logistic', 'solver': 'adam'},
          {'hidden_layer_size':(5, 5, 5), 'activation': 'tanh', 'solver': 'lbfgs'},
          {'hidden_layer_size':(3, 5, 7), 'activation': 'relu', 'solver': 'sgd'}]

for conf in config:
    nn_different_config = MLPClassifier(
        hidden_layer_sizes = conf['hidden_layer_size'],
        activation = conf['activation'],
        solver = conf['solver'],
        random_state = 1,
        max_iter = 3000
    ).fit(X_train, y_train)

    different_y_prediction = nn_different_config.predict(X_test)
    conf_accuracy = accuracy_score(y_test, different_y_prediction)
    if conf_accuracy > config_score:
        best_config = conf
        config_score = conf_accuracy

print(f"Best Configuration: {best_config}")
print(f"Best Configuration Score: {config_score}")
```

The best configuration for the neural network (not including the original configuration we started with) had a hidden_layer_size of (5, 5), an activation of 'tanh', and a solver of 'sgd'. This produced a score of 1.0. This doesn't take into consideration that some of the different configurations may also have a score of 1.0. It just takes the first instance of a 1.0 and uses it as the best model. I will say that it is very odd that both neural networks provide a score of 1.0. This may be due to the small sample size provided

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
kN_model = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
kN_probabilities = kN_model.predict_proba(sample)
print(f"K-Nearest Neighbors Sample Predictions: {kN_probabilities}")

# iii. Report on the score for kNN, what does the score measure?
kN_y_prediction = kN_model.predict(X_test)
kN_score = accuracy_score(y_test, kN_y_prediction)
print(f"K-Nearest Neighbors Score: {kN_score}")
```
