```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
df = pd.read_csv('iris.csv')
# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(df.info())
```

```python
print(df['species'].unique())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?
* The dataset contains measurements of iris flower and the species.
* The features present in the dataset are `sepal_length`, `sepal_width`, `petal_length` and `petal_width`.
* The label for this dataset is the `species` column. There are 3 different species, `setosa`, `versicolor`, and `virginica`.
* The dataset classifies an iris flower based on its measurements.
* Mapping of Labels:
    * `setosa` ‚Üí Class Setosa
    * `versicolor` ‚Üí Class Versicolor
    * `virginica` ‚Üí Class Virginica

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = df['species']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=41)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model = LogisticRegression()
model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame([X_test.iloc[0].values], columns=X_test.columns)
probabilities = model.predict_proba(sample_datapoint)
print(f'Predicted probabilities for each class: {probabilities}')
# iii. Report on the score for Logistic regression model, what does the score measure?
score = model.score(X_test, y_test)
print(f'Accuracy score: {score}')
# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = model.coef_
intercepts = model.intercept_
print(f'Coefficients: {coefficients}')
print(f'Intercepts: {intercepts}')
```

* The accuracy score that was outputed measures the proportion of correct predictions the model makes on the test set, indicating its overall performance in classifying the data.
* In our case we can see that the accuracy score in 0.93 which means the model makes correct predictions 93% of time. This shows us that the model fits the data well and is performing well.

* Boundary Lines (3 present) (Rounded to 4 decimal places):
    1. `Setosa` or `Class 1`: ‚àí0.4375‚ãÖx1 + 0.8737‚ãÖx2 ‚àí 2.4208‚ãÖx3 ‚àí 1.0501‚ãÖx4 + 9.8374
    2. `Versicolor` or `Class 2`: 0.5694‚ãÖx1‚Äã ‚àí 0.4854‚ãÖx2 ‚àí 0.2000‚ãÖx3 ‚àí 0.8777‚ãÖx4 + 2.3443
    3.  `Virginica` or `Class 3`: ‚àí0.1319‚ãÖx1 ‚àí 0.3883‚ãÖx2 + 2.6208‚ãÖx3 + 1.9278‚ãÖx4 ‚àí 12.1817

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc_model = SVC(probability=True)
svc_model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame([X_test.iloc[0].values], columns=X_test.columns)
svc_probabilities = svc_model.predict_proba(sample_datapoint)
print(f'Predicted probabilities for each class (SVC): {svc_probabilities}')
# iii. Report on the score for the SVM, what does the score measure?
svc_score = svc_model.score(X_test, y_test)
print(f'SVC Accuracy score: {svc_score}')
```

* The score is 0.93 which means that the SVM model correctly classified 93% of the test samples.
* This shows us the model fit the data well and was able to generalize well.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(random_state=42, max_iter=600)
mlp_model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame([X_test.iloc[0].values], columns=X_test.columns)
mlp_probabilities = mlp_model.predict_proba(sample_datapoint)
print(f'Predicted probabilities for each class (MLP): {mlp_probabilities}')
# iii. Report on the score for the Neural Network, what does the score measure?
mlp_score = mlp_model.score(X_test, y_test)
print(f'MLP Accuracy score: {mlp_score}')
# iv: Experiment with different options for the neural network, report on your best configuration
configurations = [
    {'hidden_layer_sizes': (10,), 'activation': 'relu', 'solver': 'adam', 'learning_rate': 'constant', 'learning_rate_init': 0.01},
    {'hidden_layer_sizes': (50, 30), 'activation': 'tanh', 'solver': 'adam', 'learning_rate': 'adaptive', 'learning_rate_init': 0.001},
    {'hidden_layer_sizes': (100, 50, 25), 'activation': 'relu', 'solver': 'sgd', 'learning_rate': 'adaptive', 'learning_rate_init': 0.01},
    {'hidden_layer_sizes': (100,), 'activation': 'logistic', 'solver': 'adam', 'learning_rate': 'constant', 'learning_rate_init': 0.1}
]


random_state = 7

#Using different random state for the train/test split here. The only way to produce fair results.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=random_state)
print(f"\nTesting with train/test split random_state={random_state}")

# Initialize variables for tracking the best configuration
best_score = 0
best_config = None

# Loop through configurations
for config in configurations:
    # Create and train the MLP model
    temp_model = MLPClassifier(random_state=random_state, max_iter=2000, **config)
    temp_model.fit(X_train, y_train)
    score = temp_model.score(X_test, y_test)
    print(f"Configuration: {config}, Accuracy: {score}")

    if score > best_score:
        best_score = score
        best_config = config

print(f'Best Configuration with random_state {random_state}: {best_config} with Accuracy: {best_score}')
```

* There were 2 configurations that performed the best:
    1. `hidden_layer_sizes': (50, 30), 'activation': 'tanh', 'solver': 'adam', 'learning_rate': 'adaptive', 'learning_rate_init': 0.001`
    2. `hidden_layer_sizes': (100, 50, 25), 'activation': 'relu', 'solver': 'sgd', 'learning_rate': 'adaptive', 'learning_rate_init': 0.01`.
* Both of these configurations produced an accuracy score of 0.967, which shows how well the model is perforing and able to generalize.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame([X_test.iloc[0].values], columns=X_test.columns)
knn_probabilities = knn_model.predict_proba(sample_datapoint)
print(f'Predicted probabilities for each class (k-NN): {knn_probabilities}')
# iii. Report on the score for kNN, what does the score measure?
knn_score = knn_model.score(X_test, y_test)
print(f'k-NN Accuracy score: {knn_score}')
```

* The k-NN accuracy score of `0.93` indicates that the model correctly classified 93% of the test dataset, which shows the model reflects the data well and is able to generalize and perform well.
