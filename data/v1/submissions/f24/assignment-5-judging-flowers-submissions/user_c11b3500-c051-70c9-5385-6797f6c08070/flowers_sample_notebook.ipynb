```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression 
from sklearn.metrics import accuracy_score, confusion_matrix
import numpy as np
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
dataset = pd.read_csv("iris.csv")
# Output the first 15 rows of the data
dataset.head(15)
```

```python
# Display a summary of the table information (number of datapoints, etc.)
dataset.describe(include='all')
```

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
x = dataset.drop('species', axis=1)
y = dataset['species']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=5)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model = LogisticRegression()
model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
probabilities = model.predict_proba([[5.1, 3.5, 1.4, 0.2]])
probabilities
```

```python
# iii. Report on the score for Logistic regression model, what does the score measure?
score = model.score(X_test, y_test)
score
```

The score refers how well the model predicts the correct labeling on iterations of the testing data. A score of 1 indicates that the model predicted all instances of the data correct. If the score is 0, 0% of the predicts of label values given a feature were correct. This score only indicates the accuracy of the model on this dataset, the test dataset we have given it.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
SVCmodel = SVC(kernel = 'rbf', random_state=42, probability = True).fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
SVCmodel.predict_proba([[5.1, 3.5, 1.4, 0.2]])
```

```python
# iii. Report on the score for the SVM, what does the score measure?
score = SVCmodel.score(X_test, y_test)
score
```

The score tells you how often the SVC model accurately predicts the label for the test dataset. This model correctly predicted the test dataset 93% of the time.  

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
NNmodel = MLPClassifier(hidden_layer_sizes=(10, 10), max_iter=1000, random_state=42)
NNmodel.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
NNmodel.predict_proba([[5.1, 3.5, 1.4, 0.2]])
```

```python
# iii. Report on the score for the Neural Network, what does the score measure?
score = NNmodel.score(X_test, y_test)
score
```

The score tells you how often the neural network model accurately predicts the label for the test dataset. This model correctly predicted the test dataset 80% of the time.
