```python

# Load the dataset (load remotely or locally)
iris = fetch_ucirepo(id=53)

# Output the first 15 rows of the data
dataframe_original = iris.data.original
print(iris.data.original.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(dataframe_original.describe())
```

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = iris.data.features
y = iris.data.targets


# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
logistic_model = LogisticRegression(max_iter=1000)
logistic_model = logistic_model.fit(X_train, np.ravel(y_train))
# ii. For a sample datapoint, predict the probabilities for each possible class
sample = X_test.sample(n=1)
print("sample from test dataset: \n", sample)
print("probabilities for each class: ", logistic_model.predict_proba(sample))
print("predicted class: ", logistic_model.predict(sample))
print("actual class: ", y_test.loc[sample.index[0]])

# iii. Report on the score for Logistic regression model, what does the score measure?
# print("score: ", logistic_model.score(X_test, y_test))
cross_val_scores = cross_val_score(logistic_model, X, np.ravel(y), cv=5)
print("cross val scores: " + str(cross_val_scores))
score_mean = statistics.mean(cross_val_scores)
print("average cross val score: " + str(score_mean))
print("A score of " + str(score_mean) + " indicates that there is a strong correlation between the features and target.")
# iv. Extract the coefficents and intercepts for the boundary line(s)
print("coefficients: \n",logistic_model.coef_)
print("intercept: ", logistic_model.intercept_)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc_model = svm.SVC(probability=True);
svc_model = svc_model.fit(X_train, np.ravel(y_train))
# ii. For a sample datapoint, predict the probabilities for each possible class
sample2 = X_test.sample(n=1)
print("sample from test dataset: \n", sample)
print("probabilities for each class: ", svc_model.predict_proba(sample))
print("predicted class: ", svc_model.predict(sample))
print("actual class: ", y_test.loc[sample.index[0]])
# iii. Report on the score for the SVM, what does the score measure?
cross_vals_svc = cross_val_score(svc_model, X, np.ravel(y))
print("cross val scores: ", cross_vals_svc)
svc_mean = statistics.mean(cross_vals_svc)
print("average cross val score: ", svc_mean)
print("A score of " + str(svc_mean) + " indicates that there is a strong correlation between the features and target, though slightly worse than the logistic regression model.")
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_classifier = MLPClassifier(random_state=1, max_iter=700).fit(X_train, np.ravel(y_train))
# ii. For a sample datapoint, predict the probabilities for each possible class
sample3 = X_test.sample(n=1)
print("sample from test dataset: \n", sample)
print("probabilities for each class: ", mlp_classifier.predict_proba(sample3))
print("predicted class: ", mlp_classifier.predict(sample3))
print("actual class: ", y_test['class'].loc[sample3.index])

# iii. Report on the score for the Neural Network, what does the score measure?
cross_vals_nn = cross_val_score(mlp_classifier, X, np.ravel(y))
print("cross val scores: ", cross_vals_nn)
nn_mean = statistics.mean(cross_vals_nn)
print("average cross val score: ", nn_mean)
print("A score of " + str(nn_mean) + " represents a strong correlation between target and feature values. The neural network model has slighlty more accurate predictions than the SVM or logistic models.")

# iv: Experiment with different options for the neural network, report on your best configuration
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model = knn_model.fit(X_train, np.ravel(y_train))
# ii. For a sample datapoint, predict the probabilities for each possible class
sample4 = X_test.sample(n=1)
print("probabilities for each class: ", knn_model.predict_proba(sample4))
print("predicted class: " + knn_model.predict(sample4))
print("actual class: ", y_test["class"].loc[sample4.index])
# iii. Report on the score for kNN, what does the score measure?
cross_vals_knn = cross_val_score(knn_model, X, np.ravel(y))
print("cross val scores: ", cross_vals_knn)
knn_mean = statistics.mean(cross_vals_knn)
print("average cross val score: ", knn_mean)
print("An average score of " + str(knn_mean) + " represents that the k-nearest neighbors model has a good correlation between features and targets, but it is weaker than the logistic and neural network models.")
```
