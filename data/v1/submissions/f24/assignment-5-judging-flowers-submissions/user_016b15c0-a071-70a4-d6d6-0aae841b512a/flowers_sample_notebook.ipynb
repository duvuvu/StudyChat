```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
iris = load_iris(as_frame=True)
# Output the first 15 rows of the data
print(iris.data.head(15))
# print(iris.target.head(15))
print(iris.target_names)
# Display a summary of the table information (number of datapoints, etc.)
print(iris.data.describe())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

This contains data about iris flowers, including various lengths and widths (features) and which type of iris it is (labels). The features are the sepal length, sepal width, petal length, and petal width (all in units of cm). The label for each datapoint is an integer 0, 1, or 2. Each number represents a specfic class of flower. 0 represents the Iris-Setosa, 1 represents the Iris-Versicolour, and 2 represents the Iris-Virginica.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = iris.data
y = iris.target
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, test_size=0.1, random_state=29)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
logReg = LogisticRegression().fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = X_test.head(1)
print(f'Prediction probabilities: {logReg.predict_proba(sample_datapoint)}')
# iii. Report on the score for Logistic regression model, what does the score measure?
print(f'Score: {logReg.score(X_test, y_test)}')
# iv. Extract the coefficents and intercepts for the boundary line(s)
print(f'Coefficients: {logReg.coef_}')
print(f'Intercepts: {logReg.intercept_}')
```

### Part 3 Reports:
The score is a measure of the mean accuracy of based on the inputted test features and test labels. Essentially, it is a measure of how accurate the predicted class is to the actual class of the datapoint. 1 represents a perfect model where as 0 is the worst model predicition. Since the score is somewhat close to 1, that means are model is somewhat accurate at predicting the class given a feature vector, but not perfect.

Here are the equations of the three boundary lines (rounded to 4 significant digits):
Let $x_1$ represent the sepal length, $x_2$ represent the sepal width, $x_3$ represent the petal length, and $x_4$ represent the pedal length, all in units of cm.
$$f_1(x_1, x_2, x_3, x_4) = 10.07 - 0.4321x_1 + 0.8512x_2 - 2.463x_3 - 1.005x_4$$
$$f_2(x_1, x_2, x_3, x_4) = 1.694 + 0.5442x_1 - 0.2406x_2 - 0.1741x_3 - 0.9333x_4$$
$$f_3(x_1, x_2, x_3, x_4) = 11.76 - 0.1121x_1 - 0.6106x_2 + 2.637x_3 + 1.938x_4$$

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc_model = SVC(probability=True).fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
print(f'Prediction probabilities: {svc_model.predict_proba(sample_datapoint)}')
# iii. Report on the score for the SVM, what does the score measure?
print(f'Score: {svc_model.score(X_test, y_test)}')
```

### Part 4 Reports:
Like for logisitic regression, the score measures the mean accuracy of based on the X_test feature vectors and y_test labels. That is, it measures how accurate the predicted class is to the actual class. A score of 1 is the most accurate, where as 0 is the least accurate. Since the score is close to 1, the model is mostly accurate but has some room for improvement.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
nn_model = MLPClassifier().fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
print(f'Prediction probabilities: {nn_model.predict_proba(sample_datapoint)}')
# iii. Report on the score for the Neural Network, what does the score measure?
print(f'Score: {nn_model.score(X_test, y_test)}')
# iv: Experiment with different options for the neural network, report on your best configuration
moreIter = MLPClassifier(max_iter=2000, activation='logistic').fit(X_train, y_train)
print(f'Prediction probabilities: {moreIter.predict_proba(sample_datapoint)}')
print(f'Score: {moreIter.score(X_test, y_test)}')
```

### Part 5 Reports:
Just like the last two models, the score is a measure of the mean accuracy of the prediction of a feature vector from X_test to its corresponding label in y_test. A score of 1 represents perfect predictions where as 0 represents predicition that are completely incorrect. Both neural networks used had high scores as they both were close to 1.

I configured the neural network to use a logisitic (sigmoid) function for the activation function instead of the default rectified linear unit function $f(x) = \max(0,x)$. I also increased the max number of iterations to 2000, which consistently allows the neural network optimization to converge before reaching the iteration cap. This sometimes resulted in a better score, depending on the random values the neural network was initialized with. However, it does not perform worse than the first neural network, only the same or better.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
kNN_model = KNeighborsClassifier().fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
print(f'Prediction probabilities: {kNN_model.predict_proba(sample_datapoint)}')
# iii. Report on the score for kNN, what does the score measure?
print(f'Score: {kNN_model.score(X_test, y_test)}')
```

### Part 6 Reports:
As with the other model scores, this score measures the mean accuracy of the prediction of the given test feature vector to its corresponding label in the train set. A score of 1 is the best whereas a score of 0 is the worst. The score is somewhat close to 1, but could be better. So this model is okay at predicting the class given a feature vector, but is a little imperfect.
