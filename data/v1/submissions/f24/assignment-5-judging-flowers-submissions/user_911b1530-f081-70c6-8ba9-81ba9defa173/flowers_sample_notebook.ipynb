```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
data = pd.read_csv("iris.csv")

# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
print(data.head(15))
print(f"\nNo. of rows: {data.shape[0]}")
print(f"\nNo. of columns: {data.shape[0]}\n")
print(f"Classes: {data['species'].unique()}\n")
data.info()
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The dataset consists of 150 samples of iris flowers, with each sample described by five attributes: sepal_length, sepal_width, petal_length, and petal_width. These four attributes serve as the features used to predict the label, which is the species of the iris flower.

The target label species is divided into three distinct classes: 
- Iris Setosa
- Iris Versicolor
- Iris Virginica

By analyzing the relationships between the features and the label, we aim to classify each iris flower into one of these species based on its physical characteristics.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = data.drop('species', axis=1)
y = data['species']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.10, random_state=45)

print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model_LR = LogisticRegression(max_iter=200).fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
prediction = model_LR.predict(X_test.iloc[[1]]) 
print(f"Prediction: {prediction}\n")

probability = model_LR.predict_proba(X_test.iloc[[1]])
print(f"Probability for sentosa: {probability[0][0]}")
print(f"Probability for versicolor: {probability[0][1]}")
print(f"Probability for virginica: {probability[0][2]}\n")

# iii. Report on the score for Logistic regression model, what does the score measure?
score = model_LR.score(X_test, y_test) 
print(f"Score: {score}\n")

# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficent = model_LR.coef_ 
print(f"Coefficent: {coefficent}\n")

intercept = model_LR.intercept_ 
print(f"Intercept: {intercept}")
```

The score for the Logistic Regression Model is 0.9333, representing the accuracy of the model on the given dataset. Accuracy is calculated as the ratio of correctly predicted samples to the total samples in the test set. This means the model correctly classifies 93.33% of the instances, providing a quick measure of its performance in assigning data to the correct categories.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
model_svm = svm.SVC(probability=True).fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
prediction = model_svm.predict(X_test.iloc[[1]]) 
print(f"Prediction: {prediction}\n")

probability = model_svm.predict_proba(X_test.iloc[[1]])
print(f"Probability for sentosa: {probability[0][0]}")
print(f"Probability for versicolor: {probability[0][1]}")
print(f"Probability for virginica: {probability[0][2]}\n")

# iii. Report on the score for the SVM, what does the score measure?
score = model_svm.score(X_test, y_test) 
print(f"Score: {score}\n")
```

The score for the SVM model indicates an accuracy of 93.33%, , meaning that the model correctly classified approximately 93.33% of the instances in the test dataset. This high score suggests that the model is performing very well in distinguishing between the classes. 

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
model_nn = MLPClassifier(random_state=1, max_iter=200).fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
prediction = model_nn.predict(X_test.iloc[[1]]) 
print(f"Prediction1: {prediction}\n")

# iii. Report on the score for the Neural Network, what does the score measure?
score = model_nn.score(X_test, y_test) 
print(f"Score1: {score}\n")

# iv: Experiment with different options for the neural network, report on your best configuration
model_nn1 = MLPClassifier(random_state=1, max_iter=10).fit(X_train, y_train)

prediction = model_nn1.predict(X_test.iloc[[1]]) 
print(f"Prediction2: {prediction}\n")

score = model_nn1.score(X_test, y_test) 
print(f"Score2 : {score}\n")


model_nn2 = MLPClassifier(random_state=1, max_iter=100).fit(X_train, y_train)

prediction = model_nn2.predict(X_test.iloc[[1]]) 
print(f"Prediction3: {prediction}\n")

score = model_nn2.score(X_test, y_test) 
print(f"Score3: {score}\n")
```

Prediction 1 represents the best configuration, as Prediction 2 has a low score that may result in underfitting. In contrast, Prediction 3 achieves a perfect score, which poses a risk of overfitting during prediction.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
model_knn = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
prediction = model_knn.predict(X_test.iloc[[1]]) 
print(f"Prediction: {prediction}\n")

# iii. Report on the score for kNN, what does the score measure?
score = model_knn.score(X_test, y_test) 
print(f"Score: {score}\n")
```

A K-Nearest Neighbors (KNN) score of 0.9333 indicates that the model has an accuracy of 93.33%, meaning that it correctly classified approximately 93.33% of the instances in the test dataset. This high score suggests that the KNN model is effectively distinguishing between the classes in the dataset, making reliable predictions for most of the instances.
