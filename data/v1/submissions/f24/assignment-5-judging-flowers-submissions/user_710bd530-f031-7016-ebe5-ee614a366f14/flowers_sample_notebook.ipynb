```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
dataset = pd.read_csv("iris.csv")

# Output the first 15 rows of the data
print("First 15 columns of iris.csv: ")
print(dataset.head(15))
print()

# Display a summary of the table information (number of datapoints, etc.)
print("Table summary: ")
print(dataset.info())
print("Number of datapoints:", len(dataset))
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The dataset is about classifying flowers into 3 classes: setosa, versicolor, and virginica. The dataset contains 150 datapoints, and it contains 4 features (sepal_length, sepal_width, petal_length, and petal_width) and 1 label column (species). Each class has 50 datapoints associated with it.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
x = dataset.drop("species", axis=1)
y = dataset["species"]
# print(x.head(3))
# print(y.head(3))

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model = LogisticRegression(max_iter=200)
model.fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = x_test.iloc[[0]]
# print(sample_data)
prob = model.predict_proba(sample_data)
print("Probabilities for each class: ")
print(prob)
# print(sum(prob[0]))

# iii. Report on the score for Logistic regression model, what does the score measure?
# What does the score measure? ---> in Markdown cell below
score = model.score(x_test, y_test)
print("Score of the Logistic Regression model: ", score)

# iv. Extract the coefficents and intercepts for the boundary line(s)
coeffs = model.coef_
intercept = model.intercept_
print("Model coefficients: ", coeffs)
print("Model intercept: ", intercept)
```

iii. What does the score measure?

The score measures the accuracy of the logistic regression model. Accuracy = (Number of correct predictions) / (Total number of predictions). In this case, the score is 1.0, which indicates that the model perfectly classifies all the datapoints in the test set. However, this could also indicate that the model is overfitted to the data.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc_model = SVC(probability=True)
svc_model.fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = x_test.iloc[[0]]
prob = svc_model.predict_proba(sample_data)
print("Probabilities for each class: ")
print(prob)

# iii. Report on the score for the SVM, what does the score measure?
score = svc_model.score(x_test, y_test)
print("Score of Support Vector Classifier model: ", score)
```

iii. What does the score measure?

The score measures the SVM model's accuracy. Accuracy = (number of correct predictions) / (total number of predictions). For this model, the score is 1.0, which means that the accuracy is 100% and the model correctly identifies each datapoint and puts it in the right class. However, this could also indicate overfitting.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(max_iter=750)
mlp_model.fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = x_test.iloc[[0]]
prob = mlp_model.predict_proba(sample_data)
print("Probabilities for each class: ")
print(prob)

# iii. Report on the score for the Neural Network, what does the score measure?
# Description --> markdown cell below
score = mlp_model.score(x_test, y_test)
print("Score of the MLP model: ", score)
print()

# iv: Experiment with different options for the neural network, report on your best configuration

# The score for my neural network is 1.0, which means that it correctly identifies all the datapoints in the test set.
# Thus, the current configuration is the best one.
print("Current configuration of mlp_model: ") 
print("Hidden layer sizes: ", mlp_model.hidden_layer_sizes)
print("Activation function: ", mlp_model.activation)
print("Solver: ", mlp_model.solver)
print("Number of iterations: ", mlp_model.n_iter_)
```

iii. What does the score measure?

The score measures the accuracy of the neural network. Accuracy = (number of correct predictions) / (total number of predictions). The score is 1.0, which implies that the neural network correctly categorizes all the datapoints in the test set. However, this score could also indicate overfitting.


# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn_model = KNeighborsClassifier(n_neighbors=3)
knn_model.fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = x_test.iloc[[0]]
prob = knn_model.predict_proba(sample_data)
print("Probabilities for each class: ")
print(prob)

# iii. Report on the score for kNN, what does the score measure?
# Description --> Markdown cell below
score = knn_model.score(x_test, y_test)
print("Score of the k-Neighbors classifier model: ", score)
```

iii. What does the score measure?

The score measures the accuracy of the k-neighbors classifier model. Accuracy = (number of correct predictions) / (size of the test set). When test_size = 0.1 and random_state = 42, the score is 1.0, which means that the KNN model correctly classifies every datapoint in the test set and predicts the correct label. However, like in the other models, this could also indicate overfitting.
