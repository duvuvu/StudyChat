```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

### Load the dataset (load remotely or locally)

```python
iris_data = pd.read_csv("iris.csv")
```

### Output the first 15 rows of the data

```python
print(f"Iris data (15 rows):\n{iris_data[:15].to_string()}\n")
```

### Display a summary of the table information (number of datapoints, etc.)

```python
print(f"Iris data summary:\n{iris_data.describe()}\n")
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The data is of sepal and petal lengths and widths of samples of *Iris setosa*, *Iris virginica*, and *Iris versicolor* flowers. There are fifty samples of each of the three species. This (in)famous dataset (as our professor desribes it) was collected by botanist Edgar Anderson, who obtained the samples from "the same pasture, and picked on the same day and measured at the same time by the same person with the same apparatus". The three flowers (all of the genus *Iris*) are notable for their very similar appearance.  
  
The four features are the aforementioned qualities of sepal length, sepal width, petal length, and petal width. The labels are the species of the genus *Iris*, which are either *setosa*, *virginica*, or *versicolor* in this dataset.

# Part 2: Split the dataset into train and test

### Take the dataset and split it into our features (X) and label (y)

```python
iris_X = iris_data.drop("species", axis=1)
iris_y = iris_data["species"]
```

### Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)

```python
X_train, X_test, y_train, y_test = train_test_split(iris_X, iris_y, test_size=0.1, random_state=42)
```

# Part 3: Logistic Regression

### i. Use sklearn to train a LogisticRegression model on the training set

```python
lr_model = LogisticRegression(max_iter=500)

lr_model.fit(X_train, y_train)
```

### ii. For a sample datapoint, predict the probabilities for each possible class

```python
lr_sample_input = X_test.iloc[[1]]
lr_sample_actual_class = y_test.iloc[1]
lr_sample_predicted_class = lr_model.predict(lr_sample_input)[0]

print(f"Input sample point: \n{lr_sample_input}\n")
print(f"Predicted class of sample point: \n{lr_sample_predicted_class}\n")
print(f"Actual class of sample point: \n{lr_sample_actual_class}")
```

### iii. Report on the score for Logistic regression model, what does the score measure?

```python
lr_score = lr_model.score(X_test, y_test)

print(f"Score for LogisticRegression model: {lr_score}")

lr_kf = KFold(n_splits=10, random_state=42, shuffle=True)
lr_cv_scores = cross_val_score(lr_model, X_train, y_train, cv=lr_kf)

print(f"Scores from cross-validation:\n{lr_cv_scores}")
print(f"Average score from cross-validation: {lr_cv_scores.mean()}")
```

This score measures how accurately the logistic regression model classified the test dataset after being trained on the training set. It does this by reporting what percent of classes had all of its classes identified correctly. Since this dataset only has one class for each data point (it has only one label), this is just the percent of the data it predicted the class of correctly.  
  
It scored a 1.0, which means it predicted the species correctly 100% of the time. This is to be expected for this dataset, which has classes that are quite distinct for the label. Of course, 1.0 prompts some worries of overfitting, so I ran some cross-validation. The mean score from the 10 fold cross-validation was 96.98%, which is very high and alleviates some concern for overfitting while still showing that the model is strong.

### iv. Extract the coefficents and intercepts for the boundary line(s)

```python
lr_classes = lr_model.classes_
lr_coefs = lr_model.coef_
lr_intercept = lr_model.intercept_

print(f"Boundary Lines:")
for i in range(len(lr_classes)):
    c = lr_coefs[i]
    print(f"{lr_classes[i]}: h(x) = {c[0]}x_1 + {c[1]}x_2 + {c[2]}x_3 + {c[3]}x_4 + {lr_intercept[i]}")
```

# Part 4: Support Vector Machine

### i. Use sklearn to train a Support Vector Classifier on the training set

```python
svc_model = svm.SVC(probability=True)

svc_model.fit(X_train, y_train)
```

### ii. For a sample datapoint, predict the probabilities for each possible class

```python
svc_classes = svc_model.classes_

svc_sample_input = X_test.iloc[[1]]
svc_sample_actual_class = y_test.iloc[1]
svc_sample_class_probs = svc_model.predict_proba(svc_sample_input)[0]

print(f"Input sample point: \n{svc_sample_input}\n")
print(f"Probabilities of sample point being in each class:")
for i in range(len(svc_sample_class_probs)):
    print(f"{svc_classes[i]}: {svc_sample_class_probs[i] * 100}%")
print(f"\nActual class of sample point: \n{svc_sample_actual_class}")
```

### iii. Report on the score for the SVM, what does the score measure?

```python
svc_score = svc_model.score(X_test, y_test)

print(f"Score for Support Vector Classifier model: {svc_score}")

svc_kf = KFold(n_splits=10, random_state=42, shuffle=True)
svc_scores = cross_val_score(svc_model, X_train, y_train, cv=svc_kf)

print(f"Scores from cross-validation:\n{svc_scores}")
print(f"Average score from cross-validation: {svc_scores.mean()}")
```

This score measures how accurately the SVC model classified the test dataset after being trained on the training set. It does this by reporting what percent of classes had all of its classes identified correctly. Since this dataset only has one class for each data point (it has only one label), this is just the percent of the data it predicted the class of correctly.  
  
The model scored a 1.0, which means it predicted the species correctly 100% of the time. As with the Logistic Regression model, this is to be expected for this dataset, which has classes that are quite distinct for the label. Again, 1.0 prompts some worries of overfitting, so I ran some cross-validation. The mean score from the 10 fold cross-validation was 96.98%, which is very high (the same as the Logistic Regression model) and alleviates some concern for overfitting while still showing that the model is strong.
