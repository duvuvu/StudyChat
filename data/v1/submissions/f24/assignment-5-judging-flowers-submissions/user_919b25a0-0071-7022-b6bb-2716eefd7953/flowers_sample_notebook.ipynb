```python
#Name: <redacted> <redacted>, Git username: <redacted>, commit ID: 

# Imports and pip installations (if needed)
import pandas as pd
import io 
import requests
print(pd.__version__)
import sklearn
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import PolynomialFeatures

import numpy as np
from sklearn.metrics import make_scorer, r2_score
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
df =  pd.read_csv('iris.csv')


# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
df.info()
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The features are numerical and describe the flowers: sepal length, sepal width, petal length, and petal width. Labels correspond to the species of the flowers, where label 0 is iris setosa, versicolor is label 1, and virginica is label 2. 

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = df[['sepal_length', 'sepal_width', 'petal_length','petal_width']]  # Features
y = df['species']   # Label

print("Features shape:", X.shape)
print("Label shape:", y.shape)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X = X.dropna()
y = y.dropna()

X = X.loc[y.index] 

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

print("Training set:", X_train.shape, y_train.shape)
print("Test set:", X_test.shape, y_test.shape)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = pd.DataFrame({"sepal_length":[3], "sepal_width": [3], "petal_length": [1], "petal_width":[4]})
probabilities = log_reg.predict_proba(sample_data)
# iii. Report on the score for Logistic regression model, what does the score measure?
# Calculate the score (accuracy) of the model on the test set
score = log_reg.score(X_test, y_test)

# Output the score
print("Model accuracy score:", score)
# iv. Extract the coefficents and intercepts for the boundary line(s)
# Extract the coefficients and intercept
coefficients = log_reg.coef_
intercept = log_reg.intercept_

# Output the coefficients and intercept
print("Coefficients:\n", coefficients)
print("Intercept:\n", intercept)

#THe score is very accurate which means that it was able to accurately predict the correct classification.
```

# Part 4: Support Vector Machine

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# Create and train a Support Vector Classifier model
svc = SVC(probability=True)  # Set probability=True to allow for probability estimates
svc.fit(X_train, y_train)

# Predict probabilities for a sample data point (the first point in the test set)
sample_data = pd.DataFrame({"sepal_length":[3], "sepal_width": [3], "petal_length": [1], "petal_width":[4]})
probabilities = svc.predict_proba(sample_data)

# Output the probabilities
print("Probabilities for each class:", probabilities)

# Calculate and report the score (accuracy) of the SVC on the test set
score = svc.score(X_test, y_test)
print("Model accuracy score:", score)
```

# Part 5: Neural Network

```python
from sklearn.neural_network import MLPClassifier

# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp = MLPClassifier(random_state=42)
mlp.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = pd.DataFrame({"sepal_length":[3], "sepal_width": [3], "petal_length": [1], "petal_width":[4]})
probabilities = mlp.predict_proba(sample_data)
print("Probabilities for each class:", probabilities)

# iii. Report on the score for the Neural Network, what does the score measure?
score = mlp.score(X_test, y_test)
print("Model accuracy score:", score)

# iv. Experiment with different options for the neural network
best_score = 0
best_model = None

configurations = [
    {'hidden_layer_sizes': (10,), 'activation': 'relu', 'solver': 'adam'},
    {'hidden_layer_sizes': (10, 10), 'activation': 'relu', 'solver': 'adam'},
    {'hidden_layer_sizes': (20,), 'activation': 'tanh', 'solver': 'sgd'},
    {'hidden_layer_sizes': (15, 15), 'activation': 'logistic', 'solver': 'adam'},
    {'hidden_layer_sizes': (30,), 'activation': 'relu', 'solver': 'adam'},
]

for config in configurations:
    mlp = MLPClassifier(random_state=42, max_iter=1000, **config)
    mlp.fit(X_train, y_train)
    score = mlp.score(X_test, y_test)
    print(f"Configuration: {config}, Score: {score}")

    if score > best_score:
        best_score = score
        best_model = config

print("\nBest Configuration:", best_model, "with Score:", best_score)
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data = pd.DataFrame({"sepal_length":[3], "sepal_width": [3], "petal_length": [1], "petal_width":[4]})
probabilities = knn.predict_proba(sample_data)

# iii. Report on the score for kNN, what does the score measure?
score = knn.score(X_test, y_test)

# Output the score
print("Model accuracy score:", score)
#The score (obtained using `knn.score`) measures the accuracy of the k-Neighbors Classifier. It is calculated as the proportion of correctly predicted instances to the total instances in the test set. Accuracy is a common measure used to evaluate the performance of classification models. In this case, it tells you how many of the predictions made by the kNN model match the actual labels in the test dataset.
```
