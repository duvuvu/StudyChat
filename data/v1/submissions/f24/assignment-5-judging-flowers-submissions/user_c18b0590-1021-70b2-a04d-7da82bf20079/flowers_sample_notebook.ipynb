```python
# Imports and pip installations (if needed)
import numpy as np
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, accuracy_score
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
data = pd.read_csv("iris.csv")
# Output the first 15 rows of the data
print("First 15 rows of the Iris dataset:")
print(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print("\nDataset Information:")
print(data.info())

print("\nClass distribution:")
print(data['species'].value_counts())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The Iris dataset is a classic dataset in machine learning that includes measurements of three species of iris flowers: Setosa, Versicolor, and Virginica. Each flower is described by four features: sepal length, sepal width, petal length, and petal width, all measured in centimeters. These measurements serve as input features to help distinguish between the species. The species labels (Setosa, Versicolor, Virginica) are the target variable, with 50 samples for each species, making the dataset balanced and well-suited for classification tasks. This dataset enables us to predict the species of an iris flower based on its physical dimensions.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)

X = data.drop(columns=['species']) 
y = data['species']  

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42, stratify=y)

print("Training set size:", X_train.shape[0])
print("Test set size:", X_test.shape[0])
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
log_reg = LogisticRegression(max_iter=200)
log_reg.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample = X_test.iloc[0] 
probabilities = log_reg.predict_proba([sample])

print("Predicted probabilities for each class for the sample datapoint:")
print(probabilities)
# iii. Report on the score for Logistic regression model, what does the score measure?
score = log_reg.score(X_test, y_test)
print("\nLogistic Regression model score:", score)
# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = log_reg.coef_
intercepts = log_reg.intercept_

print("\nCoefficients of the Logistic Regression model:")
print(coefficients)
print("\nIntercepts of the Logistic Regression model:")
print(intercepts)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svm = SVC(probability=True)  
svm.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample = X_test.iloc[0]  # Selecting the first datapoint in the test set
probabilities = svm.predict_proba([sample])

print("Predicted probabilities for each class for the sample datapoint:")
print(probabilities)
# iii. Report on the score for the SVM, what does the score measure?
score = svm.score(X_test, y_test)
print("\nSVM model score:", score)
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp = MLPClassifier(hidden_layer_sizes=(10,), max_iter=500, random_state=42)
mlp.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample = X_test.iloc[0] 
probabilities = mlp.predict_proba([sample])
print("Predicted probabilities for each class for the sample datapoint:")
print(probabilities)
# iii. Report on the score for the Neural Network, what does the score measure?
score = mlp.score(X_test, y_test)
print("\nNeural Network model score:", score)
# iv: Experiment with different options for the neural network, report on your best configuration
best_score = 0
best_config = None

for config in [(5,), (10,), (10, 10), (20, 20)]:
    mlp_test = MLPClassifier(hidden_layer_sizes=config, max_iter=500, random_state=42)
    mlp_test.fit(X_train, y_train)
    test_score = mlp_test.score(X_test, y_test)
    
    print(f"Configuration: {config}, Score: {test_score}")
    
    if test_score > best_score:
        best_score = test_score
        best_config = config

print(f"\nBest configuration: {best_config} with score: {best_score}")
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn = KNeighborsClassifier(n_neighbors=3) 
knn.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample = X_test.iloc[0]  
probabilities = knn.predict_proba([sample])

print("Predicted probabilities for each class for the sample datapoint:")
print(probabilities)
# iii. Report on the score for kNN, what does the score measure?
score = knn.score(X_test, y_test)
print("\nKNN model score:", score)
```

The results show that simpler models like Logistic Regression and SVM performed best on the Iris dataset, achieving high accuracy due to the dataset's linear separability.
