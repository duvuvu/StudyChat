```python
# Imports and pip installations (if needed)
import pandas as pd              

from sklearn.model_selection import train_test_split    
from sklearn.linear_model import LogisticRegression     
from sklearn.svm import SVC                        
from sklearn.neural_network import MLPClassifier 
from sklearn.neighbors import KNeighborsClassifier  
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)

# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
data = pd.read_csv("iris.csv")
print(data.head(15))
print(data.info())
```

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
x = data[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
y = data["species"]

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
logistic_model = LogisticRegression(max_iter=200)
logistic_model.fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample = pd.DataFrame({
    "sepal_length": [5], 
    "sepal_width": [3],
    "petal_length": [1],
    "petal_width": [0.2]
})

logistic_model_probabilities = logistic_model.predict_proba(sample)
logistic_model_prediction = logistic_model.predict(sample)[0]
labels = target_names = data["species"].unique()
print("Predicted Probabilities:")
for label, prob in zip(labels, logistic_model_probabilities[0]):
    print(f"Probability of {label}: {prob}")

print(f"Predicted class: {logistic_model_prediction}")
print()

# iii. Report on the score for Logistic regression model, what does the score measure?
logistic_model_score = logistic_model.score(x_test, y_test)
print("Logistic Regression Score:", logistic_model_score)
print()

# iv. Extract the coefficents and intercepts for the boundary line(s)
logistic_model_coefficients = logistic_model.coef_
logistic_model_intercepts = logistic_model.intercept_

def equation(coeffs, intercept):
    return str(intercept) + " + " + str(coeffs[0]) + "x_1 + " + str(coeffs[1]) + "x_2 + " + str(coeffs[2]) + "x_3+ " + str(coeffs[3]) + "x_4" 

for i in range(3):
    print("f_" + str(i+1) + " = " + equation(logistic_model_coefficients[i], logistic_model_intercepts[i]))
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svm = SVC(probability=True)
svm.fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
svm_probabilities = svm.predict_proba(sample)
svm_prediction = svm.predict(sample)[0]
print("Predicted Probabilities:")
for label, prob in zip(labels, svm_probabilities[0]):
    print(f"Probability of {label}: {prob}")

print(f"Predicted class: {svm_prediction}")
print()

# iii. Report on the score for the SVM, what does the score measure?
svm_score = svm.score(x_test, y_test)
print("SVM Score:", svm_score)
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp = MLPClassifier(hidden_layer_sizes=(10,), max_iter=5000, random_state=42)
mlp.fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
mlp_probabilities = mlp.predict_proba(sample)
mlp_prediction = mlp.predict(sample)[0]
print("Predicted Probabilities:")
for label, prob in zip(labels, mlp_probabilities[0]):
    print(f"Probability of {label}: {prob}")

print(f"Predicted class: {mlp_prediction}")
print()

# iii. Report on the score for the Neural Network, what does the score measure?
mlp_score = mlp.score(x_test, y_test)
print("Neural Network Score:", mlp_score)
print()

# iv: Experiment with different options for the neural network, report on your best configuration
mlp_2 = MLPClassifier(hidden_layer_sizes=(20,), max_iter=10000, random_state=42)
mlp_2.fit(x_train, y_train)

mlp_2_probabilities = mlp_2.predict_proba(sample)
mlp_2_prediction = mlp_2.predict(sample)[0]
print("Predicted Probabilities:")
for label, prob in zip(labels, mlp_2_probabilities[0]):
    print(f"Probability of {label}: {prob}")

print(f"Predicted class: {mlp_2_prediction}")
print()

mlp_2_score = mlp_2.score(x_test, y_test)
print("Alternative Neural Network Score:", mlp_2_score)
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
knn_probabilities = knn.predict_proba(sample)
knn_prediction = knn.predict(sample)[0]
print("Predicted Probabilities:")
for label, prob in zip(labels, knn_probabilities[0]):
    print(f"Probability of {label}: {prob}")

print(f"Predicted class: {knn_prediction}")
print()

# iii. Report on the score for kNN, what does the score measure?
knn_score = knn.score(x_test, y_test)
print("kNN Score:", knn_score)
```
