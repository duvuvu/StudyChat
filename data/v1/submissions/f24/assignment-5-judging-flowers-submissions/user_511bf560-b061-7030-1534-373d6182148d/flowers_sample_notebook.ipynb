```python
# Imports and pip installations (if needed)
import numpy as np
import pandas as pd

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
df = pd.read_csv('iris.csv')
# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(df.info())
```

## About the dataset
The dataset contains 150 entries and 5 columns. The first four columns represent the attributes of the Iris flowers (these are the features): sepal_length, sepal_width, petal_length, petal_width. The label is the 5th column, species. The species column contains the following three classes of Iris flowers: setosa, versicolor, and virginica. The goal of the classification task is to predict the species of the flower (label) based on its features (measurements).

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = df.drop('species', axis=1)  
y = df['species']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
#Displaying the shape of the resulting dataset
print("Training feature set shape:", X_train.shape)
print("Test feature set shape:", X_test.shape)
print("Training label set shape:", y_train.shape)
print("Test label set shape:", y_test.shape)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
logistic_model = LogisticRegression(max_iter=200) 
logistic_model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data_point = X_test.iloc[[0]] 
predicted_probabilities = logistic_model.predict_proba(sample_data_point)
print("Predicted probabilities for each class:", predicted_probabilities)
# iii. Report on the score for Logistic regression model, what does the score measure?
y_pred = logistic_model.predict(X_test)  
score = accuracy_score(y_test, y_pred)    
print("Accuracy score of the Logistic Regression model:", score)
print("The accuracy score measures the proportion of correctly classified samples in the test set.")
# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = logistic_model.coef_         
intercepts = logistic_model.intercept_       
print("Coefficients for each class:\n", coefficients)
print("Intercepts for each class:\n", intercepts)
```

In this case, the accuracy score is 1. This means all the samples in the test set were classified correctly!

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svm_model = SVC(probability=True, random_state=42)
svm_model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data_point = X_test.iloc[[0]] 
predicted_probabilities = svm_model.predict_proba(sample_data_point)
print("Predicted probabilities for each class:", predicted_probabilities)
# iii. Report on the score for the SVM, what does the score measure?
y_pred = svm_model.predict(X_test)  
score = accuracy_score(y_test, y_pred)  
print("Accuracy score of the Support Vector Classifier model:", score)
```

In this case, the accuracy score is 1. This means all the samples in the test set were classified correctly!

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(max_iter=1000, random_state=42)  
mlp_model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data_point = X_test.iloc[[0]]  
predicted_probabilities = mlp_model.predict_proba(sample_data_point)
print("Predicted probabilities for each class:", predicted_probabilities)
# iii. Report on the score for the Neural Network, what does the score measure?
y_pred = mlp_model.predict(X_test)  # Predict labels for the test set
score = accuracy_score(y_test, y_pred)
print("Accuracy score of the Neural Network model:", score)
# iv: Experiment with different options for the neural network, report on your best configuration
best_score = 0
best_params = {}

for hidden_layers in [(5,), (10,), (5, 5), (10, 10), (5, 10)]:
    for activation in ['identity', 'logistic', 'tanh', 'relu']:
        model = MLPClassifier(hidden_layer_sizes=hidden_layers, 
                              activation=activation, 
                              max_iter=1000, 
                              random_state=42)
        model.fit(X_train, y_train)
        score = model.score(X_test, y_test)

        print(f"Testing configuration - Hidden layers: {hidden_layers}, Activation: {activation}")
        print("Accuracy score:", score)
        if score > best_score:
            best_score = score
            best_params = {'hidden_layers': hidden_layers, 'activation': activation}

print("Best Configuration:")
print("Hidden layers:", best_params['hidden_layers'])
print("Activation function:", best_params['activation'])
print("Best accuracy score:", best_score)
```

The probabilities recieved seem reasonable, given that this dataset is well separated in feature space. The accuracy score of 1.0 means that the model perfectly classified all instances in the test set. Given that the Iris dataset is relatively small, simple, and clean, achieving perfect accuracy does not seem unusual. The warnings indicating that the maximum number of iterations was reached without convergence. It suggests fine-tuning may be needed. 

The best configuration :
Testing configuration - Hidden layers: (5, 5), Activation: tanh
Accuracy score: 1.0

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn_model = KNeighborsClassifier(n_neighbors=3)  
knn_model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data_point = X_test.iloc[[0]] 
predicted_probabilities = knn_model.predict_proba(sample_data_point)
print("Predicted probabilities for each class:", predicted_probabilities)
# iii. Report on the score for kNN, what does the score measure?
y_pred = knn_model.predict(X_test)
score = accuracy_score(y_test, y_pred) 
print("Accuracy score of the k-Neighbors Classifier model:", score)
```

The accuracy score measures the proportion of correctly classified samples in the test set.
