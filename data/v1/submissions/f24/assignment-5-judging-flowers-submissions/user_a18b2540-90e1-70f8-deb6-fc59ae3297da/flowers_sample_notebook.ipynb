```python
# Imports and pip installations (if needed)
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
df = pd.read_csv("iris.csv")
# Output the first 15 rows of the data
display(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
display(df.describe())
```

The features are sepal_length, sepal_width, petal_length, and petal_width. The label is the species of flower.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = df.drop('species', axis=1)
y = df['species']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.9, test_size=0.1, random_state=42)
display(X_train)
display(y_train)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
logistic_reg_model = LogisticRegression(max_iter=85, random_state=42).fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
median_features = np.median(X_train, axis=0).reshape(1, -1)
log_reg_y_pred = logistic_reg_model.predict_proba(median_features) 
    # Used the mean of both the sepal and petal length and width
print("Probabilities of each class:", log_reg_y_pred)
# iii. Report on the score for Logistic regression model, what does the score measure?
train_score = logistic_reg_model.score(X_test, y_test)
print("Score:", train_score)
# iv. Extract the coefficents and intercepts for the boundary line(s)
print("Coefficients:", logistic_reg_model.coef_)
print("Intercepts:", logistic_reg_model.intercept_)
```

The score measures the accuracy of the logistic regression model using the test data

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc = SVC(probability=True).fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
median_features = np.median(X_train, axis=0).reshape(1, -1)
svc_y_pred = svc.predict_proba(median_features)
print(svc_y_pred)
    # Used the mean of both the sepal and petal length and width
# iii. Report on the score for the SVM, what does the score measure?
svc_test_score = svc.score(X_test, y_test)
print(svc_test_score)
```

The score measures the accuracy of the support vector classifier using the test data

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
clf = MLPClassifier(random_state=1, max_iter=300).fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
median_features = np.median(X_train, axis=0).reshape(1, -1)
clf_y_pred = clf.predict_proba(median_features)
print(clf_y_pred)
# iii. Report on the score for the Neural Network, what does the score measure?
clf_test_score = clf.score(X_test, y_test)
print(clf_test_score)
# iv: Experiment with different options for the neural network, report on your best configuration
```

The score measures the accuracy of the neural network using the test data

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn = KNeighborsClassifier(n_neighbors=3).fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
median_features = np.median(X_train, axis=0).reshape(1, -1)
knn_y_pred = knn.predict_proba(median_features)
print(knn_y_pred)
# iii. Report on the score for kNN, what does the score measure?
knn_test_score = knn.score(X_test, y_test)
print(knn_test_score)
```

The score measures the accuracy of the k-nearest neighbors classifier using the test data
