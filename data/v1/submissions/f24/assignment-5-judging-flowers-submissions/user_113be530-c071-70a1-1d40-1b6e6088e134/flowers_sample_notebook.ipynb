# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

# Create the kNN model with a specific number of neighbors (e.g., k=5)
knn_model = KNeighborsClassifier(n_neighbors=5)

# Fit the model on the training data (this step essentially loads the data into the model)
knn_model.fit(X_train, y_train.values.ravel())

# ii. For a sample datapoint, predict the probabilities for each possible class
probabilities = knn_model.predict_proba(sample)
print("Predicted probabilities for the sample:", probabilities)

# iii. Report on the score for kNN, what does the score measure?
knn_accuracy = knn_model.score(X_test, y_test)
print("Score (accuracy) of the kNN model:", knn_accuracy)
```

Explanation of the Score:

The accuracy score measures the proportion of correct predictions made by the kNN model on the test data. It shows how well the model classifies the samples into the correct classes.
Accuracy is defined as:

Accuracy = Number of Correct Predictions / Total Number of Predictions
​

 
Additional Notes:
Nonparametric Nature: kNN is a nonparametric model, meaning it doesn't learn explicit parameters during training. Instead, it memorizes the training data and makes predictions by finding the k nearest data points and using them to classify new samples.
Class Probabilities: The predict_proba() method returns the probability of each class for a given input, calculated as the fraction of k nearest neighbors belonging to each class.
