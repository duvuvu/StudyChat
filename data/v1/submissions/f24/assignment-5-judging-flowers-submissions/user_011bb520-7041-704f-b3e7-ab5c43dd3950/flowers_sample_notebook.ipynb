```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
data = pd.read_csv('iris.csv')
# Output the first 15 rows of the data
print("First 15 rows")
print(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print('\nSummary')
print(data.describe())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The data is 4 measurements taken from flowers and their species. The measurements/features are sepal width, sepal length, petal length, and petal width. Their species/labels are setosa, versicolor, and verginica. Setosa corresponds with Iris-Setosa, versicolor corresponds with Iris-Versicolor, and verginica corresponds with Iris-Virginica.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = data.drop(columns=['species'])
y = data['species']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model = LogisticRegression()
model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = X_test.iloc[0:1]
print('Sample:', sample_datapoint)
predicted_output = model.predict(sample_datapoint)
print('Prediction:', predicted_output)
# iii. Report on the score for Logistic regression model, what does the score measure?
score = model.score(X_test, y_test)
print('Score:', score)
# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = model.coef_
intercept = model.intercept_
print('Intercept:', intercept, '\nCoefficients:', coefficients)
```

A score of 1 means that the model made the correct prediction for 100% of the data.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
svm_classifier = SVC(kernel='linear', probability=True)

svm_classifier.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = X_test[0:1]
print('Sample:', sample_datapoint)
# sample_datapoint = scaler.transform(sample_datapoint)
probabilities = svm_classifier.predict_proba(sample_datapoint)

# Show the predicted probabilities
print("Predicted probabilities for each class:", probabilities)
# iii. Report on the score for the SVM, what does the score measure?
score = svm_classifier.score(X_test, y_test)
print('Score:', score)
```

The score of 0.933 implies that the SVM made correct predictions for 93.3% of the test data

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp = MLPClassifier(hidden_layer_sizes=(100,), max_iter=1000, random_state=42)
mlp.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
predicted_probabilities = mlp.predict_proba(sample_datapoint)
print("Predicted probabilities for each class:", predicted_probabilities)
# iii. Report on the score for the Neural Network, what does the score measure?
score = mlp.score(X_test, y_test)
print("Model Score (Accuracy):", score)
# iv: Experiment with different options for the neural network, report on your best configuration
# A list of different configurations to test
configs = [
    {'hidden_layer_sizes': (1,), 'activation': 'relu'},
    {'hidden_layer_sizes': (3,), 'activation': 'relu'},
    {'hidden_layer_sizes': (3,3), 'activation': 'relu'},
    {'hidden_layer_sizes': (10,), 'activation': 'relu'},
    {'hidden_layer_sizes': (10,10,10), 'activation': 'relu'},
    {'hidden_layer_sizes': (50,5), 'activation': 'relu'},
    {'hidden_layer_sizes': (100,), 'activation': 'relu'},
    {'hidden_layer_sizes': (50, 50), 'activation': 'relu'},
    {'hidden_layer_sizes': (100,), 'activation': 'tanh'},
    {'hidden_layer_sizes': (100, 50), 'activation': 'logistic'},
    {'hidden_layer_sizes': (30, 30, 30), 'activation': 'logistic'},
]

best_score = 0
best_configuration = None

for config in configs:
    mlp = MLPClassifier(hidden_layer_sizes=config['hidden_layer_sizes'], 
                        activation=config['activation'], 
                        max_iter=1000, 
                        random_state=42)
    mlp.fit(X_train, y_train)
    score = mlp.score(X_test, y_test)
    print(f"Configuration: {config}, Score: {score}")

    if score > best_score:
        best_score = score
        best_configuration = config

print("Best Configuration:", best_configuration)
print("Best Score:", best_score)
```

The neural network had a score of 1.0, meaning it predicted correctly for 100% of the test data. A few configurations had a score of 1.0, the simplest one had 10 nodes in its hidden layer and a Rectified Linear Unit activation function.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn = KNeighborsClassifier(n_neighbors=5)  # You can adjust the number of neighbors (k)
knn.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
predicted_probabilities = knn.predict_proba(sample_datapoint)

print("Predicted probabilities for each class:", predicted_probabilities)
# iii. Report on the score for kNN, what does the score measure?
score = knn.score(X_test, y_test)
print("Model Score (Accuracy):", score)
```

The model's score was 1.0, meaning it predicted the correct class 100% of the time
