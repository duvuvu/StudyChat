```python
# Imports and pip installations (if needed)
import sklearn
import pandas as pd
import sklearn.model_selection
import sklearn.linear_model
import sklearn.svm
import sklearn.neural_network
import sklearn.neighbors
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
ds = pd.read_csv("iris.csv")
# shuffling the data
ds = ds.sample(frac = 1)
# Output the first 15 rows of the data
print(ds.head(n=15))
# Display a summary of the table information (number of datapoints, etc.)
ds.info()
print(ds.shape)
print(ds.describe())
```

## About the dataset
The data represents the dimensions of the petals and sepals (modified leaves that resemble petals in orchids) of various photos of orchids as the features and labels each with the correct species of orchid represented in the photo. No null values are present, and there are 150 such processed representations of photos in the dataset. Before being split, the dataset thus has 150 rows and 5 columns. 

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)

X = ds.loc[:, ["sepal_length", "sepal_width", "petal_length", "petal_width"]]
y = ds["species"]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
Xtrain, Xtest, ytrain, ytest = sklearn.model_selection.train_test_split(X, y, test_size = 0.1)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
logreg = sklearn.linear_model.LogisticRegression(max_iter=200)
logreg.fit(Xtrain, ytrain)
# ii. For a sample datapoint, predict the probabilities for each possible class
sampledp, samplecorrect = Xtest.iloc[[5]], ytest.iloc[[5]]
sampleresult = logreg.predict(sampledp)
print("The model predicted that the image was of a ", sampleresult[0], " iris. The image was of a ", 
      samplecorrect.iat[0], "iris. ")
# iii. Report on the score for Logistic regression model, what does the score measure?
logregscore = logreg.score(Xtest, ytest)
print("The score for the logistic regression model is ", logregscore)
# iv. Extract the coefficents and intercepts for the boundary line(s)
print("Boundary line coefficients: ", logreg.coef_)
print("Boundary line intercepts: ", logreg.intercept_)
```

The score for the model was 0.93333, indicating that the model was 93.333% accurate at predicting an orchid's species given its petal and sepal dimensions even when that orchid had not been trained on. The boundary coefficients and intercepts are listed above.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc = sklearn.svm.SVC()
svc.fit(Xtrain, ytrain)
# ii. For a sample datapoint, predict the probabilities for each possible class
sampledp, samplecorrect = Xtest.iloc[[11]], ytest.iloc[[11]]
sampleresult = svc.predict(sampledp)
print("The model predicted that the image was of a ", sampleresult[0], " iris. The image was of a ", 
      samplecorrect.iat[0], "iris. ")
# iii. Report on the score for the SVM, what does the score measure?
# TODO: answer the questions
svcscore = svc.score(Xtest, ytest)
print(svcscore)
```

The SVM got a score of 0.93333, indicating that when presented with new data from the testing set, the model was accurate 93.333% of the time in identifying the correct species.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
nn = sklearn.neural_network.MLPClassifier()
nn.fit(Xtrain, ytrain)
# ii. For a sample datapoint, predict the probabilities f6or each possible class
sampledp, samplecorrect = Xtest.iloc[[14]], ytest.iloc[[14]]
sampleresult = svc.predict(sampledp)
print("The model predicted that the image was of a ", sampleresult[0], " iris. The image was of a ", 
      samplecorrect.iat[0], "iris. ")
# iii. Report on the score for the Neural Network, what does the score measure?
#TODO: Answer questions
nnscore = nn.score(Xtest, ytest)
print(nnscore)
# iv: Experiment with different options for the neural network, report on your best configuration
nn2 = sklearn.neural_network.MLPClassifier(activation = 'relu', learning_rate='adaptive', max_iter=3000)
nn2.fit(Xtrain, ytrain)
sampledp, samplecorrect = Xtest.iloc[[14]], ytest.iloc[[14]]
sampleresult = svc.predict(sampledp)
print("The model predicted that the image was of a ", sampleresult[0], " iris. The image was of a ", 
      samplecorrect.iat[0], "iris. ")
nn2score = nn2.score(Xtest, ytest)
print(nn2score)
```
