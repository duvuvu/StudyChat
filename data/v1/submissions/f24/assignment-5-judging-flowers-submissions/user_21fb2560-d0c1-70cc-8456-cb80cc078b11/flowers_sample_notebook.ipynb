```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
df = pd.read_csv('iris.csv')
# Output the first 15 rows of the data
print(df.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print()
print(df.info())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

Answer: The Iris dataset constains data on the length and width of the sepals and petals of three different types of Iris flowers. The features are 'sepal_length', 'sepal_width', 'petal_length', and 'petal_width'. The label is 'species', as in which of the three types of flowers do the features correspond to. 
Each unique type of label, that is the type of flower ('setosa', 'versicolor', 'verginica'), maps to one class of the dataset.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = df.iloc[:,:-1] #featurs
y = df.iloc[:,-1] #label

print(X)
print()
print(y)

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model = LogisticRegression(random_state=42)

model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data_point = pd.DataFrame({'sepal_length': [6.3], 'sepal_width': [3.0], 'petal_length': [4.5], 'petal_width': [1.8]})

probabilities = model.predict_proba(sample_data_point)

class_labels = model.classes_

for index, prob in enumerate(probabilities[0]):
    print(f"The probability of class `{class_labels[index]}` is `{prob}`")

print()

# iii. Report on the score for Logistic regression model, what does the score measure?
score = model.score(X_test, y_test)

print(f"The accuracy of the Logistic Regression model is: {score}")
print()

# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = model.coef_ 
intercepts = model.intercept_ 

for i in range(coefficients.shape[0]):
    print(f"Coefficients for class {class_labels[i]}: {coefficients[i]}")
    print(f"Intercept for class {class_labels[i]}: {intercepts[i]}")
    print()
```

# 3iii. Report on the score for Logistic regression model, what does the score measure?
The score returned by the "score" method represents the accuracy of the model. It is calculated as the number of correct predictions divided by the total number of predictions. In other words, it measures how often the classifier is correct. We got a score of 1.0 for the Logistic Regression model, this indicates that the model made perfect predictions on the test dataset, meaning it correctly classified every instance without any errors, which suggests that the model has achieved full accuracy and effectively learned to distinguish between the different classes in the dataset.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

model2 = SVC(probability=True, random_state=42)
model2.fit(X_train_scaled, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data_point = pd.DataFrame({'sepal_length': [6.3], 'sepal_width': [3.0], 'petal_length': [4.5], 'petal_width': [1.8]})
sample_data_point_scaled = scaler.transform(sample_data_point)

probabilities = model2.predict_proba(sample_data_point_scaled)

class_labels = model2.classes_

for index, prob in enumerate(probabilities[0]):
    print(f"The probability of class `{class_labels[index]}` is `{prob}`")
print()
# iii. Report on the score for the SVM, what does the score measure?
y_pred = model2.predict(X_test_scaled)

accuracy = accuracy_score(y_test, y_pred)

print(f"The accuracy of the Support Vector Classifier is: {accuracy}")
```

# 4iii. Report on the score for the SVM, what does the score measure?
The accuracy score returned by the "accuracy_score" function represents the proportion of correct predictions made by the model. It is calculated as the number of correct predictions divided by the total number of predictions. In other words, it measures how often the classifier is correct. We obtained a score of 1.0 for the Support Vector Machine (SVM) model, indicating that the model accurately predicted every instance in the test dataset without any errors. This suggests that the SVM has achieved full accuracy and effectively learned to distinguish between the different classes in the dataset.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(max_iter=500, random_state=42)
mlp_model.fit(X_train_scaled, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data_point = pd.DataFrame({'sepal_length': [6.3], 'sepal_width': [3.0], 'petal_length': [4.5], 'petal_width': [1.8]})
sample_data_point_scaled = scaler.transform(sample_data_point)

probabilities = mlp_model.predict_proba(sample_data_point_scaled)

class_labels = mlp_model.classes_

for index, prob in enumerate(probabilities[0]):
    print(f"The probability of class `{class_labels[index]}` is `{prob}`")
    
# iii. Report on the score for the Neural Network, what does the score measure?
mlp_pred = mlp_model.predict(X_test_scaled)
mlp_accuracy = accuracy_score(y_test, mlp_pred)
print()
print(f"MLP Accuracy: {mlp_accuracy}")
print()

# iv: Experiment with different options for the neural network, report on your best configuration
print("Experiment case 1")
mlp_model = MLPClassifier(hidden_layer_sizes=1, max_iter=100, random_state=42)
mlp_model.fit(X_train_scaled, y_train)

mlp_pred = mlp_model.predict(X_test_scaled)
mlp_accuracy = accuracy_score(y_test, mlp_pred)
print(f"MLP Accuracy: {mlp_accuracy}")
print()

print("Experiment case 2")
mlp_model = MLPClassifier(hidden_layer_sizes=1, max_iter=1000, random_state=42)
mlp_model.fit(X_train_scaled, y_train)

mlp_pred = mlp_model.predict(X_test_scaled)
mlp_accuracy = accuracy_score(y_test, mlp_pred)
print(f"MLP Accuracy: {mlp_accuracy}")
print()

print("Experiment case 3")
mlp_model = MLPClassifier(hidden_layer_sizes=1000, max_iter=1, random_state=42)
mlp_model.fit(X_train_scaled, y_train)

mlp_pred = mlp_model.predict(X_test_scaled)
mlp_accuracy = accuracy_score(y_test, mlp_pred)
print(f"MLP Accuracy: {mlp_accuracy}")
print()
```

# 5iii. Report on the score for the Neural Network, what does the score measure?
The accuracy score returned by the "accuracy_score" function represents the proportion of correct predictions made by the model. It is calculated as the number of correct predictions divided by the total number of predictions. In other words, it measures how often the classifier is correct. We obtained a score of 1.0 for the Multi-Layer Perceptron (MLP) model, indicating that the model accurately predicted every instance in the test dataset without any errors. This suggests that the MLP model has achieved full accuracy and effectively learned to distinguish between the different classes in the dataset.

# 5iv: Experiment with different options for the neural network, report on your best configuration
After running a few experiments with the MLP model, it seems like the most important parameter is "hidden_layer_sizes" because the configuration that turned out to be the most successful had a high "hidden_layer_sizes", and succeeded to an extent even though I had set the "max_iter" to a very low value of 1. On the other hand, the model with a very high "max_iter" and a very low "hidden_layer_sizes", wasn't as successful. This makes sense because the size and depth of the hidden layers directly influence the model's capacity to learn complex patterns in the data. A larger number of hidden neurons/perceptrons allows the model to capture more intricate relationships between features, while insufficient hidden units may lead to underfitting, even if the training process runs for many iterations.  
Obviously, the best model would be one that has a high value for both parameters. 

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

k = 5
knn_model = KNeighborsClassifier(n_neighbors=k)
knn_model.fit(X_train_scaled, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_data_point = pd.DataFrame({'sepal_length': [6.3], 'sepal_width': [3.0], 'petal_length': [4.5], 'petal_width': [1.8]})
sample_data_point_scaled = scaler.transform(sample_data_point)

probabilities = knn_model.predict_proba(sample_data_point_scaled)

class_labels = knn_model.classes_

for index, prob in enumerate(probabilities[0]):
    print(f"The probability of class `{class_labels[index]}` is `{prob}`")

# iii. Report on the score for kNN, what does the score measure?
accuracy = knn_model.score(X_test_scaled, y_test)
print()
print(f"The accuracy of the k-Neighbors Classifier with k={k} is: {accuracy}")
```

# 6iii. Report on the score for kNN, what does the score measure?
The score returned by the "score" method represents the accuracy of the model. It is calculated as the number of correct predictions divided by the total number of predictions. In other words, it measures how often the classifier is correct. We got a score of 1.0 for the K-Nearest Neighbors model, this indicates that the model made perfect predictions on the test dataset, meaning it correctly classified every instance without any errors, which suggests that the KNN model has achieved full accuracy and effectively learned to distinguish between the different classes in the dataset.
