# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
iris = pd.read_csv('/content/drive/MyDrive/University/Academics/Fall 2024/CS 383/Projects/CS383_P5/iris.csv')
# Output the first 15 rows of the data
display(iris.head(15))
# Display a summary of the table information (number of datapoints, etc.)
display(iris.info())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The dataset used is the Iris dataset, which contains measurements of iris flowers from three different species: Setosa, Versicolor, and Virginica.
The features in the dataset are:
- sepal_length: Length of the sepal in centimeters.
- sepal_width: Width of the sepal in centimeters.
- petal_length: Length of the petal in centimeters.
- petal_width: Width of the petal in centimeters.

The label is the species of the iris flower, which can be one of the following:
- Setosa mapped as 0
- Versicolor mapped as 1
- Virginica mapped as 2

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = iris[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = iris['species']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=50)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame([[5.1, 3.5, 1.4, 0.2]], columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
sample_datapoint_proba = logreg.predict_proba(sample_datapoint)
print("Logistic Regression Prediction for Sample:", sample_datapoint_proba)

# iii. Report on the score for Logistic regression model, what does the score measure?
logreg_score = logreg.score(X_test, y_test)
display("Logistic Regression Score:", logreg_score)

# iv. Extract the coefficients and intercepts for the boundary line(s)
logreg_coefficients = logreg.coef_
logreg_intercepts = logreg.intercept_
print("Logistic Regression Coefficients:", logreg_coefficients)
print("Logistic Regression Intercepts:", logreg_intercepts)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svm = SVC(probability=True)  # enabling probability if needed
svm.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame([[5.1, 3.5, 1.4, 0.2]], columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
sample_datapoint_proba_svm = svm.predict_proba(sample_datapoint)
print("SVM Prediction for sample:", sample_datapoint_proba_svm)

# iii. Report on the score for the SVM, what does the score measure?
svm_score = svm.score(X_test, y_test)
print("SVM Score:", svm_score)
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp = MLPClassifier(max_iter=1000, hidden_layer_sizes=(100), activation='tanh')
mlp.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame([[5.1, 3.5, 1.4, 0.2]], columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
sample_datapoint_proba = mlp.predict_proba(sample_datapoint)
print("MLPClassifier Prediction for Sample: ", sample_datapoint_proba)

# iii. Report on the score for the Neural Network, what does the score measure?
score = mlp.score(X_test, y_test)
print("Score: ",score)

# iv: Experiment with different options for the neural network, report on your best configuration
# THe default MLPClassifier was getting a score of 0.97. Experimenting with the max_iter parameter = 5 is 0.4 accuracy, 50 is 0.733, 500 is score 0.866.0
# I also experimented with the activation function and tanh gave a higher accuracy of 0.933
# Changing the learning rate alpha had no apparant effect on the acuracy
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame([[5.1, 3.5, 1.4, 0.2]], columns=['sepal_length', 'sepal_width', 'petal_length', 'petal_width'])
sample_datapoint_proba = knn.predict_proba(sample_datapoint)
print("KNN Prediction for Sample: ",sample_datapoint_proba)

# iii. Report on the score for kNN, what does the score measure?
score = knn.score(X_test, y_test)
print("Score: ",score)
```
