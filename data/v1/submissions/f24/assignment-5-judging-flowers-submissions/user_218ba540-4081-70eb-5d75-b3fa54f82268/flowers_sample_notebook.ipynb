```python
# Imports and pip installations (if needed)
!python -m venv sklearn-env
!source sklearn-env/bin/activate
%pip install -U scikit-learn
# Imports section
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
df = pd.read_csv('iris.csv')
print(df.head(15))
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
summary_stats = df.describe()
print(summary_stats)
print(df.dtypes)
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

* sepal_length: The length of the flower's sepal. (feature)

* sepal_width: The width of the flower's sepal. (feature)

* petal_length: The length of the pedal. (feature)

* petal_width: The width of the petal. (feature)

* species: The species of the flower. (label)

The mapping of labels:

0: setosa

1: versicolor

2: virginica


# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = df.loc[:, df.columns != 'species']
y = df['species']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model = LogisticRegression(max_iter=200)
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_index = 0  # Change this index to select a different sample
sample_data = X_test.iloc[sample_index].values.reshape(1, -1)  # Reshape it into a 2D array
# Convert back to DataFrame with feature names
sample_data_df = pd.DataFrame(sample_data, columns=X_test.columns)
probabilities = model.predict_proba(sample_data_df)
print(f'Sample data: {sample_data}')
print(f'Predicted probabilities for the sample data point: {probabilities}')

predicted_class = np.argmax(probabilities, axis=1)
print(f'Predicted Class: {predicted_class}')
class_labels = model.classes_  # This will give you the class labels
print("Class Labels:", class_labels)
predicted_class_name = class_labels[predicted_class[0]]
print(f'Predicted Class (Name): {predicted_class_name}')

# iii. Report on the score for Logistic regression model, what does the score measure?
scores = cross_val_score(model, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')

"""
    Report: The score of accuracy on the test data is: 0.9625541125541126
"""

# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
model = SVC(kernel='linear', probability=True)  # Use a linear kernel; 'probability=True' allows for probability estimates
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_index = 0  # Change this index to select a different sample
sample_data = X_test.iloc[sample_index].values.reshape(1, -1)  # Reshape it into a 2D array
# Convert back to DataFrame with feature names
sample_data_df = pd.DataFrame(sample_data, columns=X_test.columns)
probabilities = model.predict_proba(sample_data_df)
print(f'Sample data: {sample_data}')
print(f'Predicted probabilities for the sample data point: {probabilities}')

predicted_class = np.argmax(probabilities, axis=1)
print(f'Predicted Class: {predicted_class}')
class_labels = model.classes_  # This will give you the class labels
print("Class Labels:", class_labels)
predicted_class_name = class_labels[predicted_class[0]]
print(f'Predicted Class (Name): {predicted_class_name}')

# iii. Report on the score for the SVM, what does the score measure?
# Perform 10-fold cross-validation to evaluate the model
scores = cross_val_score(model, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')

"""
    Report: The score of accuracy on the test data is: 0.9718614718614719
"""

# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = model.coef_
intercept = model.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, random_state=42)  # You can customize the architecture
mlp_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_index = 0  # Change this index to select a different sample
sample_data = X_test.iloc[sample_index].values.reshape(1, -1)  # Reshape it into a 2D array
# Convert back to DataFrame with feature names
sample_data_df = pd.DataFrame(sample_data, columns=X_test.columns)
probabilities = mlp_model.predict_proba(sample_data_df)
print(f'Sample data: {sample_data}')
print(f'Predicted probabilities for the sample data point: {probabilities}')

predicted_class = np.argmax(probabilities, axis=1)
print(f'Predicted Class: {predicted_class}')
class_labels = mlp_model.classes_  # This will give you the class labels
print("Class Labels:", class_labels)
predicted_class_name = class_labels[predicted_class[0]]
print(f'Predicted Class (Name): {predicted_class_name}')

# iii. Report on the score for the Neural Network, what does the score measure?
# Perform 10-fold cross-validation to evaluate the model
scores = cross_val_score(mlp_model, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')

"""
    Report: The score of accuracy on the test data is: 0.9764069264069264
"""

# iv: Experiment with different options for the neural network, report on your best configuration
print()

# Experiment 1: Default Settings
mlp_default = MLPClassifier(max_iter=1000, random_state=42)
mlp_default.fit(X_train, y_train)
y_pred_default = mlp_default.predict(X_test)
print("Experiment 1: Default Settings")
scores = cross_val_score(mlp_default, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')
print()

# Experiment 2: Increase the number of hidden layers and neurons
mlp_layers = MLPClassifier(hidden_layer_sizes=(100, 50, 25), max_iter=1000, random_state=42)
mlp_layers.fit(X_train, y_train)
y_pred_layers = mlp_layers.predict(X_test)
print("Experiment 2: Increased Hidden Layers")
scores = cross_val_score(mlp_layers, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')
print()

# Experiment 3: Change the activation function
mlp_tanh = MLPClassifier(hidden_layer_sizes=(100, 50), activation='tanh', max_iter=1000, random_state=42)
mlp_tanh.fit(X_train, y_train)
y_pred_tanh = mlp_tanh.predict(X_test)
print("Experiment 3: Tanh Activation Function")
scores = cross_val_score(mlp_tanh, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')
print()

# Experiment 4: Change the learning rate initialization
mlp_lr = MLPClassifier(hidden_layer_sizes=(100, 50), learning_rate_init=0.01, max_iter=1000, random_state=42)
mlp_lr.fit(X_train, y_train)
y_pred_lr = mlp_lr.predict(X_test)
print("Experiment 4: Increased Learning Rate Initialization")
scores = cross_val_score(mlp_lr, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')
print()

# Experiment 5: Add L2 regularization
mlp_alpha = MLPClassifier(hidden_layer_sizes=(100, 50), alpha=0.1, max_iter=1000, random_state=42)
mlp_alpha.fit(X_train, y_train)
y_pred_alpha = mlp_alpha.predict(X_test)
print("Experiment 5: L2 Regularization")
scores = cross_val_score(mlp_alpha, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')
print()

# Experiment 6: Use Early Stopping
mlp_early_stopping = MLPClassifier(hidden_layer_sizes=(100, 50), max_iter=1000, early_stopping=True, random_state=42)
mlp_early_stopping.fit(X_train, y_train)
y_pred_early_stopping = mlp_early_stopping.predict(X_test)
print("Experiment 6: Early Stopping")
scores = cross_val_score(mlp_early_stopping, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')
print()

# Experiment 7: Use a different Solver (e.g. 'adam')
mlp_solver = MLPClassifier(hidden_layer_sizes=(100, 50), solver='adam', max_iter=1000, random_state=42)
mlp_solver.fit(X_train, y_train)
y_pred_solver = mlp_solver.predict(X_test)
print("Experiment 7: Using 'adam' Solver")
scores = cross_val_score(mlp_solver, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')
print()

# Experiment 8: Adaptive learning rate
mlp_adaptive = MLPClassifier(hidden_layer_sizes=(100, 50), learning_rate='adaptive', max_iter=1000, random_state=42)
mlp_adaptive.fit(X_train, y_train)
y_pred_adaptive = mlp_adaptive.predict(X_test)
print("Experiment 8: Using adaptive learning rate")
scores = cross_val_score(mlp_adaptive, X_train, y_train, cv=10)
# Output the cross-validation scores and mean accuracy
print("Cross-Validation Scores:", scores)
print(f'Mean Cross-Validation Accuracy: {np.mean(scores)}')

"""
    Report: experiment 1 using the default setting and no hidden layer has the best performance out of all
    8 experiments, with an accuracy of 0.9811688311688311.
"""
```

# Part 6: K-Nearest Neighbors
