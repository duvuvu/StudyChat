```python
# Imports and pip installations (if needed)
import pandas as pd
import sklearn
from sklearn import *
rand_state = 42
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
flower_data = pd.read_csv("iris.csv")
# Output the first 15 rows of the data
print(flower_data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
flower_data.info()
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The data contains the length and width of different parts of measured iris flowers (sepal and petal, specifically). It also has the specific species of iris that those flowers belong to.
The features would be sepal_length, sepal_width, petal_length, and petal_width. The label would be the species.
The labels "setosa", "versicolor", and "virginica" represent 3 species of iris, Iris setosa, Iris versicolor, and Iris verginica.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = flower_data.iloc[:,0:4]
print(X)
Y = flower_data.iloc[:,4]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, Y_train, Y_test = sklearn.model_selection.train_test_split(X, Y, test_size=0.1, random_state=rand_state)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
LRmodel = sklearn.linear_model.LogisticRegression().fit(X_train, Y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
LR_probability_predictions = LRmodel.predict_proba(X_test)
print(LR_probability_predictions)
# iii. Report on the score for Logistic regression model, what does the score measure?
LR_Score = LRmodel.score(X_test,Y_test)
print(LR_Score)
# iv. Extract the coefficents and intercepts for the boundary line(s)
LRcoefficient = LRmodel.coef_
LRintercept = LRmodel.intercept_
print(LRcoefficient)
print(LRintercept)
```

What does the score measure?

The score measures the mean accuracy of how well the model predicts the correct categories. This is a value between 0 and 1, representing the percentage of test cases that were correctly categorized.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
SVC_model = sklearn.svm.SVC(probability = True).fit(X_train, Y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
SVC_probability_predictions = SVC_model.predict_proba(X_test)
print(SVC_probability_predictions)
# iii. Report on the score for the SVM, what does the score measure?
SVC_Score = SVC_model.score(X_test,Y_test)
print(SVC_Score)
```

What does the score measure?


The score measures the mean accuracy of how well the model predicts the correct categories. This is a value between 0 and 1, representing the percentage of test cases that were correctly categorized.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
MLP_Model = sklearn.neural_network.MLPClassifier(random_state = rand_state).fit(X_train, Y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
MLP_probability_predictions = MLP_Model.predict_proba(X_test)
print(MLP_probability_predictions)
# iii. Report on the score for the Neural Network, what does the score measure?
MLP_Score = MLP_Model.score(X_test,Y_test)
print(MLP_Score)
# iv: Experiment with different options for the neural network, report on your best configuration
MLP_High_Alpha = sklearn.neural_network.MLPClassifier(random_state = rand_state, alpha = 0.1).fit(X_train, Y_train)
alpha_score = MLP_High_Alpha.score(X_test,Y_test)
print(alpha_score)
MLP_Low_Learning_Rate = sklearn.neural_network.MLPClassifier(random_state = rand_state, learning_rate_init=0.0001).fit(X_train, Y_train)
alpha_score = MLP_Low_Learning_Rate.score(X_test,Y_test)
print(alpha_score)
MLP_High_Learning_Rate = sklearn.neural_network.MLPClassifier(random_state = rand_state, learning_rate_init=0.01).fit(X_train, Y_train)
learning_rate_score = MLP_High_Learning_Rate.score(X_test,Y_test)
print(learning_rate_score)
```

What does the score measure? 

The score measures the mean accuracy of how well the model predicts the correct categories. This is a value between 0 and 1, representing the percentage of test cases that were correctly categorized.

All configurations performed similarly, except the one with a slower learning rate, which performed very poorly.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
KNNmodel = sklearn.neighbors.KNeighborsClassifier().fit(X_train, Y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
KNN_probability_predictions = KNNmodel.predict_proba(X_test)
print(KNN_probability_predictions)
# iii. Report on the score for kNN, what does the score measure?
KNN_Score = KNNmodel.score(X_test,Y_test)
print(KNN_Score)
```
