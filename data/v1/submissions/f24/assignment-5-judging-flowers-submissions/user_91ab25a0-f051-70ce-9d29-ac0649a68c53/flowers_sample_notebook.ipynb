```python
# Imports and pip installations (if needed)
import pandas as pd
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
flower_data = pd.read_csv("iris.csv")
# Output the first 15 rows of the data
print(flower_data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(flower_data.info())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The data is using different features such as different measurements of part of the flower to label what species of Iris these plants are. The features are the sepal_length, sepal_width, petal_length, petal_width and the labels are the values in the species column. The mapping of the the labels to the actual classes would just be the species of the Iris where each class is just the label added to Iris, so one class would just be Iris Setosa as shown in the first 15 of the dataset.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
labels = flower_data.iloc[:,-1]

features = flower_data.iloc[:,:-1]

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train , X_test , y_train, y_test = train_test_split(features, labels, test_size=0.4, train_size=0.6, random_state=16)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
LRmodel = LogisticRegression()
LRmodel.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
LRpredict = LRmodel.predict_proba(X_test)
print(LRpredict)
# iii. Report on the score for Logistic regression model, what does the score measure?
LRscore = LRmodel.score(X_test, y_test)
print(LRscore)
# iv. Extract the coefficents and intercepts for the boundary line(s)
boundaryC = LRmodel.coef_
print(boundaryC)
boundaryI = LRmodel.intercept_
print(boundaryI)
```

The score represents that there is about a 95% mean accuracy in the models predictions

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
SVMmodel = svm.SVC(probability=True)
SVMmodel.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
SVCpredict = SVMmodel.predict_proba(X_test)
print(SVCpredict)
# iii. Report on the score for the SVM, what does the score measure?
SVMscore = SVMmodel.score(X_test, y_test)
print(SVMscore)
```

The score represents that there is about a 95% mean accuracy in the models predictions

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
NNmodel = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=6)
NNmodel.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
NNpredict = NNmodel.predict_proba(X_test)
print(NNpredict)
# iii. Report on the score for the Neural Network, what does the score measure?
NNScore = NNmodel.score(X_test, y_test)
print(NNScore)
# iv: Experiment with different options for the neural network, report on your best configuration
```

The score represents that there is about a 98.3% mean accuracy in the models predictions

After trying out multiple configurations, it seems like solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=6 is what was the most accurate for me. the random state went up as it approached 6-8 and after 8 went back down, the hidden_layer worked best at 5,2 and the alpha worked best for anything what was 1e-x where x was greater than 5

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
KNNmodel = KNeighborsClassifier(n_neighbors=3)
KNNmodel.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
KNNpredict = KNNmodel.predict_proba(X_test)
print(KNNpredict)
# iii. Report on the score for kNN, what does the score measure?
KNNscore = KNNmodel.score(X_test, y_test)
print(KNNscore)
```

The score represents that there is about a 96.7% mean accuracy in the models predictions
