```python
# Imports and pip installations (if needed)
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier


import warnings
warnings.filterwarnings("ignore")
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
df = pd.read_csv("iris.csv")

# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

print(df.head(15))
print(df.describe())
```

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = df[["sepal_length", "sepal_width", "petal_length", "petal_width"]]
y = df["species"]

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=69)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
model = LogisticRegression()
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
y_pred = model.predict_proba(X_test[0:1])
print("prediction probabilities: ", [(model.classes_[i], y_pred[0][i]) for i in range(len(model.classes_))]) # pair each class with the corresponding probability
print("true label: ", y_test.iloc[0])

# iii. Report on the score for Logistic regression model, what does the score measure?
#score measures fraction of correct classifications
print("accuracy: ", model.score(X_test, y_test))

# iv. Extract the coefficents and intercepts for the boundary line(s)
coef = model.coef_
intercept = model.intercept_
print("coefficients: ", coef)
print("intercepts: ", intercept)
```

Part iii:

Score measures fraction of correct classifications.

Part iv:

There are 3 boundary lines; one for each classification type.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
model = SVC(probability=True)
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
y_pred = model.predict_proba(X_test[0:1])
print("prediction probabilities: ", [(model.classes_[i], y_pred[0][i]) for i in range(len(model.classes_))])
print("true label: ", y_test.iloc[0])

# iii. Report on the score for the SVM, what does the score measure?
#score measures fraction of correct classifications
print("accuracy: ", model.score(X_test, y_test))
```

Part iii:

Score measures fraction of correct classifications.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
model = MLPClassifier(hidden_layer_sizes=(10,), max_iter=1000, random_state=1, activation="logistic", solver="adam")
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
y_pred = model.predict_proba(X_test[0:1])
print("prediction probabilities: ", [(model.classes_[i], y_pred[0][i]) for i in range(len(model.classes_))])
print("true label: ", y_test.iloc[0])

# iii. Report on the score for the Neural Network, what does the score measure?
print("accuracy: ", model.score(X_test, y_test))

# iv: Experiment with different options for the neural network, report on your best configuration
activations = ["relu", "logistic", "identity", "tanh"]
num_sizes = 4
hidden_sizes = np.logspace(0,num_sizes - 1,num_sizes,base=2, dtype=int)

#note: it seems fairly trivial to get 1.0 accuracy with most configurations with high hidden layer size
print("\n-------------")
for activation in activations:
    print(f'Models using {activation} activation:')
    for size in hidden_sizes:
        model = MLPClassifier(hidden_layer_sizes=(size,), max_iter=1000, random_state=1, activation=activation, solver="adam")
        model.fit(X_train, y_train)
        acc = model.score(X_test, y_test)
        print(f'NN with hidden layer of size {size}', "accuracy: ", acc)
```

Part iii:

Score measures fraction of correct classifications.

Part iv: 

I tested a few different hyperparameters: the activation function (relu, logistic, identity, and tanh), as well as the number of nodes in the hidden layer. All models only had 1 hidden layer.

Although all the models were able to reach a perfect 1.0 accuracy, only the tanh activation was able to reach it with a hidden size of 2. With the Occam's Razor principle (simpler models of equal performance are better), the best configuration is the tanh activation function with hidden size of 2.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
k = 5  # Number of neighbors
model = KNeighborsClassifier(n_neighbors=k)
model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
y_pred = model.predict_proba(X_test[0:1])
print(y_pred)
print("prediction probabilities: ", [(model.classes_[i], y_pred[0][i]) for i in range(len(model.classes_))])


# iii. Report on the score for kNN, what does the score measure?
#score measures fraction of correct classifications
print("accuracy: ", model.score(X_test, y_test))
```
