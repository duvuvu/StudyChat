```python
# Imports and pip installations (if needed)
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
iris_df = pd.read_csv('iris.csv')
# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)
print(iris_df.head(15))

iris_df.info()
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

In our data we have columns named Sepal_length, Sepal_width, Petal_Length, Petal_width, and Species. The first 4 are our features that then map to a specific species. We can use this information to train our model to see what features are used to classify flowers.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = iris_df.iloc[:, :-1]
y = iris_df.iloc[:, -1]
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_point = X_test.iloc[0]
sample_point_reshape = sample_point.values.reshape(1, -1)
probabilities = logistic_model.predict_proba(sample_point_reshape)
#First is Setosa, Second is Versicolor, Third is Virginica
print("Probabilities: ",probabilities)
# iii. Report on the score for Logistic regression model, what does the score measure?
logistic_score = logistic_model.score(X_test, y_test)
print("Model Score: ", logistic_score)
#The score measures how effectively  Logistic Regression can predict the species of iris flowers based on the features. A high accuracy suggests that the model performs well on the test data, while a low score may indicate inadequate performance.
#The score measures the accuracy of the predictions, meaning the number of correct predicitions divided by the total number of predictions. This model's score is a 1 which means that it predicted it 100% correctly.

# iv. Extract the coefficents and intercepts for the boundary line(s)
print("Coefficients: ",logistic_model.coef_)
print("Intercept", logistic_model.intercept_)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svc_model = SVC(probability=True)
svc_model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_point = X_test.iloc[0]
sample_point_reshaped = sample_point.values.reshape(1, -1)
probabilities = svc_model.predict_proba(sample_point_reshaped)
print("Probabilities: ", probabilities)
# iii. Report on the score for the SVM, what does the score measure?
svc_score = svc_model.score(X_test, y_test)
print("Model Score: ", svc_score)
#The score measures how effectively the SVC can predict the species of iris flowers based on the features. A high accuracy suggests that the model performs well on the test data, while a low score may indicate inadequate performance.
#The score measures the accuracy of the predictions, meaning the number of correct predicitions divided by the total number of predictions. This model's score is a 1 which means that it predicted it 100% correctly.
```

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set
mlp_model = MLPClassifier()
mlp_model.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_point = X_test.iloc[0]
sample_point_reshape = sample_point.values.reshape(1, -1)
probabilities = mlp_model.predict_proba(sample_point_reshape)
print("Probabilities: ", probabilities)

# iii. Report on the score for the Neural Network, what does the score measure?
mlp_score = mlp_model.score(X_test, y_test)
print("Model Score: ", mlp_score)
#The score measures how effectively the NN  can predict the species of iris flowers based on the features. A high accuracy suggests that the model performs well on the test data, while a low score may indicate inadequate performance.
#The score measures the accuracy of the predictions, meaning the number of correct predicitions divided by the total number of predictions. This model's score is a 1 which means that it predicted it 100% correctly.

# iv: Experiment with different options for the neural network, report on your best configuration
from sklearn.model_selection import GridSearchCV
param_grid = {
    'hidden_layer_sizes': [(5,), (10,), (5, 5), (10, 5)],  # Different layer configurations
    'activation': ['relu', 'tanh', 'logistic'],  # Different activation functions
    'alpha': [0.0001, 0.001, 0.01],  # Different regularization strengths
    'max_iter': [500, 1000]  # Number of iterations
}

grid_search = GridSearchCV(MLPClassifier(), param_grid, cv=5, n_jobs=-1) 
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
best_params = grid_search.best_params_
best_score = grid_search.best_score_


print("Grid Search Results:")
# Display all results
results = grid_search.cv_results_

for mean_score, params in zip(results['mean_test_score'], results['params']):
    print(f"Mean Test Score: {mean_score:.4f} | Parameters: {params}")

print("\nBest configuration:")
print(f"Best parameters: {best_params}")
print(f"Best cross-validated accuracy: {best_score:.4f}")

# Evaluate on the test set
test_score = best_model.score(X_test, y_test)
print(f"Test set accuracy for best model: {test_score:.4f}")
```

The Best Configuration ended up being utilizing a ReLu activation function. Using an alpha of .001 which is a penalty applied to larger weights to help prevent overfitting.  This means not much regularization was needed to acheive good results. We utilized 2 hidden layers, first one with 10 and the second one with 5. This allowed us to get a cross validated accuracy of 97.5% which means that the model correctly predicted the species of the flowers 97.5% of the time ocress different subsets of training data. We achieved a test accuracy of 96.67% which meant that the model correctly predicted 96.67% of the species.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

k_neighbors_model = KNeighborsClassifier(n_neighbors=10)
k_neighbors_model.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_point = X_test.iloc[0]
sample_point_reshape = sample_point.values.reshape(1, -1)
probabilities = k_neighbors_model.predict_proba(sample_point_reshape)
print("Probabilities: ", probabilities)

# iii. Report on the score for kNN, what does the score measure?
knn_score = k_neighbors_model.score(X_test, y_test)
print("Model Score: ", knn_score)
#The score measures how effectively the k-NN model can predict the species of iris flowers based on the features. A high accuracy suggests that the model performs well on the test data, while a low score may indicate inadequate performance.
#The score measures the accuracy of the predictions, meaning the number of correct predicitions divided by the total number of predictions. This model's score is a 1 which means that it predicted it 100% correctly.
```
