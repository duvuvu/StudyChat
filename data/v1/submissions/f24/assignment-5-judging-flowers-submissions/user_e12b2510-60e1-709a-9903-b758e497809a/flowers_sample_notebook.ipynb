```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
data = pd.read_csv(url)
# Output the first 15 rows of the data
print(data.head(15))
# Display a summary of the table information (number of datapoints, etc.)
print(data.info())
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

The dataset contains 150 entries with 5 cols, where 4 cols are numerical features used to describe the iris flower and the other column is the categorical label indicating the species of each flower.
Features: sepal_length, sepal_width, petal_length, petal_width

Label: species (setosa, versicolor, virginica)

The mapping of the labels to the actual classe is 0: setosa, 1:versicolor, 2: virginica.

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
X = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = data['species']
# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set
log_reg = LogisticRegression(max_iter=200)
log_reg.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame(X_test.iloc[0].values.reshape(1, -1), columns=X.columns)
probabilities = log_reg.predict_proba(sample_datapoint)
print("Probabilities:", probabilities)
# iii. Report on the score for Logistic regression model, what does the score measure?
score = log_reg.score(X_test, y_test)
print("Logistic Regression Score:", score)
# iv. Extract the coefficents and intercepts for the boundary line(s)
coefficients = log_reg.coef_
intercept = log_reg.intercept_

print("Coefficients:", coefficients)
print("Intercept:", intercept)
```

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set
svm = SVC(probability=True)
svm.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
svm_probabilities = svm.predict_proba(sample_datapoint)
print("SVM Probabilities:", svm_probabilities)
# iii. Report on the score for the SVM, what does the score measure?
svm_score = svm.score(X_test, y_test)
print("SVM Score:", svm_score)
```

SVM achieved a score of 1, which means perfect accuracy on the test dataset.So the model has successfully distinguished between the different species of iris flowers without any errors.

# Part 5: Neural Network

```python
# i. Train a Neural Network (MLP Classifier) on the training set
nn = MLPClassifier(hidden_layer_sizes=(20,), activation='logistic', alpha=0.001, max_iter=3000, learning_rate_init=0.001, random_state=42)
nn.fit(X_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
sample_datapoint = pd.DataFrame(X_test.iloc[0].values.reshape(1, -1), columns=X.columns)
nn_probabilities = nn.predict_proba(sample_datapoint)
print("Neural Network Probabilities:", nn_probabilities)

# iii. Report on the score for the Neural Network
nn_score = nn.score(X_test, y_test)
print("Neural Network Score:", nn_score)

# iv. Experiment with different options for the neural network
configs = [
    {"hidden_layer_sizes": (10,), "activation": "relu", "alpha": 0.001},
    {"hidden_layer_sizes": (10, 10), "activation": "tanh", "alpha": 0.01},
    {"hidden_layer_sizes": (20,), "activation": "logistic", "alpha": 0.001},
]

best_score = 0
best_config = None

for config in configs:
    nn = MLPClassifier(hidden_layer_sizes=config["hidden_layer_sizes"], 
                       activation=config["activation"], 
                       alpha=config["alpha"], 
                       max_iter=3000, 
                       random_state=42)
    nn.fit(X_train, y_train)
    score = nn.score(X_test, y_test)
    print(f"Configuration: {config}, Score: {score}")
    
    if score > best_score:
        best_score = score
        best_config = config

print("Best Configuration:", best_config)
print("Best Score:", best_score)
```

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
# ii. For a sample datapoint, predict the probabilities for each possible class
knn_probabilities = knn.predict_proba(sample_datapoint)
print("KNN Probabilities:", knn_probabilities)
# iii. Report on the score for kNN, what does the score measure?
knn_score = knn.score(X_test, y_test)
print("KNN Score:", knn_score)
```
