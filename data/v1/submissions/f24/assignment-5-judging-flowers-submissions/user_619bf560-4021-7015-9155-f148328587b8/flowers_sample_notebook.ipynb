```python
# Imports and pip installations (if needed)
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier
from sklearn.neighbors import KNeighborsClassifier
```

# Part 1: Load the dataset

```python
# Load the dataset (load remotely or locally)

iris_data = pd.read_csv('iris.csv')

# Output the first 15 rows of the data
# Display a summary of the table information (number of datapoints, etc.)

print(iris_data[0:15])
iris_data.info()
```

## About the dataset
Explain what the data is in your own words. What are your features and labels? What is the mapping of your labels to the actual classes?

This data is on the relationship between sepal and petal sizes with the plant species. The features are the sepal_length, sepal_width, petal_length, and petal_width. The label is the species column. The labels setosa, versicolor, and virginica directly map to the three classes of flowers shown in the readme. 

# Part 2: Split the dataset into train and test

```python
# Take the dataset and split it into our features (X) and label (y)
feature_cols = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
features = iris_data[feature_cols]
label = iris_data['species']

# Use sklearn to split the features and labels into a training/test set. (90% train, 10% test)
x_train, x_test, y_train, y_test = train_test_split(features, label, test_size=0.1, random_state=42)
```

# Part 3: Logistic Regression

```python
# i. Use sklearn to train a LogisticRegression model on the training set

regression = LogisticRegression(random_state=42).fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class

sample_data_point = pd.DataFrame([[4.0, 4.0, 3.0, 1.0]], columns=feature_cols)
print(f'Class probability order: {regression.classes_}')
print(f'Probabilities: {regression.predict_proba(sample_data_point)}')

# iii. Report on the score for Logistic regression model, what does the score measure?

print(f'Score: {regression.score(x_test, y_test)}')

# iv. Extract the coefficents and intercepts for the boundary line(s)
print(f'Intercepts: {regression.intercept_}')
print(f'Coefficients: {regression.coef_}')

for i in range(0, 3):
    print(f'Boundary for {regression.classes_[i]}:')
    print(f'{regression.intercept_[i]}', end='')
    for j in range(0, 4):
        print(f' + {regression.coef_[i][j]} * x_{j+1}')

    print()
```

iii. The score represents the average of the accuracy for each of the datapoints in the test data. In this case, we can see that on average, the model correctly classified the test data points 100% of the time.

# Part 4: Support Vector Machine

```python
# i. Use sklearn to train a Support Vector Classifier on the training set

vector_classifier = SVC(probability=True, random_state=42).fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class

print(f'SVC Class Order: {vector_classifier.classes_}')
print(f'SVC Probabilities: {vector_classifier.predict_proba(sample_data_point)}')

# iii. Report on the score for the SVM, what does the score measure?
print(f'Score: {vector_classifier.score(x_test, y_test)}')
```

iii. The score again measures the average accuracy over all of the datapoints in the test set. In this case, it was again 100% accurate at classifying each of the test data points.

# Part 5: Neural Network

```python
# i. Use sklearn to train a Neural Network (MLP Classifier) on the training set

mlp_classifier = MLPClassifier(random_state=42, max_iter=500).fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class

print(f'Class Order: {mlp_classifier.classes_}')
print(f'Probabilities: {mlp_classifier.predict_proba(sample_data_point)}')

# iii. Report on the score for the Neural Network, what does the score measure?

print(f'Score: {mlp_classifier.score(x_test, y_test)}')

# iv: Experiment with different options for the neural network, report on your best configuration
print(f'10 hidden layers (instead of 100): {MLPClassifier(hidden_layer_sizes=10, max_iter=500, random_state=42).fit(x_train, y_train).score(x_test, y_test)}')
print(f'lbfgs instead of gradient descent: {MLPClassifier(solver='lbfgs', max_iter=500, random_state=42).fit(x_train, y_train).score(x_test, y_test)}')
print(f'High learning rate alpha=1: {MLPClassifier(alpha=1, max_iter=500, random_state=42).fit(x_train, y_train).score(x_test, y_test)}')
```

iv: Out of the four configurations I tried, the only one that was not completely accurate on the test data was the MLP with only 10 hidden layers. That model only predicted the test data on average 86.67% of the time.

# Part 6: K-Nearest Neighbors

```python
# i. Use sklearn to 'train' a k-Neighbors Classifier
# Note: KNN is a nonparametric model and technically doesn't require training
# fit will essentially load the data into the model see link below for more information
# https://stats.stackexchange.com/questions/349842/why-do-we-need-to-fit-a-k-nearest-neighbors-classifier

kn_classifier = KNeighborsClassifier().fit(x_train, y_train)

# ii. For a sample datapoint, predict the probabilities for each possible class
print(f'Class Order: {kn_classifier.classes_}')
print(f'Probabilities: {kn_classifier.predict_proba(sample_data_point)}')

# iii. Report on the score for kNN, what does the score measure?
print(f'Score: {kn_classifier.score(x_test, y_test)}')
```

iii. Once again, the score measures the average accuracy for the model's predictions on the test data. It was once again 100%.
